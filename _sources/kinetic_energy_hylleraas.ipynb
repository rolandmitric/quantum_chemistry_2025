## Hylleraas-Methode: Kinetische Energie

Der Operator für die kinetische Energie zweier Elektronen lautet

$$
\hat T = -\frac{1}{2}\big(\nabla_1^2 + \nabla_2^2\big).
$$

Wir wollen nun den Erwartungswert der kinetischen Energie für den Hylleraas-Ansatz berechnen. Unter der Annahme, dass die Wellenfunktion im Unendlichen, kann der Erwartungswert von $\hat T$ durch Integration der ersten Ableitungen geschrieben werden:

$$
\langle T\rangle
=\frac{1}{2}\int |\nabla\Psi|^2\,d^3r_1\,d^3r_2.
$$

Dieser Ausdruck soll nun in die Hylleraas-Koordinaten transformiert werden. 

Wir betrachten dazu zwei Elektronen mit den kartesischen Koordinaten $(x_1,y_1,z_1)$ und $(x_2,y_2,z_2)$. Die radialen Abstände sind definiert als  

$$
r_1 = \sqrt{x_1^2+y_1^2+z_1^2}, \qquad
r_2 = \sqrt{x_2^2+y_2^2+z_2^2}, \qquad
r_{12} = \sqrt{(x_2-x_1)^2+(y_2-y_1)^2+(z_2-z_1)^2}.
$$

Wir erinnern uns an die Definition von Hylleraas-Koordinaten

$$
s = r_1 + r_2, \qquad
t = r_1 - r_2, \qquad
u = r_{12}
$$


Da der Hylleraas-Ansatz nur $\Psi$ nur von $s,t,u$ abhängt, gilt für jede kartesische Ableitung die Kettenregel. So erhalten wir zum Beispiel:

$$
\frac{\partial \Psi}{\partial x_1}
=\frac{\partial \Psi}{\partial s}\frac{\partial s}{\partial x_1}
+\frac{\partial \Psi}{\partial t}\frac{\partial t}{\partial x_1}
+\frac{\partial \Psi}{\partial u}\frac{\partial u}{\partial x_1}.
$$

Die analogen Beziehungen gelten auch für die Ableitungen nach $y_1,z_1,x_2,y_2,z_2$. Zunächst berechnen wir die Ableitungen der Koordinaten $r_1,r_2,r_{12}$ nach den kartesischen Komponenten:

$$
\frac{\partial r_1}{\partial x_1} = \frac{x_1}{r_1}, \qquad
\frac{\partial r_1}{\partial y_1} = \frac{y_1}{r_1}, \qquad
\frac{\partial r_1}{\partial z_1} = \frac{z_1}{r_1},
$$

$$
\frac{\partial r_2}{\partial x_2} = \frac{x_2}{r_2}, \qquad
\frac{\partial r_2}{\partial y_2} = \frac{y_2}{r_2}, \qquad
\frac{\partial r_2}{\partial z_2} = \frac{z_2}{r_2},
$$

und

$$
\frac{\partial r_{12}}{\partial x_1} = -\frac{x_2-x_1}{r_{12}}, \qquad
\frac{\partial r_{12}}{\partial x_2} =  \frac{x_2-x_1}{r_{12}},
$$

mit analogen Ausdrücken für $y$ und $z$.

Da $s=r_1+r_2$, $t=r_1-r_2$, $u=r_{12}$, ergibt sich:

$$
\frac{\partial s}{\partial x_1} = \frac{x_1}{r_1}, \qquad
\frac{\partial t}{\partial x_1} = \frac{x_1}{r_1}, \qquad
\frac{\partial u}{\partial x_1} = -\frac{x_2-x_1}{r_{12}},
$$

$$
\frac{\partial s}{\partial x_2} = \frac{x_2}{r_2}, \qquad
\frac{\partial t}{\partial x_2} = -\frac{x_2}{r_2}, \qquad
\frac{\partial u}{\partial x_2} = \frac{x_2-x_1}{r_{12}}.
$$

Damit können wir alle ersten Ableitungen der Wellenfunktion nach den kartesischen Variablen ausdrücken:

$$
\frac{\partial \Psi}{\partial x_1}
= \left(\frac{x_1}{r_1}\right)(\Psi_s+\Psi_t)
-\frac{x_2-x_1}{r_{12}}\Psi_u,
$$

$$
\frac{\partial \Psi}{\partial x_2}
= \left(\frac{x_2}{r_2}\right)(\Psi_s-\Psi_t)
+\frac{x_2-x_1}{r_{12}}\Psi_u,
$$

und analog für $y$ und $z$. Die Norm des gesamten Gradienten, die wir für die Berechnung der kinetischen Energie benötigen lautet:

$$
|\nabla\Psi|^2 =
\left(\frac{\partial \Psi}{\partial x_1}\right)^2
+\left(\frac{\partial \Psi}{\partial y_1}\right)^2
+\left(\frac{\partial \Psi}{\partial z_1}\right)^2
+\left(\frac{\partial \Psi}{\partial x_2}\right)^2
+\left(\frac{\partial \Psi}{\partial y_2}\right)^2
+\left(\frac{\partial \Psi}{\partial z_2}\right)^2.
$$

Da die Formeln für $x,y,z$ identisch sind, kann man alle Komponenten durch Produkte wie

$$
x_1x_2+y_1y_2+z_1z_2 = r_1r_2\cos\theta
$$

zusammenfassen, wobei $\theta$ der Winkel zwischen den beiden Ortsvektoren ist und  

$$
\cos\theta=\frac{r_1^2+r_2^2-u^2}{2r_1r_2}.
$$

Nach Einsetzen von

$$
r_1=\frac{s+t}{2}, \qquad r_2=\frac{s-t}{2}, \qquad r_{12}=u,
$$

erhält man schließlich

$$
|\nabla\Psi|^2
= 2(\Psi_s^2+\Psi_t^2+\Psi_u^2)
+\frac{4}{u}\frac{s(u^2-t^2)}{s^2-t^2}\Psi_s\Psi_u
+\frac{4}{u}\frac{t(s^2-u^2)}{s^2-t^2}\Psi_t\Psi_u.
$$


Damit können wir nun den Erwartungswert der kinetischen Energie berechnen. Wir verwenden dazu das Volumenelement in Hylleraas-Koordinaten:

$$
d^3r_1\,d^3r_2
= 2\pi^2\,u\,(s^2-t^2)\,ds\,dt\,du,
$$

wobei die Integrationsgrenzen durch die Dreiecksungleichungen gegeben sind:

$$
0\le u \le s, \qquad -u\le t \le u.
$$

Damit ergibt sich der Erwartungswert der kinetischen Energie als

$$
\langle T\rangle
= \pi^2
\int_{0}^{\infty}\!ds
\int_{0}^{s}\!du
\int_{-u}^{u}\!dt\;
u\,(s^2-t^2)
\Biggl[
2(\Psi_s^2+\Psi_t^2+\Psi_u^2)
+\frac{4}{u}\frac{s(u^2-t^2)}{s^2-t^2}\Psi_s\Psi_u
+\frac{4}{u}\frac{t(s^2-u^2)}{s^2-t^2}\Psi_t\Psi_u
\Biggr].
$$

Dieser Ausdruck erlaubt die direkte Berechnung der kinetischen Energie für beliebige Hylleraas-Basisfunktionen $\Psi(s,t,u)$ nur durch erste Ableitungen nach $s,t,u$.



## Sympy-Programm zur symbolischen Transformation des Gradienten

```python
import sympy as sp

x1, y1, z1 = sp.symbols('x1 y1 z1', real = True)
x2, y2, z2 = sp.symbols('x2 y2 z2', real = True)
r1_vec = sp.Matrix([x1,y1,z1])
r2_vec = sp.Matrix([x2,y2,z2])
r12_vec = r2_vec - r1_vec

r1 = r1_vec.norm()
r2 = r2_vec.norm()
r12 = r12_vec.norm()

s = r1 + r2
t = r1 - r2
u = r12

var_new = sp.Matrix([s,t,u])
var_old = sp.Matrix([x1, y1, z1, x2, y2, z2])
J = var_new.jacobian(var_old)

D_s,D_t,D_u = sp.symbols('D_s D_t D_u',real = True)
D_stu = sp.Matrix([D_s,D_t,D_u])


K  = J @J.T
D2 = (D_stu.T @ K @D_stu)[0].simplify()


R1 = sp.symbols('r1', real = True)
R2 = sp.symbols('r2', real = True)
R12 = sp.symbols('r12', real = True)
subsdict = {r1:R1,
            r2:R2,
            r12:R12,
            r1**2:R1**2,
            r2**2:R2**2,
            r12**2:R12**2,
            r1_vec.dot(r2_vec):(R12**2-R1**2-R2**2)/(2*R1*R2)
           }
D2 =D2.subs(subsdict)
D2 = sp.collect(D2.expand(),(D_s**2, D_u**2, D_t**2, D_s*D_t, D_s*D_u,D_t*D_u))
D2 = D2.simplify()
D2 =D2.subs(subsdict)

s,t,u = sp.symbols('s t u', real = True)
subsdict = {R1: (s+t)/2, R2:(s-t)/2, R12:u}
D2 = D2.subs(subsdict).simplify()
```

```python
D2 = sp.collect(D2,(D_s**2, D_u**2, D_t**2, D_s*D_t, D_s*D_u,D_t*D_u))
```

```python
D2 = sp.factor_terms(D2)
```

```python
D2.cancel()
```
