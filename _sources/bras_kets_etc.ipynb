## Definition: Skalarprodukt

Sei $ V $ ein komplexer Vektorraum. Ein **Skalarprodukt** auf $ V $ ist eine Abbildung

$$
\langle \,\cdot\,\mid\,\cdot\,\rangle : V \times V \rightarrow \mathbb{C},
$$

**(1) Hermitizit√§t:**  
    
$$ \langle\psi\mid\phi\rangle = \langle\phi\mid\psi\rangle^\ast $$

**(2) Linearit√§t im zweiten Argument:**  
    
$$ \langle\psi\mid(a\,\phi + \chi)\rangle = a\,\langle\psi\mid\phi\rangle + \langle\psi\mid\chi\rangle $$

**(3) Homogenit√§t im zweiten Argument:**  
$$ \langle\psi\mid(a\,\phi)\rangle = a\,\langle\psi\mid\phi\rangle $$

**(4) Positive Definitheit:**  
    
$$ \langle\psi\mid\psi\rangle \ge 0 $$ und  
    
$$ \langle\psi\mid\psi\rangle = 0 \Longleftrightarrow \lvert\psi\rangle = \lvert 0\rangle $$

üìù *Hinweis:* In der Mathematik wird die Linearit√§t h√§ufig im ersten Argument definiert. In der Physik ist es jedoch √ºblich, sie (wie hier) im zweiten Argument einzuf√ºhren. Ein Vektorraum, der mit einem Skalarprodukt ausgestattet ist, hei√üt **unit√§rer Raum**. 
</div>

Mit dieser Definition k√∂nnen den Vektorraum der inneren Zust√§nde eines Elektrons zu einem unit√§ren Raum erweitern indem wir ein Skalarprodukt wie folgt einf√ºhren. Betrachten wir zwei beliebige Vektoren, die durch folgende Spaltenvektoren dargestellt werden k√∂nnen:

\end{array}\right]\\

\end{array}\right] $$

Wir *definieren*  deren Skalarprodukt als :

$$ \langle\phi|\psi\rangle=d_{\alpha}^{*}c_{\alpha}+d_{\beta}^{*}c_{\beta} $$

Diese Definition erf√ºllt alle Definitionseigenschaften eines Skalarprodukts wie man direkt √ºberpr√ºfen kann:

|\psi\rangle \rightarrow\left[\begin{array}{l}

\end{array}\right]
$$

\begin{aligned}

Dies ist aber immer positiv und erf√ºllt damit die Definitionseigenschaft 1. Es gilt also

$$

Des Weiteren ist $\langle\psi \mid \psi\rangle=0$ √§quivalent mit $c_\alpha=c_\beta=0$, womit auch die Eigenschaft 2 erf√ºllt ist. Betrachten wir einen beliebigen, aber fixierten Vektor $|\phi\rangle \in V$. Mit Hilfe dieses Vektors definieren wir die Abbildung

Diese Zuordnung ordnet jedem Vektor $|\psi\rangle$ die Zahl zu, die durch das Skalarprodukt mit $|\phi\rangle$ entsteht. Wir wollen nun zeigen, dass $F_{|\phi\rangle}$ eine **lineare Abbildung** ist, d.\,h. dass **Additivit√§t** und **Homogenit√§t** erf√ºllt sind.

Zun√§chst zeigen wir die **Additivit√§t**. F√ºr beliebige $|\psi\rangle, |\chi\rangle \in V$ gilt

### *Definition: Dualraum*

### Riesz‚Äôscher Darstellungssatz (endlichdimensionale Version)

Sei $(V,\langle \cdot \mid \cdot \rangle)$ ein endlichdimensionaler unit√§rer Raum. Dann gilt:

> **Zu jedem linearen Funktional $F \in V^\ast$ existiert genau ein Vektor $|u\rangle \in V$, so dass**

$$

\qquad \text{f√ºr alle } |v\rangle \in V.
$$

Der Vektor $|u\rangle$ hei√üt die **Riesz-Darstellung** des Funktionals $F$.

###  Beweis

Sei $\{\,|e_1\rangle,\dots,|e_n\rangle\,\}$ eine Orthonormalbasis von $V$.  
Jeder Vektor $|v\rangle \in V$ l√§sst sich eindeutig als Linearkombination dieser Basisvektoren schreiben:

$$
|v\rangle = \sum_{k=1}^n v_k\,|e_k\rangle, 
\qquad v_k = \langle e_k \mid v \rangle.
$$

Wendet man das gegebene lineare Funktional $F$ auf $|v\rangle$ an, so erh√§lt man wegen der Linearit√§t von $F$:

$$
F(|v\rangle)
= F\!\left(\sum_{k=1}^n v_k\,|e_k\rangle\right)
= \sum_{k=1}^n v_k\,F(|e_k\rangle).
$$

Nun wollen wir $F(|v\rangle)$ in der Form eines Skalarprodukts $\langle u \mid v \rangle$ schreiben. Dazu definieren wir den folgenden Vektor:

$$
|u\rangle = \sum_{k=1}^n F(|e_k\rangle)^\ast\,|e_k\rangle.
$$

Damit folgt f√ºr jedes $|v\rangle$:
$$
\langle u \mid v \rangle
= \sum_{k=1}^n u_k^\ast\,v_k
= \sum_{k=1}^n F(|e_k\rangle)\,v_k
= F(|v\rangle).
$$

Damit haben wir gezeigt, dass es zu jedem linearen Funktional $F$ tats√§chlich einen Vektor $|u\rangle$ gibt, so dass

$$
F(|v\rangle) = \langle u \mid v \rangle 
\qquad \text{f√ºr alle } |v\rangle \in V.
$$

### Eindeutigkeit

Die **Eindeutigkeit** von $|u\rangle$ folgt aus der Nichtentartung des Skalarprodukts.  
Angenommen, es g√§be zwei Vektoren $|u\rangle$ und $|u'\rangle$ in $V$ mit

$$
\langle u \mid v \rangle = F(|v\rangle) = \langle u' \mid v \rangle 
\qquad \text{f√ºr alle } |v\rangle \in V.
$$

Dann gilt f√ºr alle $|v\rangle$:
$$
\langle u - u' \mid v \rangle = 0.
$$
Insbesondere folgt f√ºr $|v\rangle = |u - u'\rangle$:

$$
\langle u - u' \mid u - u' \rangle = 0.
$$

Da das Skalarprodukt positiv definit ist, folgt $|u - u'\rangle = 0$, also $|u\rangle = |u'\rangle$.  
Die Darstellung ist somit eindeutig.

Jedem Vektor $|u\rangle$ k√∂nnen wir ein lineares Funktional $F_{|u\rangle}$ zuordnen, und umgekehrt besitzt jedes lineare Funktional $F$ genau eine Riesz-Darstellung $|u\rangle$. Die Abbildung

$$
V \longrightarrow V^\ast, \qquad
|u\rangle \mapsto F_{|u\rangle}, 
\qquad F_{|u\rangle}(|v\rangle) = \langle u \mid v \rangle,
$$

ist damit ein Isomorphismus von Vektorr√§umen. In der Sprache der Physik motiviert der Rieszsche Satz die Einf√ºhrung der sog. Dirac-Notation.

### Dirac-Notation

Die Dirac-Notation trennt systematisch zwischen Vektoren und den zugeh√∂rigen linearen Funktionalen.  Ein **Ket** $|v\rangle$ bezeichnet dabei einen Vektor in $V$. Durch den Rieszschen Darstellungssatz entspricht jedem Ket ein eindeutig bestimmtes lineares Funktional, das man als **Bra** schreibt:

$$
\langle v| : V \rightarrow \mathbb{C}, 
\qquad
\langle v|\,|w\rangle = \langle v \mid w \rangle.
$$

Der Bra $\langle v|$ ist also nichts anderes als die Riesz-Darstellung des Funktionals 
$F_{|v\rangle}$. Die Bra‚ÄìKet-Schreibweise

Er erlaubt, jedem Vektor $|v\rangle$ eindeutig ein ‚Äûlinkswirkendes‚Äú lineares Funktional $\langle v|$ zuzuordnen und so die Dualit√§t zwischen Vektoren und Funktionalen in einer besonders eleganten und physikalisch intuitiven Form zu schreiben. Dies illustrieren wir nun am Beispiel des zweidimensionalen Vektroraumes mit dem wir die inneren Zust√§nde des Elektrons beschreiben.

Nachdem wir nun Kets als Vektoren, Bras als zugeh√∂rige Funktionale und damit die Dualit√§t zwischen $V$ und $V^\ast$ verstanden haben, wenden wir uns nun den **linearen Operatoren** zu, also den Abbildungen, die Vektoren wieder auf Vektoren abbilden und in der Quantenmechanik die Observablen repr√§sentieren.

### Definition: Linearer Operator

Ein **linearer Operator** auf einem komplexen Vektorraum $V$ ist eine Abbildung

$$
\hat A : V \longrightarrow V,
$$

die f√ºr alle $|v\rangle, |w\rangle \in V$ und alle $a,b \in \mathbb{C}$ die folgende Eigenschaft hat:

$$
\hat A\,(a|v\rangle + b|w\rangle)
= a\,\hat A|v\rangle + b\,\hat A|w\rangle.
$$


### Matrixdarstellung eines Operators in einer Orthonormalbasis

Sei nun $V$ zweidimensional mit der Orthonormalbasis $\{|\alpha\rangle,|\beta\rangle\}$.  
Ein Operator $\hat A$ ist vollst√§ndig bestimmt durch seine Wirkung auf die Basisvektoren:

$$

$$

Da diese wieder Vektoren in $V$ sind, k√∂nnen sie in der Basis $\{|\alpha\rangle,|\beta\rangle\}$ zerlegt werden:

$$
\hat A|\alpha\rangle = A_{\alpha\alpha}|\alpha\rangle + A_{\beta\alpha}|\beta\rangle,
$$

$$
\hat A|\beta\rangle = A_{\alpha\beta}|\alpha\rangle + A_{\beta\beta}|\beta\rangle.
$$

Die vier komplexen Zahlen, die bei der Zerlegung auftauchen

$$
A_{\alpha\alpha},\; A_{\beta\alpha},\; A_{\alpha\beta},\; A_{\beta\beta}
$$

In Matrixschreibweise entspricht dies der folgenden Darstellung:

$$
\hat A \;\longleftrightarrow\;
\begin{pmatrix}
A_{\alpha\alpha} & A_{\alpha\beta} \\
A_{\beta\alpha} & A_{\beta\beta}
\end{pmatrix}.
$$

F√ºr einen beliebigen Vektor

$$
|\psi\rangle = c_\alpha|\alpha\rangle + c_\beta|\beta\rangle
\quad \leftrightarrow \quad
\begin{bmatrix}
c_\alpha \\ c_\beta
\end{bmatrix}
$$

l√§sst sich die Wirkung des Operators auf den Vektor
durch gew√∂hnliche Matrixmultiplikation beschreiben:

$$

\quad\leftrightarrow\quad
\begin{pmatrix}
A_{\alpha\alpha} & A_{\alpha\beta} \\

\end{pmatrix}
\begin{bmatrix}
c_\alpha \\ c_\beta
\end{bmatrix}.
$$

Damit k√∂nnen wir jeden linearen Operator in einem endlichdimensionalen Raum direkt mit einer Matrix relativ zu einer gew√§hlten Orthonormalbasis identifizieren!

### Das adjungierte (hermitesche konjugierte) eines Operators

Um Observablen beschreiben zu k√∂nnen, ben√∂tigen wir den Begriff des **Adjungierten** eines Operators. F√ºr jeden Operator $\hat A$ definieren wir den adjungierten Operator $\hat A^\dagger$ durch die folgende Bedingung:

$$

$$

Der adjungierte Operator ist also derjenige, der die Wirkung von $\hat A$ ‚Äûauf das Bra umw√§lzt‚Äú. Die Matrixdarstellung dieses Operators erh√§lt man indem man die Matrix des Operators $\hat A$  **konjugiert und transponierten**:

$$
\hat A^\dagger \;\longleftrightarrow\;
\begin{pmatrix}
A_{\alpha\alpha} & A_{\alpha\beta} \\
A_{\beta\alpha} & A_{\beta\beta}
\end{pmatrix}^\dagger
=
\begin{pmatrix}
A_{\alpha\alpha}^\ast & A_{\beta\alpha}^\ast \\
A_{\alpha\beta}^\ast & A_{\beta\beta}^\ast
\end{pmatrix}.
$$

### Hermitesche Operatoren (Observablen)

Ein Operator $\hat A$ hei√üt **hermitesch** (oder selbstadjungiert),
$$

$$

In Matrixdarstellung bedeutet dies:

$$
A_{\alpha\alpha}=A_{\alpha\alpha}^\ast,\qquad
A_{\beta\beta}=A_{\beta\beta}^\ast,\qquad
A_{\beta\alpha}=A_{\alpha\beta}^\ast.
