# Magnetic dipole in a uniform magnetic 

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

ex = np.array([1,0,0])
ey = np.array([0,1,0])
ez = np.array([0,0,1])

R = 1.0  # Radius des Ringes

def loop_points(theta):
    n = -np.sin(theta)*ey + np.cos(theta)*ez
    t2 = np.cross(n, ex)

    phi = np.linspace(0, 2*np.pi, 300)
    r = (np.cos(phi)[:,None]*ex 
         + np.sin(phi)[:,None]*t2) * R
    return r

def plot_loop(theta_deg=0):
    theta = np.deg2rad(theta_deg)

    pts = loop_points(theta)

    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot(pts[:,0], pts[:,1], pts[:,2], 'b', lw=2)

    n = -np.sin(theta)*ey + np.cos(theta)*ez
    ax.quiver(0,0,0, n[0], n[1], n[2], color='r', lw=3)

    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    ax.set_zlim(-1.2, 1.2)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")

    plt.show()

interact(plot_loop, theta_deg=(0,180,1))
```
