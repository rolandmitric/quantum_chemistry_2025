# Das Pauli-Prinzip 

## Raum- und Spin-Zustände

Wir haben nun gelernt, wie man Spinzustände für ein Zweielektronensystem bildet. Diese müssen wir nun durch räumliche Zustände (Wellenfunktionen) erweitern, damit wir molekulare Systeme beschreiben können. Dazu führen wir den Begriff eines Spin-Orbitals ein.

Betrachten wir zunächst ein einzelnes Elektron, das sich in irgendeinem gegebenen Potential bewegt. Dies könnte z.B. ein Elektron im H-Atom oder im H$_2$-Molekül sein. Die räumlichen Zustände bilden einen Vektorraum, mit einer Basis, die z.B. aus den Eigenfunktionen des Hamiltonoperators gebildet werden kann.

Wie kombiniert man nun solche räumlichen Zustände mit den Spinzuständen?

$$
\begin{array}{c|c}
\text{Raum-Basis} & \text{Spin-Basis} \\
\hline\hline
|\psi_1\rangle & |\alpha\rangle \\
|\psi_2\rangle & |\beta\rangle \\
|\psi_3\rangle &  \\
\vdots &
\end{array}
$$


Da die Raum- und Spin-Zustände in zwei unterschiedlichen Vektorräumen leben, können wir sie zu einem neuen Vektorraum kombinieren, indem wir das Tensorprodukt bilden. Als Basis dieses Tensorproduktraums erhalten wir die sogenannten **Spinorbitale**

$$
|\psi_{i,\sigma}\rangle \equiv |\psi_i\rangle \otimes |\sigma\rangle ,
$$

wobei $|\sigma\rangle$ entweder dem $|\alpha\rangle$- oder dem $|\beta\rangle$-Spinzustand entspricht. Enthält der Hamiltonoperator keinen Spin – wie im nichtrelativistischen Grenzfall – so sind Spinorbitale, die sich nur im Spin unterscheiden, entartet.

Betrachten wir nun ein Zweielektronensystem. Der allgemeine elektronische Hamiltonoperator hat die bekannte Form

$$
\hat{H}_{\mathrm{el}}
=
-\underbrace{\frac{1}{2}\nabla_1^2-\frac{1}{2}\nabla_2^2}_{T_e}
-\underbrace{V_{ne}(1)-V_{ne}(2)}_{V_{ne}}
+\underbrace{\frac{1}{r_{12}}}_{V_{ee}}
+\underbrace{V_{nn}}_{V_{nn}} .
$$

Die Bestimmung der räumlichen Zustände eines Zweielektronensystems ist wegen der Präsenz des $1/r_{12}$-Terms im Hamiltonoperator generell ein kompliziertes, nicht analytisch lösbares Problem. Um dieses Problem zu vereinfachen, betrachten wir zunächst den Fall zweier **nichtwechselwirkender** Elektronen. In diesem Fall können wir den $1/r_{12}$-Term vernachlässigen, was zu einem separablen Hamiltonoperator führt:

$$
\hat{H}_{\mathrm{el}}
=
\left[-\frac{1}{2}\nabla_1^2 - V_{ne}(1)\right]
+
\left[-\frac{1}{2}\nabla_2^2 - V_{ne}(2)\right].
$$

Die beiden Operatoren in den Klammern hängen jeweils nur von den Koordinaten eines Elektrons ab.Für nichtwechselwirkende Elektronen lässt sich, wie bereits bekannt, der exakte räumliche Zustand nun als Tensorprodukt zweier Einelektron-Zustände schreiben:

$$
|\psi(1,2)\rangle
=
|\psi_i(1)\rangle \otimes |\psi_j(2)\rangle .
$$

Dabei sind $|\psi_i\rangle$ und $|\psi_j\rangle$ Eigenfunktionen des Einelektronoperators 

$\hat{h}$:
$$
\begin{aligned}
\hat{h}(1)|\psi_i(1)\rangle &= \varepsilon_i |\psi_i(1)\rangle, \\
\hat{h}(2)|\psi_j(2)\rangle &= \varepsilon_j |\psi_j(2)\rangle .
\end{aligned}
$$

Der Zustand $|\psi(1,2)\rangle$ ist damit ein Eigenzustand von
$$
\hat{H}_{\mathrm{el}} = \hat{h}(1) + \hat{h}(2)
$$
mit der Energie $\varepsilon_i + \varepsilon_j$.

Betrachten wir nun den energetisch niedrigsten **räumlichen** Zustand des Zweielektronensystems. Dieser entsteht, wenn beide Elektronen das niedrigste Orbital besetzen (z. B. das 1s-Orbital im He-Atom):

$$
|\psi_{\mathrm{raum}}(1,2)\rangle
=
|\psi_0(1)\rangle \otimes |\psi_0(2)\rangle .
$$

Da der nichtrelativistische Hamiltonoperator keinen Spin enthält, kann dieser räumliche Zustand formal mit einem beliebigen Zweielektron-Spinzustand kombiniert werden. Der Spinraum eines Elektrons ist zweidimensional und wird von den Zuständen $|\alpha\rangle$ und $|\beta\rangle$ aufgespannt. Der Spinraum des Zweielektronensystems ist daher das Tensorprodukt der beiden Einelektron-Spinräume und ist, wie wir gesehen werden vierdimensional mit den Basiszuständen:

$$
|\alpha(1)\rangle \otimes |\alpha(2)\rangle,\quad
|\alpha(1)\rangle \otimes |\beta(2)\rangle,\quad
|\beta(1)\rangle \otimes |\alpha(2)\rangle,\quad
|\beta(1)\rangle \otimes |\beta(2)\rangle .
$$

Diese lassen sich zu Eigenzuständen von $S^2$ und $S_z$ kombinieren. Man erhält einen Singulett- und drei Triplett-Zustände:

$$
\begin{aligned}
|\chi_S\rangle
&=
\frac{1}{\sqrt{2}}
\left(
|\alpha(1)\rangle \otimes |\beta(2)\rangle
-
|\beta(1)\rangle \otimes |\alpha(2)\rangle
\right), \\
|\chi_{T,+1}\rangle
&=
|\alpha(1)\rangle \otimes |\alpha(2)\rangle, \\
|\chi_{T,0}\rangle
&=
\frac{1}{\sqrt{2}}
\left(
|\alpha(1)\rangle \otimes |\beta(2)\rangle
+
|\beta(1)\rangle \otimes |\alpha(2)\rangle
\right), \\
|\chi_{T,-1}\rangle
&=
|\beta(1)\rangle \otimes |\beta(2)\rangle .
\end{aligned}
$$

Formal ergeben sich damit vier Gesamtzustände gleicher Energie:

$$
\begin{aligned}
|\Psi_S(1,2)\rangle
&=
|\psi_0(1)\rangle \otimes |\psi_0(2)\rangle \otimes |\chi_S\rangle, \\
|\Psi_{T,M}(1,2)\rangle
&=
|\psi_0(1)\rangle \otimes |\psi_0(2)\rangle \otimes |\chi_{T,M}\rangle,
\qquad M = +1,0,-1 .
\end{aligned}
$$

Wären alle diese Zustände physikalisch erlaubt, so wäre der Grundzustand eines Zweielektronensystems wie des He-Atoms oder des $\mathrm{H}_2$-Moleküls vierfach entartet. Experimentell wird diese Vierfachentartung jedoch **nicht** beobachtet: Der Grundzustand ist eindeutig nicht entartet. Offensichtlich werden die Triplett-Zustände nicht realisiert.



Der fundamentale Unterschied zeigt sich in der Vertauschung der Elektronen. Für den Singulett-Zustand gilt

$$
P_{12}|\psi_{\mathrm{Singulett}}(1,2)\rangle
=
-|\psi_{\mathrm{Singulett}}(1,2)\rangle ,
$$

der Gesamtzustand ist antisymmetrisch. Die Triplett-Zustände sind dagegen symmetrisch:

$$
P_{12}|\psi_{\mathrm{Triplett}}(1,2)\rangle
=
+|\psi_{\mathrm{Triplett}}(1,2)\rangle .
$$

<div style="border: 2px solid #b00020; background-color: #fdecea; padding: 12px; border-radius: 6px;">

<strong>Pauli-Prinzip</strong>  

Der Zustand eines Systems aus $N$ Elektronen muss antisymmetrisch bezüglich der Vertauschung beliebiger Elektronen sein.

</div>


Eine unmittelbare Konsequenz ist, dass zwei Elektronen mit gleichem Spin niemals dasselbe räumliche Orbital besetzen können. Der antisymmetrisierte Zustand verschwindet in diesem Fall identisch.

Woher kommt dieses Prinzip? Die Antwort liegt in der **Ununterscheidbarkeit von Quantenteilchen**. Messwahrscheinlichkeiten dürfen nicht davon abhängen, wie identische Teilchen nummeriert werden.

Betrachten wir dazu zwei nichtwechselwirkende Elektronen in einem eindimensionalen Kastenpotential:

$$
\Psi_1(1,2)
=
\sqrt{\frac{2}{L}}
\sin\!\frac{\pi x_1}{L}
\sin\!\frac{2\pi x_2}{L},
$$

$$
\Psi_2(1,2)
=
\sqrt{\frac{2}{L}}
\sin\!\frac{2\pi x_1}{L}
\sin\!\frac{\pi x_2}{L}.
$$

Die Wahrscheinlichkeit, das „Elektron 1“ bei $x_1 = L/4$ und das „Elektron 2“ bei $x_2 = L/2$ zu finden, ist für diese beiden Zustände unterschiedlich. Dieses Ergebnis hängt jedoch nur von der gewählten Bezeichnung der Elektronen ab und ist daher physikalisch unzulässig.

Die Wellenfunktion muss daher die Eigenschaft besitzen

$$
|\psi(1,2)|^2 = |\psi(2,1)|^2 ,
$$

was impliziert

$$
\psi(2,1) = \pm \psi(1,2).
$$

Die Ununterscheidbarkeit identischer Teilchen erzwingt somit Zustände, die entweder symmetrisch oder antisymmetrisch bezüglich der Vertauschung sind.

Warum aber müssen Elektronen antisymmetrische Zustände besitzen? Erst die relativistische Quantenfeldtheorie liefert die Antwort:
$$
\begin{aligned}
\text{halbzahliger Spin} &\Rightarrow \psi(2,1) = -\psi(1,2), \\
\text{ganzzahliger Spin} &\Rightarrow \psi(2,1) = \phantom{-}\psi(1,2).
\end{aligned}
$$

Teilchen mit halbzahligem Spin heißen **Fermionen**, solche mit ganzzahligem Spin **Bosonen**.  

**Elektronen besitzen Spin $\tfrac{1}{2}$, sind Fermionen und erfordern daher eine antisymmetrische Beschreibung ihrer Gesamtzustände.**
