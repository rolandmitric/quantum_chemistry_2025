Wir haben gelernt, dass Wolfgang Pauli das dem Periodensystem der Elemente zugrunde liegende Aufbauprinzip erklärt hat, indem er annahm, dass ein Elektron eine Eigenschaft (Observable) hat, die nur **zwei mögliche Werte** annehmen kann. Darüber hinaus nahm Pauli an, dass sich in jedem Zustand, der durch eine räumliche Wellenfunktion definiert ist, nur Elektronen aufhalten können, die sich durch den Wert dieser etwas geheimnisvollen ''zweideutigen'' Eigenschaft unterscheiden. Dies war historisch die erste Formulierung des (Paulischen) Ausschlussprinzips, das später allgemeiner formuliert wurde (wie wir noch sehen werden). 


Wir versuchen ausgehend von dieser Annahme eine konsistente quantenmechanische Beschreibung solcher Observablen aufzubauen, die nur zwei mögliche (Mess)-Werte haben können. Dabei werden wir den Spin des Elektrons entdecken! 

Unsere Betrachtungen folgen nicht dem historischen Weg, der zur Entdeckung des Spins geführt hat. Wir versuchen vielmehr die Existenz des Spins basierend auf der modernen Formulierung der Quantenmechanik plausibel zu machen. 

Eine quantenmechanische Observable, die nur zwei mögliche Werte annehmen kann,  wird in der Quantenmechanik durch einen hermiteschen Operator mit zwei zueinander orthogonalen Eigenzuständen beschrieben. 

Diese zwei Basiszustände bezeichnen wir im Folgenden mit $|\alpha \rangle  $ und $| \beta \rangle $. Dabei haben wir die sogennante **Diracsche Bra-Ket-Notation** für die quantenmechanischen Zustände eingeführt. 

In der Dirac-Notation wird ein allgemeiner Quantenzustand durch einen sog. Ket-Vektor $|\psi\rangle$ beschrieben, den man als eine basisunabhängige Darstellung des Quantenzustands auffassen kann. Im Gegensatz dazu entspricht die übliche Beschreibung durch raumabhängige Wellenfunktionen einer speziellen Darstellung des Ket-Vektors in der Basis der Eigenzustände des Ortsoperators. Zur Erinnerung: Die Beschreibung der Quantenzustände durch Vektoren ist der mathematische Ausdruck für die experimentelle Tatsache, dass man Quantenzustände linear kombinieren kann, wodurch neue Quantenzustände entstehen. Dies ist die fundamentalste Eigenschaft der Quantenmechanik, die im völligen Gegensatz zu der klassischen Beschreibung steht. 

<h1> Komplexer Vektorraum der Quantenzustände </h1> 

Die grundlegendste Eigenschaft quantenmechanischer Zustände ist also, dass wir Linearekombinationen aus ihnen aufbauen können, wodurch neue quantenmechanische Zustände erzeugt werden. Mathematisch gesehen, sind die quantenmechanischen Zustände **Vektoren in einem komplexen Vektorraum**.

Da unsere bisher etwas mysteriöse zweideutige Observable mit zwei Basiszuständen beschrieben werden kann, können wir durch Bildung beliebiger Linearkombinationen der Basiszustände neue Quantenzustände erzeugen: 

$$ \psi = c_\alpha |\alpha \rangle  + c_\beta |\beta \rangle. $$

So eine Linearkombination ist vollständig durch zwei **komplexe Zahlen**  $c_\alpha $ and $ c_\beta $ charakterisiert. Diese können wir kompakt z. B. durch einen Spaltenvektor darstellen. So kann jedem Zustand des Systems ein Spaltenvektor zugeordnet werden:

$$ \psi\rightarrow\left[\begin{array}{c}
c_{\alpha}\\
c_{\beta}
\end{array}\right]$$

So einen Spaltenvektor können wir auch als:

$$\left[\begin{array}{c}
c_{\alpha}\\
c_{\beta}
\end{array}\right]=c_{\alpha}\left[\begin{array}{c}
1\\
0
\end{array}\right]+c_{\beta}\left[\begin{array}{c}
0\\
1
\end{array}\right] $$

darstellen. 

Daraus sehen wir, dass wir den Basiszuständen $|\alpha \rangle $ und $|\beta \rangle $ auch zwei entsprechende Spaltenvektoren zuordnen können:

$$|\alpha\rangle\rightarrow\left[\begin{array}{c}
1\\
0
\end{array}\right]\quad|\beta\rangle\rightarrow\left[\begin{array}{c}
0\\
1
\end{array}\right]$$

<u> Bemerkung </u> : Einen allgemeinen Vektor in einem zweidimesionalen komplexen Vektorraum bezeichnen man manchmal in der Literatur als **Spinor**.

<h1> Hermitesche Operatoren in einem zweidimensionalen komplexen Vektorraum </h1>

Wenn wir nun der Idee von Wolfgang Pauli folgen, müssen wir zur Beschreibung der ''Zweideutigkeit'' einen hermiteschen Operator einführen, der nur zwei mögliche Eigenzustände und Eigenwerte hat. Dieser Operator, den wir zunächst mit $\hat{\sigma}$ bezeichnen, sollte durch die folgende Wirkung auf die Basiszustände definiert sein: 

$$\hat{\sigma}|\alpha\rangle=+1|\alpha\rangle$$

$$ \hat{\sigma}|\beta\rangle=-1|\beta\rangle $$ 

Diesem Operator können wir also in der Matrix-Darstellung die folgende 2x2 Matrix zuordnen:


$$ \hat{\sigma}\rightarrow\left[\begin{array}{cc}
1 & 0\\
0 & -1
\end{array}\right] $$

Wir wollen uns im Folgenden die allgemeine Struktur der hermiteschen Operatoren in einem komplexen 2D-Vektorraum anschauen. 

Damit wir aber überhaupt den Begriff des hermiteschen Operators einführen können, müssen wir an unserem Vektorraum ein **Skalarprodukt**  definieren. 

Zur Erinnerung: Das Skalarprodukt wird in der Mathematik axiomatisch eingeführt, indem man auf dem gegebenen Vektorraum eine Abbildung (Funktion) definiert, die jedem Paar von Vektoren eine (komplexe) Zahl zuordnet. Zusätzlich muss diese Abbildung gewisse Eigenschaften haben, die analog zur üblichen Definition des Skalarproduktes aus der Geometrie sind. Die formale Definition lautet wie folgt:

Ein Skalarprodukt auf einem Vektorraum V über einem Zahlenkörper C ist eine Abbildung von einem geordneten Vektorenpaar in V auf eine Zahl im Körper C, die die folgenden Eigenschaften erfüllt:

<ol>
  <li> $$ \langle\psi|\psi\rangle\ge0\quad\forall|\psi\rangle  $$ </li>
  <li> $$\langle\psi|\psi\rangle=0\Leftrightarrow|\psi\rangle=0 $$ </li>
  <li> $$\langle\phi|\left(\lambda|\psi\rangle+\mu|\chi\rangle\right)=\lambda\langle\phi|\psi\rangle+\mu\langle\phi|\chi\rangle $$ </li>
    <li> $$ \langle\phi|\psi\rangle=\langle\psi|\phi\rangle^{*} $$ </li>
</ol>




Betrachten wir nun zwei beliebige Vektoren im zweidimensionalen komplexen Vektorraum, die durch folgende Spaltenvektoren dargestellt werden können: 

$$|\psi\rangle\rightarrow\left[\begin{array}{c}
c_{\alpha}\\
c_{\beta}
\end{array}\right]\\
|\phi\rangle\rightarrow\left[\begin{array}{c}
d_{\alpha}\\
d_{\beta}
\end{array}\right] $$

Wir <u> definieren </u>  das Skalarprodukt als :

$$ \langle\phi|\psi\rangle=d_{\alpha}^{*}c_{\alpha}+d_{\beta}^{*}c_{\beta} $$

Diese Definition erfüllt alle Definitionseigenschaften eines Skalarprodukts.

Bemerkung: Das Skalarprodukt kann formal auch in Form eines Matrixproduktes geschrieben werden :

$$
\langle\psi \mid \psi\rangle=c_\alpha^* c_\alpha+c_\beta^* c_\beta=\left[\begin{array}{ll}
c_\alpha^* & c_\alpha
\end{array}\right]\left[\begin{array}{l}
c_\alpha \\
c_\beta
\end{array}\right]
$$

Wir sehen, dass man genauso, wie man den Ket-Vektoren einen Spaltenvektor $\left[\begin{array}{l}c_\alpha \\ c_\beta\end{array}\right]$ zugeordnet hat, formal den Ausdruck $\langle\psi|$ als einen Zeilenvektor auffassen kann:

$$
\begin{aligned}
& |\psi\rangle \rightarrow\left[\begin{array}{l}
c_\alpha \\
c_\beta
\end{array}\right] \\
& \langle\psi| \rightarrow\left[\begin{array}{ll}
c_\alpha^* & c_\beta^*
\end{array}\right]
\end{aligned}
$$

Diese neuen Zeilenvektoren bezeichnet man als Bra-Vektoren. Die Bra-Vektoren "leben" in einem anderen (dualen) Vektorraum V$^*$. Mathematiker identifizieren solche Bra-Vektoren mit linearen Funktionalen (Formen), die jedem Vektor aus dem Vektorraum V eine komplexe Zahl zuordnen (durch Bildung des Skalarprodukts). Wir haben also eine allgemeine Vorschrift, wie man aus einen Ket-Vektor im 2D-komplexen Vektorraum einen Bra-Vektor bildet.

Betrachten wir einen allgemeinen Zustand $|\psi\rangle=c_\alpha|\alpha\rangle+c_\beta|\beta\rangle$, dem wir in der Matrix-Darstellung den Spaltenvektor

$$
|\psi\rangle \rightarrow\left[\begin{array}{l}
c_\alpha \\
c_\beta
\end{array}\right]
$$

zugeordnet haben. Laut unserer Definition wäre das Skalarprodukt $\langle\psi \mid \psi\rangle$ durch die folgende Beziehung gegeben:

$$
\begin{aligned}
\langle\psi \mid \psi\rangle & =c_\alpha^* c_\alpha+c_\beta^* c_\beta \\
& =\left|c_\alpha^2\right|^2+\left|c_\beta\right|^2
\end{aligned}
$$

Dies ist aber immer positiv und erfüllt damit die Definitionseigenschaft 1. Es gilt also

$$
\langle\psi \mid \psi\rangle \geqslant 0 \quad \forall \psi \text {. }
$$

Des Weiteren ist $\langle\psi \mid \psi\rangle=0$ äquivalent mit $c_\alpha=c_\beta=0$, womit auch die Eigenschaft 2 erfüllt ist. Die Linearitat lässt sich ebenfalls durch direkte Anwendung der Definition überprüfen, so dass wir hiermit eine valide Definition eines Skalarprodukts auf unserem Vektorraum haben!

Wir können nun analog zum Ket-Vektorraum eine Basis im dualen Vektorraum der Bras konstruieren:

$$
\begin{aligned}
& |\psi\rangle=c_\alpha|\alpha\rangle+c_\beta|\beta\rangle \Longleftrightarrow\langle\psi|=c_\alpha^*\langle\alpha|+c_\beta^*\langle\beta| \\
& \text { Matrix-Darstellung }|\psi\rangle \rightarrow c_\alpha\left[\begin{array}{l}
1 \\
0
\end{array}\right]+c_\beta\left[\begin{array}{l}
0 \\
1
\end{array}\right] \Leftrightarrow\langle\psi|=c_\alpha^*[1,0]+c_\beta[0,1] \\
&
\end{aligned}
$$

Dabei stellen wir fest, dass die Zeilenvektoren $\langle\alpha| \equiv[1,0]$ und $\langle\beta| \equiv[0,1]$ eine Basis des dualen Raumes bilden. Für die Skalarprodukte der Basisvektoren gilt speziell:

$$
\begin{aligned}
& \langle\alpha \mid \alpha\rangle=[1,0]\left[\begin{array}{l}
1 \\
0
\end{array}\right]=1 \\
& \langle\beta \mid \beta\rangle=[0,1]\left[\begin{array}{l}
0 \\
1
\end{array}\right]=1 \\
& \langle\alpha \mid \beta\rangle=[1,0]\left[\begin{array}{l}
0 \\
1
\end{array}\right]=0 \\
& \langle\beta \mid \alpha\rangle=[0,1]\left[\begin{array}{l}
1 \\
0
\end{array}\right]=0
\end{aligned}
$$

Im Folgenden führen wir lineare Operatoren im zweidimensionalen komplexen Vektorraum ein. Dazu erinnern wir uns zunächst an die allgemeine Definition eines linearen Operators. 

<u> Definition </u> : Ein linearer Operator $\hat{A}: V \rightarrow V$ ist eine Funktion mit folgenden Eigenschaften:

1) Linearität $\hat{A}(|u\rangle+|v\rangle)=\hat{A}|u\rangle+\hat{A}|v\rangle,\; \forall|u\rangle,|v\rangle$

2) Homogenität $\hat{A}(\lambda|u\rangle)=\lambda \hat{A}|u\rangle,\; \forall|u\rangle \in V, \forall \lambda \in \mathbb{C}$


Dies ist äquivalent zu der Aussage, dass

$$
\hat{A}\left(\lambda|u\rangle+\mu|v\rangle\right)=\lambda \hat{A}|u\rangle+\mu \hat{A}|v\rangle\\
\forall |u\rangle,|v\rangle \in V\\
\forall \lambda,\mu \in C
$$

Wir erinnern uns an die Definition eines hermiteschen Operators, die wir bereits aus den vorherigen Vorlesungen kennen. Ein Operator $\hat{A}$ ist hermitesch wenn die folgende Beziehung für zwei beliebige Wellenfunktionen erfüllt ist: 

$$
\int d \tau\, \psi^* \hat{A} \phi=\left[\int d \tau\, \phi^* \hat{A} \psi\right]^* \quad \forall \phi, \psi
$$

Dies können wir nun allgemein in der basisfreien Dirac-Schreibweise wie folgt ausdrücken: 

$$
\langle\psi|\hat{A}| \phi\rangle=\langle\phi|\hat{A}| \psi\rangle^*
$$


Wie sieht die Matrix-Darstellung eines hermiteschen Operators in unserem 2D-Vektorraum aus?

Betrachten wir zunächst einen beliebigen Zustand $|\psi\rangle=c_\alpha|\alpha\rangle+c_\beta|\beta\rangle$.
Es gilt:

$$
\begin{aligned}
\hat{A}|\psi\rangle & =\hat{A}\left(c _\alpha|\alpha\rangle+c_\beta|\beta\rangle\right) \quad \text { wegen der Linearität } \\
& =c_\alpha \hat{A}|\alpha\rangle+c_\beta \hat{A}|\beta\rangle
\end{aligned}
$$

Die Vektoren $\hat{A}|\alpha\rangle$ und $\hat{A}|\beta\rangle$ sind auch zwei Vektoren im 2D-Vektorraum und können in die Basis $|\alpha\rangle,|\beta\rangle$ entwickelt werden:

$$
\begin{aligned}
& \hat{A}|\alpha\rangle=A_{11}|\alpha\rangle+A_{21}|\beta\rangle \\
& \hat{A}|\beta\rangle=A_{12}|\alpha\rangle+A_{22}|\beta\rangle
\end{aligned}
$$

Der Operator $\hat{A}$ ist also im Allgemeinen durch 4 komplexe Zahlen ( $A_{11}, A_{21}, A_{12}, A_{22}$ ), die man auch in einer $2 \times 2$ Matrix zusammenfassen kann, definiert:

$$
\hat{A} \rightarrow\left[\begin{array}{ll}
A_{11} & A_{12} \\
A_{21} & A_{22}
\end{array}\right]
$$

$$
\begin{aligned}
\hat{A}|\psi\rangle & =c_\alpha\left(A_{11}|\alpha\rangle+A_{21}|\beta\rangle\right)+c_\beta\left(A_{12}|\alpha\rangle+A_{22}|\beta\rangle\right) \\
& =\left(A_{11} c_\alpha+A_{12} c_\beta\right)|\alpha\rangle+\left(A_{21} c_\alpha+A_{22} c_\beta\right)|\beta\rangle
\end{aligned}
$$

In der Matrix-Darstellung kann der Operatorausdruck $ \hat{A}|\psi\rangle$ in Form eines Matrix-Vektor Produktes ausgedrückt werden:

$$
\hat{A}|\psi\rangle \rightarrow\left[\begin{array}{l}
A_{11} c_\alpha+A_{12} c_\beta \\
A_{21} c_\alpha+A_{22} c_\beta
\end{array}\right]=\left[\begin{array}{ll}
A_{11} & A_{12} \\
A_{21} & A_{12}
\end{array}\right]\left[\begin{array}{l}
c_\alpha \\
c_\beta
\end{array}\right]
$$

Wir sehen, dass man einen beliebigen lineare Operator A durch eine $2 \times 2$ Matrix dartellen bann. Das Ergebnis der Anwendung von $\hat{A}$ auf einen belietigen Vector $|\psi\rangle$ erhaten wir in der Matrix-Darstellung durch Matrix-Vektor Multiplikation.

Betrachten wir nun den Ausdruck $\langle\phi|\hat{A}| \psi\rangle$ für zwei beliebige Zustände:

$$
\begin{aligned}
& |\psi\rangle=\left[\begin{array}{l}
c_\alpha \\
c_\beta
\end{array}\right] \\
& |\phi\rangle=\left[\begin{array}{l}
d_\alpha \\
d_\beta
\end{array}\right]
\end{aligned}
$$

Es gilt :

$$
\begin{aligned}
& \langle\phi|\hat{A}| \psi\rangle=\left[\begin{array}{ll}
d_\alpha^* & d_\beta^*
\end{array}\right]\left[\begin{array}{ll}
A_{11} & A_{12} \\
A_{21} & A_{22}
\end{array}\right]\left[\begin{array}{l}
c_\alpha \\
c_\beta
\end{array}\right] \\
& =\left[d_\alpha^*, d_\beta^*\right]\left[\begin{array}{l}
A_{11} c_\alpha+A_{12} c_\beta \\
A_{11} c_\alpha+A_{22} c_\beta
\end{array}\right] \\
&
\end{aligned}
$$

Damit kann man das obige Skalarprodukt in der folgenden allgemeinen Form schreiben:

$$
\langle\phi|\hat{A}| \psi\rangle=\left(A_{11} c_\alpha+A_{12} c_\beta\right) d_\alpha^*+\left(A_{21} c_\alpha+A_{22} c_\beta\right) d_\beta^*
$$

Betrachten wir nun:

$$
\begin{aligned}
\langle\psi|A| \phi\rangle & =\left[\begin{array}{ll}
c_\alpha^* & c_\beta^*
\end{array}\right]\left[\begin{array}{ll}
A_{11} & A_{12} \\
A_{21} & A_{12}
\end{array}\right]\left[\begin{array}{l}
d_\alpha \\
d_\beta
\end{array}\right] \\
& =\left(A_{11} d_\alpha+A_{12} d_\beta\right) c_\alpha^*+\left(A_{21} d_\alpha+A_{22} d_\beta\right) c_\beta^*
\end{aligned}
$$

Damit der Operator $\hat{A}$ die Eigenschaft der Hermitezität besitzt, muss für alle Paare von Zuständen die folgende Bedingung gelten:

$$
\begin{aligned}
\langle\psi|\hat{A}| \phi\rangle=\langle\phi|\hat{A}| \psi\rangle^* \Rightarrow & \left(A_{11} c_\alpha+A_{12} c_\beta\right) d_\alpha^*+\left(A_{21} c_\alpha+A_{22} c_\beta\right) d_\beta^* \\
& =\left(A_{11} d_\alpha+A_{12} d_\beta\right)^* c_\alpha+\left(A_{21} d_\alpha+A_{22} d_\beta\right)^* c_\beta
\end{aligned}
$$


Daraus folgt die Identität:

$$
\left(A_{11}-A_{11}^*\right) c_\alpha d_\alpha^*+\left(A_{22}-A_{22}^*\right) c_\beta d_\beta^*+\left(A_{12}-A_{21}\right)^* c_\beta d_\alpha^*+\left(A_{21}-A_{12}^*\right) c_\alpha d_\beta^*=0
$$

Diese muss nun für beliebige $c_\alpha, c_\beta, d_\alpha, d_\beta$ gelten. Dies ist aber nur dann der Fall, wenn die folgenden Beziehungen zwischen den Matrixelementen $A_{ij}$ bestehen:

$$
\begin{aligned}
& A_{11}=A_{11}^* \Rightarrow A_{11} \text { ist reell! } \\
& A_{22}=A_{22}^* \Rightarrow A_{22} \text { ist reell! } \\
& A_{21}=A_{12}^* \\
& A_{12}=A_{21}^*
\end{aligned}
$$

Die Matrix-Darstellung eines hermiteschen Operators hat unter diesen Bedingungen eine allgemeine Form:

$$
A=\left[\begin{array}{cc}
a & c-i d \\
c+i d & b
\end{array}\right]
$$

$a, b, c, d$ sind dabei beliebige reelle Zahlen.

Ein beliebiger hermitescher Operator an einem 2D-Vektorraum kann als eine Linearkombination von 4 Basisoperatoren dargestellt werden:

$$
A=\left[\begin{array}{cc}
a & c-i d \\
c+i d & b
\end{array}\right]=\left[\begin{array}{ll}
a & 0 \\
0 & b
\end{array}\right]+\left[\begin{array}{cc}
0 & c-i d \\
c+i d & 0
\end{array}\right]
$$

oder

$$
\begin{aligned}
& A=\left[\begin{array}{cc}
\frac{a+b}{2}+\frac{a-b}{2} & 0 \\
0 & \frac{a+b}{2}-\left(\frac{a-b}{2}\right)
\end{array}\right]+c\left[\begin{array}{cc}
0 & 1 \\
1 & 0
\end{array}\right]+d\left[\begin{array}{cc}
0 & -i \\
i & 0
\end{array}\right] \\
\end{aligned}
$$

$$
\begin{aligned}
& A=\frac{a+b}{2}\left[\begin{array}{ll}
1 & 0 \\
0 & 1
\end{array}\right]+\frac{a-b}{2}\left[\begin{array}{cc}
1 & 0 \\
0 & -1
\end{array}\right]+c\left[\begin{array}{ll}
0 & 1 \\
1 & 0
\end{array}\right]+d\left[\begin{array}{cc}
0 & -i \\
i & 0
\end{array}\right]
\end{aligned}
$$

Wir definieren nun neue Parameter als

$$
\begin{aligned}
& a_1=c \\
& a_2=d \\
& a_3=\frac{a-b}{2} \\
& a_0=\frac{a+b}{2}
\end{aligned}
$$


Mit den neuen Parametern könen wir A auch in der folgenden Form schreiben:
$$
A=a_0 \cdot \underbrace{\left[\begin{array}{ll}
1 & 0 \\
0 & 1
\end{array}\right]}_I+a_1 \underbrace{\left[\begin{array}{ll}
0 & 1 \\
1 & 0
\end{array}\right]}_{\sigma_1}+a_2 \underbrace{\left[\begin{array}{cc}
0 & -i \\
i & 0
\end{array}\right]}_{\sigma_2}+a_{3}\underbrace{\left[\begin{array}{cc}
1 & 0 \\
0 & -1
\end{array}\right]}_{\sigma_3}
$$

Dies liefert uns die Basis für die Darstellung eines beliebigen hermiteschen Operators, die aus 4 Matrizen bestent:


$$
\begin{aligned}
& I=\left[\begin{array}{ll}
1 & 0 \\
0 & 1
\end{array}\right] \\
& \sigma_1=\left[\begin{array}{ll}
0 & 1 \\
1 & 0
\end{array}\right] \\
& \sigma_2=\left[\begin{array}{cc}
0 & -i \\
i & 0
\end{array}\right] \\
& \sigma_3=\left[\begin{array}{cc}
1 & 0 \\
0 & -1
\end{array}\right]
\end{aligned}
$$
Die Matrizen $\sigma_1, \sigma_2, \sigma_3$ bezeichnet man als <b> Pauli-Matrizen </b> !


Welchen physikalischen Inhalt haben diese mathematischen Erkenntnisse über die allgemeine Form eines hermiteschen Operators im zweidimensionalen komplexen Vektorraum?

<u> Jede physikalische Observable </u>  in einem Quantensystem mit nur zwei Basiszuständen kann als Linearkombination von drei Basisobservablen dargestellt werden (ohne den trivialen Identitätsoperator $I$). Die Pauli-Annahme über die Existenz einer Eigenschaft mit nur zwei möglichen Werten führt also auf natürlichem Wege zur Existenz von nicht einem, sondern drei verwandten hermiteschen Operatoren (Observablen), die durch die drei Pauli-Matrizen dargestellt werden können. Diese stellen gewissermaßen die ''natürlichsten'' Observablen überhaupt dar, die in einem System mit zwei Basiszuständen existieren können.  

Aber welche physikalische Observable stellen diese drei Operatoren dar? Wie kann man diese mysteriöse Observable messen und können alle drei $\sigma$'s gleichzeitig gemessen werden oder nicht? 

Die Antworten auf diese Fragen werden uns zur Entdeckung des Spins führen, der, wie wir zeigen werden, eine eigentümliche Form eines Drehimpulses darstellt, für die es kein klassisches Analogon gibt!

<h1> Einige Eigenschaften von Pauli-Matrizen </h1> 

Wir untersuchen im Folgenden einige mathematische Eigenschaften von Pauli-Matrizen.

Zunächst bemerken wir, dass alle drei Pauli-Matrizen hermitesch sind und somit auch quantenmechanische Observablen darstellen. Die Messwerte von quantenmechanischen Observablen sind im Allgemeinen durch die Eigenwerte der entsprechenden Operatoren bestimmt.

Welche Eigenwerte haben also $\sigma_1, \sigma_2$ und $\sigma_3$ ?

Für alle drei Pauli-Matrizen gilt die folgende Beziehung, die man durch direkte Multiplikation überprüfen kann: 

$$\sigma_i^2=I, \quad i=1,2,3. $$

Zum Beispiel haben wir für $\sigma_1$ :

$$
\sigma_1^2=\left[\begin{array}{ll}
0 & 1 \\
1 & 0
\end{array}\right] \cdot\left[\begin{array}{ll}
0 & 1 \\
1 & 0
\end{array}\right]=\left[\begin{array}{ll}
1 & 0 \\
0 & 1
\end{array}\right]=I
$$

Aus der Relation $\sigma_i^2=I$ folgt unmittelbar, dass die Eigenwerte der Pauli-Matrizen $\pm 1$ sind. Dies können wir wie folgt beweisen: 

Sei $|v\rangle$ ein Eigenvektor der Pauli-Matrix $\sigma_i$ mit dem Eigenwert $\lambda$ :

$$
\sigma_i|v\rangle=\lambda|v\rangle
$$

 Dann gilt:
 
$$
\begin{aligned}
\sigma_i^2|v\rangle & =\sigma_i\left(\sigma_i|v\rangle\right) \\
& =\sigma_i(\lambda|v\rangle) \\
& =\lambda \sigma_i|v\rangle \\
& =\lambda^2|v\rangle
\end{aligned}
$$

Andererseits gilt aber $\sigma_i^2=I$. Daraus folgt:

$$
I|v\rangle=\lambda^2|v\rangle \Rightarrow\left(\lambda^2-1\right)|v\rangle=0
$$

Daraus ergibt sich, für $|v\rangle \neq|0\rangle$,  $\lambda=\pm 1$.


<b>Kommutatoren der Pauli-Matrizen</b>

Wir haben gesehen, dass Pauli-Matrizen als hermitesche Matrizen quantenmechanische Observablen darstellen. Nun fragen wir uns, ob unterschiedliche $\sigma_i$'s gemeinsame Eigenvektoren haben und damit gleichzeitig beliebig scharf messbar sind. Um diese Frage zu beantworten, berechnen wir den Kommutator zweier Pauli-Matrizen.

Wir rechnen zunächst $\sigma_1 \sigma_2$ aus:
$$
\sigma_1 \sigma_2=\left[\begin{array}{ll}
0 & 1 \\
1 & 0
\end{array}\right]\left[\begin{array}{cc}
0 & -i \\
i & 0
\end{array}\right]=\left[\begin{array}{cc}
i & 0 \\
0 & -i
\end{array}\right]=i\left[\begin{array}{cc}
1 & 0 \\
0 & -1
\end{array}\right]
$$
Für $\sigma_1 \sigma_2$ bekommen wir also:
$$
\sigma_1 \sigma_2=i \sigma_3
$$
Nun berechnen wir $\sigma_2 \sigma_1$ :
$$
\begin{aligned}
& \sigma_2 \sigma_1=\left[\begin{array}{cc}
0 & -i \\
i & 0
\end{array}\right]\left[\begin{array}{ll}
0 & 1 \\
1 & 0
\end{array}\right]=\left[\begin{array}{cc}
-i & 0 \\
0 & i
\end{array}\right]=-i\left[\begin{array}{cc}
1 & 0 \\
0 & -1
\end{array}\right] \\
& \sigma_2 \sigma_1=-i \sigma_3
\end{aligned}
$$
Wir sehen, dass $\sigma_1$ und $\sigma_2$ nicht vertauschbar sind!

Der Kommutator von $\sigma_1$ und $\sigma_2$ ist damit:

$$
\left[\sigma_1, \sigma_2\right]=\sigma_1 \sigma_2-\sigma_2 \sigma_1=2 i \sigma_3
$$
Genauso kann man durch direkte Rechnung zeigen, dass

$$
\begin{aligned}
& {\left[\sigma_2, \sigma_3\right]=2 i \sigma_1} \\
& {\left[\sigma_3, \sigma_1\right]=2 i \sigma_2}
\end{aligned}
$$

Wir haben nun ein sehr interessantes Ergebnis erhalten: Pauli-Matrizen erfüllen ähnliche Vertauschungsrelationen wie der quantenmechanische Drehimpuls, den wir bei der quantenmechanischen Behandlung der Drehbewegung kennengelernt haben:

$$
\left[\sigma_1, \sigma_2\right]=2 i \sigma_3 \leftrightarrow\left[L_x, L_y\right]=i \hbar L_z \text { usw. }
$$

Die mysteriöse zweideutige Eigenschaft der Elektronen, die Pauli eingeführt hat, um das Aufbau-Prinzip der Atome zu erklären, ist also eine Art des Drehimpulses! Diesen etwas besonderen Drehimpuls bezeichnet man als "<b>Spin</b>".

Die Vertauschungsrelationen der Pauli-Matrizen können wir ganz analog zu denen des Drehimpulses, indem wir folgende Spin-Operatoren definieren:

$$
\begin{aligned}
& S_x=\frac{\hbar}{2} \sigma_1 \\
& S_y=\frac{\hbar}{2} \sigma_2 \\
& S_z=\frac{\hbar}{2} \sigma_3
\end{aligned}
$$

Diese unterscheiden sich von den Pauli-Matrizen nur durch die geschickt eingefügte Plancksche Konstante.
Mit diesen Definitionen gilt z. B.:

$$
\begin{aligned}
{\left[S_x, S_y\right] } & =\left[\frac{\hbar}{2} \sigma_1, \frac{\hbar}{2} \sigma_2\right] \\
& =\left(\frac{\hbar}{2}\right)^2\left[\sigma_1, \sigma_2\right] \\
& =\left(\frac{\hbar}{2}\right)^2 \cdot 2 i \sigma_3 \\
& =i \hbar S_z
\end{aligned}
$$

also völlig analog zum Drehimpuls $\hat{\mathbf{L}}$!

Nachdem wir den Kommutator der Pauli-Matrizen berechnet haben, bemerken wir auch die folgende Relation:

$$
\begin{aligned}
& \sigma_1 \sigma_2+\sigma_2 \sigma_1=0 \\
& \sigma_2 \sigma_3+\sigma_3 \sigma_2=0 \\
& \sigma_3 \sigma_1+\sigma_1 \sigma_3=0
\end{aligned}
$$

Multiplikation von Pauli-Matrizen ist also antikommutativ! Wir definieren den <b>Anti-Kommutator</b> als:

$$
\left\{\sigma_i, \sigma_j\right\} \equiv \sigma_i \sigma_j+\sigma_j \sigma_i=0 \text { wenn } i \neq j
$$

Die Vertauschungsrelationen für den Drehimpuls und Spin können mit Hilfe des <b>Levi-Civita-Symbols</b> in einer kompakten allgemeinen Form umschreiben, die viele spätere Berechnungen erleichtert. Wir definieren das Levi-Civita-Symbol $\varepsilon_{i j k}$ als

$$
\varepsilon_{i j k}=\left\{\begin{aligned}
1, & \text { wenn } i j k=123,231,312 \\
-1, & \text { wenn } i j k=213,132,321 \\
0, & \text { ansonsten }
\end{aligned}\right.
$$

Das Levi-Civita-Symbol hat insgesamt 27 Komponenten. Wenn mindestens zwei Indizes gleich sind, ist jedoch $\varepsilon_{i j k}=0$. Für gerade Permutationen aus der Menge $\{1,2,3\}$ ist $\varepsilon_{i j k}=+1$ und für ungerade Permutationen ist $\varepsilon_{i j k}=-1$.

Die Vertauschungsrelationen zwischen Pauli-Matrizen können mit Hilfe des Levi-Civitas Symbols in der folgenden kompakten Form geschrieben werden:

$$
\left[\sigma_i, \sigma_j\right]=2 i \sum_k \varepsilon_{i j k} \sigma_k
$$

oder noch kürzer unter Verwendung der Einsteinschen Summenkonvention (über doppelt auftretende Indizes innerhalb eines Produkts wird summiert):

$$
\left[\sigma_i, \sigma_j\right]=2 i \varepsilon_{i j k} \sigma_k \quad \text { (Summation über k implizit!) }
$$

Wir überprüfen an einem Beispiel, ob die obige Relation tatsächlich stimmt:

$$
\begin{aligned}
{\left[\sigma_1, \sigma_2\right] } & =2 i(\underbrace{\varepsilon_{121} \sigma_1}_{=0}+\underbrace{\varepsilon_{122} \sigma_2}_{=0}+\underbrace{\varepsilon_{123}}_{=1} \sigma_3) \\
& =2 i \sigma_3
\end{aligned}
$$

Eine weitere nützliche Beziehung erhalten wir, indem wir das Produkt der Pauli-Matrizen durch die Einführung des Levi-Civita-Symbols ausdrücken:

$$
\begin{aligned}
\sigma_i \sigma_j & =\frac{1}{2}\left(\sigma_i \sigma_j+\sigma_j \sigma_i\right)+\frac{1}{2}\left(\sigma_i \sigma_j-\sigma_j \sigma_i\right) \\
& =\frac{1}{2}\left\{\sigma_i, \sigma_j\right\}+\frac{1}{2}\left[\sigma_i, \sigma_j\right] \\
& =\frac{1}{2}\left(2 \delta_{i j} I\right)+\frac{1}{2} \cdot 2 i \sum_k \varepsilon_{i j k} \sigma_k
\end{aligned}
$$

Daraus folgt die Pauli-Identität:
$$
\sigma_i \sigma_j=\delta_{i j}+i \sum_k \varepsilon_{i j k} \sigma_k
$$
