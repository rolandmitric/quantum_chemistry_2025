# Spin: Einführung

Wir haben bereits gesehen, dass sich die elektronische Struktur des einfachsten Atoms jenseits von Helium (also Lithium) nicht dadurch erklären lässt, dass alle drei Elektronen dasselbe räumliche Orbital besetzen. Wir erreichen eine zumindest qualitative Übereinstimmung mit Experiment indem wir das dritte Elektron in das nächsthöhere 2s-Orbital zu setzen. Offenbar neigen also Elektronen dazu, jedes räumliche Orbital nur paarweise zu besetzen; sobald ein Orbital „voll“ ist, wandern weitere Elektronen in energetisch höhere Zustände. 

Genau diese Beobachtung führte Wolfgang Pauli zu der Einsicht, dass ein Elektron nicht nur durch seine räumliche Bewegung charakterisiert ist, sondern darüber hinaus eine interne Eigenschaft besitzt (von Pauli ursprünglich ''Zweideutigkeit'' genannt), die zwei unterschiedliche Werte annehmen kann. 

Man kann sich das bildlich gesprochen  wie einen kleinen Schalter vorstellen: Auch bei festgelegter Position kann das Elektron in zwei physikalisch verschiedenen Zuständen auftreten, gleichsam „ON“ oder „OFF“. Auf dieser Grundlage formulierte Pauli seine ursprüngliche Version des Ausschlussprinzips: In einem gegebenen räumlichen Zustand dürfen sich höchstens zwei Elektronen aufhalten, und dies nur dann, wenn sie sich genau in dieser inneren, zunächst rätselhaft anmutenden Zweideutigkeit unterscheiden. 

Wir versuchen ausgehend von dieser Annahme eine konsistente quantenmechanische Beschreibung eines Systems mir nur zwei möglichen Zuständen zu entwickeln. Insbesonder werden wir uns mit der Frage befassen, welche Art von Observablen kann ein solches System haben Dabei werden wir den Spin des Elektrons "entdecken"! 

Unsere Betrachtungen folgen nicht dem historischen Weg, der zur Entdeckung des Spins geführt hat. Wir versuchen vielmehr die Existenz des Spins basierend auf der modernen Formulierung der Quantenmechanik plausibel zu machen. 

Die Annahme, dass sich ein Elektron in zwei physikalisch unterscheidbaren internen Zuständen befinden kann, so ist dies gleichbedeutend mit der Annahme, dass eine quantenmechanische Observable existiert, die genau zwei mögliche Werte annehmen kann. Eine solche Größe wird in der Quantenmechanik durch einen hermiteschen Operator beschrieben, dessen Spektrum aus zwei zueinander orthogonalen Eigenzuständen besteht. Im Folgenden werden wir den bisher eingeführten Formalismus der Quantenmechanik am Beispiel eines solchen Zweizustandssystems leicht verallgemeinern und zugleich präziser formulieren. Dies bildet die Grundlage für eine systematische Behandlung des Spins. 

## Bras, Kets, Vektorräume, Operatoren, Dualräume usw...

Die Basiszustände der misteriösen ''Zweideutigkeit'' bezeichnen wir im Folgenden mit $|\alpha\rangle$ und $|\beta\rangle$. Damit führen wir zugleich die **Diracsche Bra–Ket-Notation** ein, die es erlaubt, Quantenzustände unabhängig von ihrer konkreten Basis-Darstellung zu formulieren.

In dieser Notation wird ein beliebiger Quantenzustand durch einen sog. Ket-Vektor $|\psi\rangle$ beschrieben, der als basisunabhängige Repräsentation des Zustands dient. 

Die übliche bisher bekannte Formulierung über raumabhängige Wellenfunktionen entspricht lediglich einer speziellen Darstellung dieses Ket-Vektors in der Basis der Eigenzustände des Ortsoperators. Zur Erinnerung: Die Beschreibung von Quantenzuständen als Vektoren spiegelt die experimentelle Tatsache wider, dass sich Quantenzustände linear überlagern lassen und so neue Zustände entstehen. Diese Fähigkeit zur Superposition ist eine der fundamentalsten Eigenschaften der Quantenmechanik und unterscheidet sie grundlegend von der klassischen Physik.

### Komplexer Vektorraum der Quantenzustände 

Mathematisch gesehen, sind die quantenmechanischen Zustände aslo **Vektoren in einem komplexen Vektorraum**. Da unsere bisher etwas mysteriöse zweideutige Observable mit zwei Basiszuständen beschrieben werden kann, können wir durch Bildung beliebiger Linearkombinationen der Basiszustände neue Quantenzustände erzeugen: 

$$ \psi = c_\alpha |\alpha \rangle  + c_\beta |\beta \rangle. $$
