# Zusammengesetzte Quantensysteme: Tensorprodukt

Wir betrachten zunächst zwei voneinander getrennte Quantensysteme, die wir System A und System B nennen. Zur Vereinfachung nehmen wir an, dass jedes dieser Systeme aus genau einem Spin-$\tfrac12$ besteht. Sind die Systeme getrennt, so wird jedes durch seinen eigenen Zustandsraum beschrieben; wir bezeichnen diese Vektorräume mit $V_A$ bzw. $V_B$. Für die folgenden Diskussionen ist es zweckmäßig, die Zustände explizit als Spaltenvektoren in der Standardbasis der Eigenzustände von $\hat S_z$ darzustellen, wobei
$|\alpha\rangle \equiv \begin{pmatrix}1\\0\end{pmatrix}$ und
$|\beta\rangle \equiv \begin{pmatrix}0\\1\end{pmatrix}$.
Der Zustand von System A ist dann vollständig durch

$$
|\psi_A\rangle = c^A_{\alpha}\,|\alpha\rangle_A + c^A_{\beta}\,|\beta\rangle_A
= \begin{pmatrix}
c^A_{\alpha} \\
c^A_{\beta}
\end{pmatrix}
$$

gegeben, und entsprechend ist der Zustand von System B charakterisiert durch

$$
|\psi_B\rangle = c^B_{\alpha}\,|\alpha\rangle_B + c^B_{\beta}\,|\beta\rangle_B
= \begin{pmatrix}
c^B_{\alpha} \\
c^B_{\beta}
\end{pmatrix}.
$$

Die Amplituden $c^A_{\alpha}, c^A_{\beta}$ sowie $c^B_{\alpha}, c^B_{\beta}$ kodieren die Wahrscheinlichkeiten für Messergebnisse an den jeweiligen Spins. Beispielsweise ist $|c^A_{\alpha}|^2$ die Wahrscheinlichkeit dafür, bei einer Messung der $z$-Komponente von Spin A (z.\,B. im Stern–Gerlach-Experiment) den Wert $+\hbar/2$ zu erhalten.



Was ändert sich, wenn wir diese beiden isolierten Spins als ein zusammengesetztes Quantensystem betrachten? Die Grundprinzipien der Quantenmechanik verlangen dann, dass Zustände des Gesamtsystems ebenfalls durch Vektoren in einem (größeren) Zustandsraum beschrieben werden. Wie konstruiert man diesen Raum? Er muss mit $V_A$ und $V_B$ so verknüpft sein, dass Messvorhersagen für die einzelnen Spins weiterhin aus den entsprechenden Amplituden folgen. Daher müssen wir festlegen, wie der Zustandsraum des Gesamtsystems $A{+}B$ aus $V_A$ und $V_B$ so aufgebaut wird, dass die physikalischen Prinzipien der Quantenmechanik korrekt widerspiegelt werden.

## Ein Irrweg: Direkte Summe von Vektorräumen

Ein möglicher Ansatz zur Konstruktion des Zustandsraums des zusammengesetzten Systems bestünde darin, die beiden Zustandsvektoren einfach zu einem größeren Vektor „aufeinanderzustapeln“, etwa in der Form

$$
\begin{pmatrix}
|\psi_A\rangle \\
|\psi_B\rangle
\end{pmatrix}
=
\begin{pmatrix}
c^A_{\alpha} \\
c^A_{\beta} \\
c^B_{\alpha} \\
c^B_{\beta}
\end{pmatrix}.
$$

Mathematisch entspricht dies der Bildung der **direkten Summe** der beiden Vektorräume, also $V_A \oplus V_B$. Obwohl dies formal möglich ist, führt es nicht zu einem physikalisch sinnvollen Zustandsraum für das Gesamtsystem. Dies lässt sich bereits am Superpositionsprinzip erkennen.

Angenommen, Spin A befinde sich in einer echten quantenmechanischen Überlagerung zweier (beliebiger) Zustände,

$$
|\psi_A\rangle = c\,|\psi\rangle_A + d\,|\phi\rangle_A,
$$

während Spin B sich in einem festen Zustand $|\chi\rangle_B$ befindet. In einem physikalisch korrekten Zustandsraum des Gesamtsystems müsste diese Überlagerung dazu führen, dass auch der Gesamtzustand eine Überlagerung zweier unterschiedlicher **Gesamtzustände** ist. In der hier verwendeten Spaltenvektor-Notation würde man daher erwarten, dass der Gesamtzustand als Linearkombination zweier *vollständiger* Zustandsvektoren des zusammengesetzten Systems geschrieben werden kann, etwa in der Form

$$
c\,
\begin{pmatrix}
|\psi\rangle_A \\
|\chi\rangle_B
\end{pmatrix}
+
d\,
\begin{pmatrix}
|\phi\rangle_A \\
|\chi\rangle_B
\end{pmatrix}.
$$

Gerade diese Struktur kann jedoch innerhalb der direkten Summe nicht konsistent realisiert werden: Das bloße „Stapeln“ der Zustände erzwingt stattdessen die Form

$$
\begin{pmatrix}
c\,|\psi\rangle_A + d\,|\phi\rangle_A \\
|\chi\rangle_B
\end{pmatrix},
$$

die sich nicht als Überlagerung zweier unterschiedlicher Zustände des Gesamtsystems interpretieren lässt. Die Superposition bleibt auf System A beschränkt und hebt sich nicht zu einer Superposition von Gesamtzuständen an.



## Lösung: Tensorprodukt der Vektorräume

Ein Ausweg ergibt sich aus einer einfachen, aber grundlegenden physikalischen Überlegung. Wir fragen nach der Wahrscheinlichkeit, bei einer gleichzeitigen Messung der $z$-Komponente **beider** Spins jeweils den positiven Eigenwert zu erhalten. Da die beiden Spins als unabhängig angenommen werden, muss diese gemeinsame Wahrscheinlichkeit gleich dem Produkt der Einzelwahrscheinlichkeiten sein und ist somit gegeben durch

$$
|c^A_{\alpha}\,c^B_{\alpha}|^2.
$$

Völlig analog lassen sich die Wahrscheinlichkeiten für alle übrigen möglichen Kombinationen der Messergebnisse bestimmen, etwa „Spin A positiv, Spin B negativ“ oder „beide negativ“. Diese Ereignisse schließen sich gegenseitig aus und schöpfen alle möglichen Messergebnisse einer gemeinsamen Spinmessung aus.

Diese einfache Forderung hat eine unmittelbare Konsequenz für die mathematische Struktur des Zustandsraums: Der Zustand des zusammengesetzten Systems muss so beschaffen sein, dass seine Komponenten genau diese gemeinsamen Wahrscheinlichkeiten reproduzieren. Dies motiviert, den Zustand des Gesamtsystems durch einen Vektor zu beschreiben, dessen Komponenten aus den **Produkten** der Amplituden der Einzelsysteme bestehen, nämlich

$$
c^A_{\alpha}c^B_{\alpha},\quad
c^A_{\alpha}c^B_{\beta},\quad
c^A_{\beta}c^B_{\alpha},\quad
c^A_{\beta}c^B_{\beta}.
$$

Der Gesamtzustand des Systems A+B ist daher natürlich mit dem folgenden Spaltenvektor zu assoziieren:

$$
|\Psi_{AB}\rangle
=
\begin{pmatrix}
c^A_{\alpha}c^B_{\alpha} \\
c^A_{\alpha}c^B_{\beta} \\
c^A_{\beta}c^B_{\alpha} \\
c^A_{\beta}c^B_{\beta}
\end{pmatrix}.
$$

In dieser Darstellung ist unmittelbar, dass die Betragsquadrate der einzelnen Komponenten genau die Wahrscheinlichkeiten für die jeweils zugehörigen kombinierten Messergebnisse liefern. Gleichzeitig ist der Gesamtzustand normiert, sofern die Einzelzustände normiert sind. Damit erfüllt diese Konstruktion sowohl die probabilistische Interpretation der Quantenmechanik als auch das Superpositionsprinzip und liefert einen konsistenten Zustandsraum für das zusammengesetzte Quantensystem. 


## Tensorprodukt: Verallgemeinerung

Diese Konstruktion erlaubt nun eine präzise formale Fassung der bisher rein physikalisch motivierten Überlegungen. Sie führt zur Definition des sogenannten **Tensorprodukts** zweier Vektorräume. Darunter versteht man einen neuen Vektorraum $V_A \otimes V_B$ zusammen mit einer **bilinearen Abbildung**

$$
\otimes \;:\; V_A \times V_B \;\longrightarrow\; V_A \otimes V_B,
$$

die jedem geordneten Paar von Vektoren $|\psi\rangle_A \in V_A$ und $|\chi\rangle_B \in V_B$ einen Vektor $|\psi\rangle_A \otimes |\chi\rangle_B \in V_A \otimes V_B$ zuordnet. Bilinearität bedeutet dabei, dass die Abbildung in jedem Argument linear ist, also gilt

$$
(c\,|\psi\rangle_A + d\,|\phi\rangle_A)\otimes|\chi\rangle_B
=
c\,|\psi\rangle_A \otimes |\chi\rangle_B
+
d\,|\phi\rangle_A \otimes |\chi\rangle_B,
$$

sowie

$$
|\psi\rangle_A \otimes (c\,|\chi\rangle_B + d\,|\eta\rangle_B)
=
c\,|\psi\rangle_A \otimes |\chi\rangle_B
+
d\,|\psi\rangle_A \otimes |\eta\rangle_B.
$$

Genau diese Eigenschaft stellt sicher, dass Superpositionen der Einzelsysteme sich als Superpositionen von Zuständen des Gesamtsystems widerspiegeln.

Im vorliegenden Beispiel zweier Spin-$\tfrac12$-Systeme ist der Tensorproduktraum vierdimensional. Eine natürliche Basis von $V_A \otimes V_B$ wird durch die Tensorprodukte der Basiszustände der Einzelsysteme gegeben, also durch

$$
|\alpha\rangle_A \otimes |\alpha\rangle_B,\quad
|\alpha\rangle_A \otimes |\beta\rangle_B,\quad
|\beta\rangle_A \otimes |\alpha\rangle_B,\quad
|\beta\rangle_A \otimes |\beta\rangle_B.
$$

In dieser Basis entspricht der zuvor hergeleitete Spaltenvektor mit Komponenten
$c^A_{\alpha}c^B_{\alpha}$,
$c^A_{\alpha}c^B_{\beta}$,
$c^A_{\beta}c^B_{\alpha}$ und
$c^A_{\beta}c^B_{\beta}$
genau dem Tensorproduktzustand
$|\psi_A\rangle \otimes |\psi_B\rangle$.
Damit wird deutlich, dass das Tensorprodukt nicht nur eine mathematische Konstruktion ist, sondern die einzig mögliche lineare Struktur darstellt, die die probabilistische Interpretation und das Superpositionsprinzip der Quantenmechanik für zusammengesetzte Systeme konsistent realisiert.


Diese Definition lässt sich ohne Weiteres auf eine beliebige Anzahl beliebiger Quantensysteme verallgemeinern. Seien dazu $V_1, V_2, \dots, V_N$ die Zustandsräume von $N$ einzelnen Quantensystemen. Dann ist der Zustandsraum des Gesamtsystems durch den Tensorproduktraum

$$
V_1 \otimes V_2 \otimes \cdots \otimes V_N
$$

gegeben, zusammen mit einer multilinearen Abbildung

$$
\otimes \;:\; V_1 \times V_2 \times \cdots \times V_N
\;\longrightarrow\;
V_1 \otimes V_2 \otimes \cdots \otimes V_N,
$$

die jedem Tupel von Zustandsvektoren

$\bigl(|\psi_1\rangle,|\psi_2\rangle,\dots,|\psi_N\rangle\bigr)$

einen zusammengesetzten Zustand

$|\psi_1\rangle \otimes |\psi_2\rangle \otimes \cdots \otimes |\psi_N\rangle$

zuordnet. Multilinearität bedeutet dabei, dass die Abbildung in *jedem* Argument linear ist, sodass Superpositionen in einzelnen Teilsystemen stets zu Superpositionen von Zuständen des Gesamtsystems führen.

Ist für jedes Teilsystem eine Basis $\{|e^{(k)}_i\rangle\}$ von $V_k$ gewählt, so bildet die Menge aller Tensorprodukte
$$
|e^{(1)}_{i_1}\rangle \otimes |e^{(2)}_{i_2}\rangle \otimes \cdots \otimes |e^{(N)}_{i_N}\rangle
$$
eine Basis des Gesamtraums. Die Dimension des zusammengesetzten Zustandsraums ist damit das Produkt der Dimensionen der Einzelsysteme,
$$
\dim\!\left(V_1 \otimes \cdots \otimes V_N\right)
=
\prod_{k=1}^N \dim(V_k).
$$


Die Konstruktion des Tensorproduktraums erlaubt  auch eine systematische Definition von Operatoren für zusammengesetzte Quantensysteme. Sei dazu $\hat A$ ein linearer Operator auf $V_A$ und $\hat B$ ein linearer Operator auf $V_B$. Dann definiert man einen zugehörigen Operator auf dem Tensorproduktraum $V_A \otimes V_B$ zunächst auf Produktzuständen durch

$$
(\hat A \otimes \hat B)\,\bigl(|\psi\rangle_A \otimes |\chi\rangle_B\bigr)
=
(\hat A|\psi\rangle_A)\otimes(\hat B|\chi\rangle_B),
$$

und erweitert diese Vorschrift anschließend durch Linearität auf den gesamten Tensorproduktraum. Diese Definition ist wohldefiniert, da jedes Element von $V_A \otimes V_B$ als Linearkombination von Produktzuständen geschrieben werden kann.

Von besonderer Bedeutung sind Operatoren, die nur auf eines der Teilsysteme wirken. Ein Operator $\hat A$ auf $V_A$ wird in natürlicher Weise zu einem Operator auf dem Gesamtraum fortgesetzt, indem man

$$
\hat A \;\longrightarrow\; \hat A \otimes \mathbb{I}_B
$$

definiert, wobei $\mathbb{I}_B$ der Identitätsoperator auf $V_B$ ist. Entsprechend wirkt ein Operator, der ausschließlich System B betrifft, als

$$
\mathbb{I}_A \otimes \hat B.
$$

Diese Konstruktion stellt sicher, dass Messungen an einem Teilsystem unabhängig vom Zustand des jeweils anderen Teilsystems beschrieben werden können. Insbesondere gilt für Erwartungswerte in Produktzuständen

$$
\langle \psi_A \otimes \psi_B | \hat A \otimes \mathbb{I}_B | \psi_A \otimes \psi_B \rangle
=
\langle \psi_A | \hat A | \psi_A \rangle,
$$

und analog für Operatoren auf System B.

Für $N$ Teilsysteme mit Zustandsräumen $V_1,\dots,V_N$ verallgemeinert sich diese Konstruktion unmittelbar. Ein Operator $\hat A^{(k)}$, der nur auf dem $k$-ten Teilsystem wirkt, wird als

$$
\mathbb{I}_1 \otimes \cdots \otimes \mathbb{I}_{k-1}
\otimes \hat A^{(k)}
\otimes \mathbb{I}_{k+1} \otimes \cdots \otimes \mathbb{I}_N
$$

auf dem Gesamtraum dargestellt. Produkte solcher Operatoren beschreiben Messungen oder Dynamiken, die mehrere Teilsysteme gleichzeitig betreffen. Damit ist auch die Operatorstruktur der Quantenmechanik konsistent auf zusammengesetzte Systeme übertragen.
