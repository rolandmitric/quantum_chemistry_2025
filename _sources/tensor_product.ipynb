Betrachten wir zwei unterscheidbare Spin-$\tfrac12$-Teilchen $A$ und $B$. 
Jedes Teilchen wird für sich durch einen zweidimensionalen komplexen Hilbertraum beschrieben. 
Ein allgemeiner Zustand von $A$ bzw. $B$ lässt sich schreiben als

$$
|\psi_A\rangle = c_{A1}\,|\uparrow\rangle_A + c_{A2}\,|\downarrow\rangle_A,
\qquad
|\psi_B\rangle = c_{B1}\,|\uparrow\rangle_B + c_{B2}\,|\downarrow\rangle_B.
$$

Die komplexen Koeffizienten bestimmen über die Bornsche Regel die Wahrscheinlichkeiten für Messergebnisse.

Die zentrale Frage ist nun, wie der Zustandsraum des Gesamtsystems aus den beiden Einteilchenräumen zu konstruieren ist. 
Eine scheinbar naheliegende Möglichkeit besteht darin, die beiden Zustandsvektoren einfach zu einem größeren Vektor zusammenzufügen,

$$
(c_{A1},c_{A2}) \oplus (c_{B1},c_{B2})
= (c_{A1},c_{A2},c_{B1},c_{B2}),
$$

was mathematisch dem direkten Summenraum der beiden Hilberträume entspricht. 
Physikalisch beschreibt diese Konstruktion jedoch kein zusammengesetztes System, sondern eine Alternative zwischen zwei Systemen. 
Die Komponenten des Vektors stehen für sich gegenseitig ausschließende Möglichkeiten („System $A$ *oder* System $B$“), nicht für gleichzeitige Eigenschaften beider Teilchen.

Der Widerspruch zeigt sich unmittelbar bei Messungen. 
Sind die beiden Teilchen unabhängig präpariert, so muss die Wahrscheinlichkeit für ein gemeinsames Messergebnis faktorisieren. 
Misst man etwa $|\uparrow\rangle_A$ an Teilchen $A$ und $|\downarrow\rangle_B$ an Teilchen $B$, so fordert die Quantenmechanik

$$
P(\uparrow_A,\downarrow_B)=|c_{A1}|^2\,|c_{B2}|^2.
$$

Ein Zustandsraum, der durch direkte Summen entsteht, enthält jedoch keine Amplituden für solche gemeinsamen Ereignisse und kann daher die grundlegenden probabilistischen Anforderungen nicht erfüllen.

Damit ist die zulässige Struktur praktisch festgelegt. 
Für jedes Paar von Messergebnissen $(i,j)$ muss es eine komplexe Amplitude geben, deren Betragsquadrat die gemeinsame Wahrscheinlichkeit liefert. 
Für zwei Spin-$\tfrac12$-Teilchen ergibt sich so ein Zustandsvektor der Form

$$
|\psi_{AB}\rangle \sim
(c_{A1}c_{B1},\,c_{A1}c_{B2},\,c_{A2}c_{B1},\,c_{A2}c_{B2}),
$$

der genau die korrekte Faktorisierung unabhängiger Wahrscheinlichkeiten reproduziert. 
Diese Zuordnung ist zudem linear in jedem Argument und damit vollständig kompatibel mit dem Superpositionsprinzip.

Abstrahiert man von der konkreten Spinbasis, so lassen sich die physikalischen Anforderungen präzise formulieren: 
Die Abbildung, die einem Paar von Einteilchenzuständen einen Zustand des Gesamtsystems zuordnet, muss in jedem Argument linear sein. 
Identifiziert man Zustände ausschließlich aufgrund dieser Bilinearität, so erhält man einen Vektorraum, der von formalen Produkten 
$|\phi\rangle_A\otimes|\chi\rangle_B$ erzeugt wird und die Relationen

$$
(\alpha|\phi\rangle_A+\beta|\phi'\rangle_A)\otimes|\chi\rangle_B
=
\alpha|\phi\rangle_A\otimes|\chi\rangle_B
+
\beta|\phi'\rangle_A\otimes|\chi\rangle_B
$$

sowie die entsprechenden Relationen im zweiten Faktor erfüllt.

Der so konstruierte Raum ist exakt das Tensorprodukt

$$
\mathcal H_A\otimes\mathcal H_B.
$$

Das Tensorprodukt ist damit keine mathematische Zusatzannahme, sondern die einzig mögliche Zustandsraumstruktur, die mit der Linearität der Quantenmechanik und der Forderung konsistenter gemeinsamer Messwahrscheinlichkeiten vereinbar ist. 
Dass dieser Raum Zustände enthält, die sich nicht als Produkte von Einteilchenzuständen schreiben lassen, ist eine unmittelbare Konsequenz dieser physikalisch motivierten Konstruktion und führt zwangsläufig zum Begriff der Verschränkung.


Gern. Im Folgenden ist eine **streng mathematische Formalisierung**, die **explizit aus der physikalischen Motivation heraus** entwickelt wird und **nahtlos in die übliche Definition des Tensorprodukts mündet**. Der Text ist bewusst so geschrieben, dass er direkt im Anschluss an deinen bisherigen Abschnitt stehen kann und den Übergang von Physik zu Mathematik sauber vollzieht.

Wir fassen die bisherigen Überlegungen nun in präzise mathematische Anforderungen. Sei dazu (V_A) der komplexe Zustandsraum des Teilchens (A) und (V_B) der des Teilchens (B). Gesucht ist ein Vektorraum (V_{AB}), der den Zustandsraum des zusammengesetzten Systems beschreibt, zusammen mit einer Abbildung
[
\mu : V_A \times V_B \to V_{AB},
\qquad
(|\psi_A\rangle,|\psi_B\rangle) \mapsto |\psi_A,\psi_B\rangle,
]
die einem Paar von Einteilchenzuständen einen Zustand des Gesamtsystems zuordnet.

Die physikalischen Prinzipien der Quantenmechanik legen die Eigenschaften dieser Abbildung weitgehend fest. Erstens muss sie **bilinear** sein. Linearität in jedem Argument ist zwingend, da Superpositionen von Zuständen eines Teilsystems bei festgehaltenem Zustand des anderen Teilsystems wiederum zulässige Zustände des Gesamtsystems ergeben müssen. Konkret fordert man
[
\mu(\alpha|\phi\rangle_A+\beta|\phi'\rangle_A,|\chi\rangle_B)
=============================================================

\alpha,\mu(|\phi\rangle_A,|\chi\rangle_B)
+
\beta,\mu(|\phi'\rangle_A,|\chi\rangle_B),
]
sowie die entsprechende Linearität im zweiten Argument. Diese Forderung ist nichts anderes als die mathematische Umsetzung der Tatsache, dass Messstatistiken und Dynamik linear von den Zustandsamplituden abhängen.

Zweitens darf der Zustandsraum des Gesamtsystems **keine zusätzlichen Identifikationen enthalten**, die nicht aus dieser Bilinearität folgen. Physikalisch würde jede weitere Identifikation bedeuten, dass unterschiedliche Präparationen desselben Paars von Einteilchenzuständen nicht unterscheidbar wären, obwohl sie unterschiedliche Messstatistiken liefern könnten. Der Zustandsraum des Gesamtsystems muss also der *allgemeinste* Raum sein, der eine bilineare Abbildung aus (V_A\times V_B) erlaubt.

Diese beiden Forderungen lassen sich nun zu einer universellen Charakterisierung zusammenfassen. Gesucht ist ein Paar ((V_{AB},\mu)) mit der Eigenschaft, dass für jeden komplexen Vektorraum (W) und jede bilineare Abbildung
[
b : V_A \times V_B \to W
]
eine eindeutig bestimmte lineare Abbildung
[
\tilde b : V_{AB} \to W
]
existiert, so dass
[
b = \tilde b \circ \mu
]
gilt. In Diagrammform:
[
\begin{array}{ccc}
V_A \times V_B & \xrightarrow{;\mu;} & V_{AB} \
& \searrow b & \downarrow \tilde b \
& & W
\end{array}
]

Diese universelle Eigenschaft besagt, dass jede bilineare Abbildung aus den Einteilchenräumen eindeutig über eine lineare Abbildung aus dem Zustandsraum des Gesamtsystems faktorisierbar ist. Genau dies kodiert die physikalische Idee, dass alle zulässigen Observablen, Übergangsamplituden und Erwartungswerte des Gesamtsystems bereits vollständig durch lineare Operatoren auf (V_{AB}) beschrieben werden können.

Ein Paar ((V_{AB},\mu)), das diese Eigenschaft erfüllt, ist bis auf kanonische Isomorphie eindeutig bestimmt. Man bezeichnet (V_{AB}) als das **Tensorprodukt** der Vektorräume (V_A) und (V_B) und schreibt
[
V_{AB} \equiv V_A \otimes V_B,
\qquad
\mu(|\phi\rangle_A,|\chi\rangle_B)=|\phi\rangle_A\otimes|\chi\rangle_B.
]

Explizit lässt sich dieser Raum konstruieren, indem man den freien Vektorraum über der Menge aller formalen Paare ((|\phi\rangle_A,|\chi\rangle_B)) bildet und anschließend genau jene Relationen faktorisiert, die die Bilinearität erzwingen. Die Äquivalenzklassen dieser formalen Produkte sind die Tensoren ( |\phi\rangle_A\otimes|\chi\rangle_B ), und endliche Linearkombinationen solcher Produkte spannen den gesamten Raum (V_A\otimes V_B) auf.

Damit ist das Tensorprodukt nicht als abstrakte algebraische Konstruktion eingeführt, sondern als die mathematisch präzise Realisierung der physikalischen Forderung, unabhängige Quantensysteme so zu kombinieren, dass Superposition, Wahrscheinlichkeitsfaktorisierung und lineare Observablenstruktur erhalten bleiben. Die Existenz von Zuständen, die sich nicht als einfache Produkte schreiben lassen, ist keine zusätzliche Annahme, sondern eine unmittelbare Konsequenz dieser universellen Definition.
