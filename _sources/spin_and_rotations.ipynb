# Spin & Rotationen

Wir möchten nun etwas Intuition darüber gewinnen, weshalb der innere Freiheitsgrad des Elektrons „Spin“ genannt wird und weshalb diese Größe in so enger Beziehung zu Rotationen steht. Schon der Name legt nahe, dass die drei Operatoren:

$$
\hat{S_x},\; \hat{S_y},\; \hat{S_z}
$$

eine Rolle spielen, die an kartesische Komponenten eines Drehimpulses erinnert.

Für einen allgemeinen Spinzustand

$$
|\psi\rangle = c_\alpha|\alpha\rangle + c_\beta|\beta\rangle
$$

liefern die Spinoperatoren drei reele Erwartungswerte:

$$
\begin{aligned}
\langle S_x\rangle &= \frac{\hbar}{2}\langle\psi|\sigma_x|\psi\rangle,\\
\langle S_y\rangle &= \frac{\hbar}{2}\langle\psi|\sigma_y|\psi\rangle,\\
\langle S_z\rangle &= \frac{\hbar}{2}\langle\psi|\sigma_z|\psi\rangle .
\end{aligned}
$$

Die Notation legt nahe, dass diese drei Größen gemeinsam wie die Komponenten eines dreidimensionalen Vektors transformieren sollten. Um dies zu zeigen, untersuchen wir zunächst die Transformation eines gewöhnlichen geometrischen Vektors unter einer Rotation.


### Rotation eines Vektors

Eine Rotation verändert weder Abstände noch Winkel. Mathematisch entspricht dies einem unitären Operator

$$
U:V\to V,\qquad U^\dagger U = \mathbb{1}.
$$

Für einen Vektor

$$
\vec v = v_x\vec e_x + v_y\vec e_y + v_z\vec e_z \in\mathbb{R}^3,
$$

und eine Rotation um den Winkel $ \varphi $ um die $z$-Achse wissen wir:

$$
\begin{aligned}
R\vec e_x &= \cos\varphi\,\vec e_x + \sin\varphi\,\vec e_y,\\
R\vec e_y &= -\sin\varphi\,\vec e_x + \cos\varphi\,\vec e_y,\\
R\vec e_z &= \vec e_z.
\end{aligned}
$$

Damit folgt

$$
\begin{aligned}
v_x' &= \cos\varphi\, v_x - \sin\varphi\, v_y,\\
v_y' &= \sin\varphi\, v_x + \cos\varphi\, v_y,\\
v_z' &= v_z.
\end{aligned}
$$

In Matrixform:

$$
\begin{pmatrix}
v_x'\\ v_y'\\ v_z'
\end{pmatrix}
=
\begin{pmatrix}
\cos\varphi & -\sin\varphi & 0\\
\sin\varphi & \cos\varphi & 0\\
0 & 0 & 1
\end{pmatrix}
\begin{pmatrix}
v_x\\ v_y\\ v_z
\end{pmatrix}.
$$

Ein Tripel von Größen, das genau so transformiert, nennen wir einen Vektor.

### Rotation eines Spinzustands

Wir möchten nun zeigen, dass eine geeignete unitäre Transformation im zweidimensionalen Spinraum die Erwartungswerte der Spinoperatoren genau wie einen geometrischen Vektor transformiert.

Wir definieren den Rotationsoperator im Spinraum als

$$
R(\varphi) = e^{-i\sigma_z \varphi/2}.
$$

Die transformierten Erwartungswerte lauten

$$
\begin{aligned}
\langle S_x'\rangle &= \frac{\hbar}{2}\langle\psi|\, R^\dagger\sigma_x R\,|\psi\rangle,\\
\langle S_y'\rangle &= \frac{\hbar}{2}\langle\psi|\, R^\dagger\sigma_y R\,|\psi\rangle,\\
\langle S_z'\rangle &= \frac{\hbar}{2}\langle\psi|\, R^\dagger\sigma_z R\,|\psi\rangle.
\end{aligned}
$$

Wir verwenden die Identität

$$
e^{i\sigma_z \varphi/2}
= \cos\!\left(\frac\varphi2\right)\mathbb{1}
+ i\sin\!\left(\frac\varphi2\right)\sigma_z,
$$

und erhalten

$$
\begin{aligned}
\sigma_x e^{i\sigma_z\varphi/2}
&= \cos\!\left(\frac\varphi2\right)\sigma_x
+ \sin\!\left(\frac\varphi2\right)\sigma_y.
\end{aligned}
$$

Daraus folgt nach Multiplikation mit $e^{-i\sigma_z\varphi/2}$:
$$
R^\dagger\sigma_x R
= \cos\varphi\,\sigma_x + \sin\varphi\,\sigma_y .
$$

Damit ist

$$
\langle S_x'\rangle
= \cos\varphi\,\langle S_x\rangle
+ \sin\varphi\,\langle S_y\rangle.
$$

Analog erhält man
$$
\langle S_y'\rangle
= -\sin\varphi\,\langle S_x\rangle
+ \cos\varphi\,\langle S_y\rangle,
$$
und
$$
\langle S_z'\rangle = \langle S_z\rangle.
$$

Wir sehen also: die Operatoren $S_x, S_y, S_z$ bilden gemeinsam einen **Vektoroperator**, denn ihre Erwartungswerte transformieren bei einer Rotation im Spinraum wie ein geometrischer dreidimensionaler Vektor.

### Rotationsoperator im Spinraum

Da
$$
S_z = \frac{\hbar}{2}\sigma_z,
$$
können wir auch schreiben:
$$
R(\varphi)
= e^{-i\sigma_z\varphi/2}
= e^{-\frac{i}{\hbar} S_z \varphi}.
$$


### Eine bemerkenswerte Eigenschaft des Spins

Aus dieser Darstellung folgt unmittelbar
$$
R(2\pi) = e^{-i\sigma_z \pi} = -\mathbb{1}.
$$

Ein Spin-$\tfrac12$-Zustand kehrt daher **erst nach einer Drehung um $4\pi$** vollständig in seinen ursprünglichen Zustand zurück.

Dies ist ein rein quantenmechanisches Phänomen ohne klassische Analogie.

```python
import numpy as np
import ipywidgets as widgets
import matplotlib.pyplot as plt

sigma_z = np.array([[1,0],[0,-1]], dtype=complex)
psi0 = np.array([1,0], dtype=complex)

def U(theta):
    return np.cos(theta/2)*np.eye(2) - 1j*np.sin(theta/2)*sigma_z

def update(theta):
    theta = float(theta)
    psi = U(theta) @ psi0

    overlap = np.vdot(psi0, psi)
    magnitude = np.abs(overlap)
    phase = np.angle(overlap)

    print(f"θ = {theta:.4f} rad")
    print(f"ψ(θ) = {psi}")
    print(f"⟨ψ(0)|ψ(θ)⟩ = {overlap:.4f}")
    print(f"|⟨ψ(0)|ψ(θ)⟩| = {magnitude:.4f}")
    print(f"arg = {phase:.4f} rad   ({phase/np.pi:.3f} π)")

    # Compute full curves from 0 to 4π
    thetas = np.linspace(0, 4*np.pi, 400)
    overlaps = []
    for t in thetas:
        U_t = U(t)
        psi_t = U_t @ psi0
        overlaps.append(np.vdot(psi0, psi_t))

    overlaps = np.array(overlaps)
    mag = np.abs(overlaps)
    ph = np.unwrap(np.angle(overlaps))

    fig, ax = plt.subplots(1,2, figsize=(12,4))

    # Magnitude plot
    ax[0].plot(thetas, mag)
    ax[0].set_title("|⟨ψ(0)|ψ(θ)⟩|")
    ax[0].set_xlabel("θ")
    ax[0].set_ylim(0,1.1)
    ax[0].axvline(2*np.pi, color='gray', linestyle='--')
    ax[0].axvline(4*np.pi, color='gray', linestyle='--')

    # Phase plot
    ax[1].plot(thetas, ph/np.pi)
    ax[1].set_title("Phase / π")
    ax[1].set_xlabel("θ")
    ax[1].axvline(2*np.pi, color='gray', linestyle='--')
    ax[1].axvline(4*np.pi, color='gray', linestyle='--')

    plt.show()


widgets.interact(
    update,
    theta=widgets.FloatSlider(
        min=0, max=4*np.pi, step=0.01, value=0,
        description="θ"
    )
);
```
