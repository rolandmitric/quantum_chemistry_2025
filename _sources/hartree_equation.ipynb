# Optimales Hartree-Produkt und Hartree-Gleichung

In den vorangegangenen Vorlesungen haben wir das Heliumatom mit Hartree- und 
Hylleraas-Ansätzen untersucht, bei denen die Gesamtwellenfunktion aus einem Produkt 
zweier wasserstoffartiger Orbitale bestand, der durch systematisches Hinzufügen 
$r_{12}$-abhängiger Korrelationsterme verbessert werden konnte. Dies motiviert die 
Frage, ob es *optimale* Orbitale gibt, die im Rahmen des einfachsten Produktansatzes 
die Energie minimieren, und wie man diese Orbitale findet. Wir werden sehen, dass diese 
optimalen Orbitale tatsächlich existieren und als Lösung einer nichtlinearen 
Hartree-Gleichung erscheinen, die **selbstkonsistent** gelöst werden muss.

Wir beginnen mit einem allgemeinen Produktansatz

$$
\Psi_H(1,2) = \phi(1)\,\phi(2)\,.
$$

Im Folgenden verwenden wir die Kurzschreibweise

$$
d1 \equiv d^3 r_1, \qquad d2 \equiv d^3 r_2,
$$

also Integrationsmaße über die Ortskoordinaten der Elektronen.

Der elektronische Hamiltonoperator ist

$$
\hat H = \hat h(1) + \hat h(2) + \frac{1}{r_{12}}\,,
$$

mit

$$
\hat h(i)= -\frac{1}{2}\nabla_i^2 - \frac{Z}{r_i}, \qquad Z=2\,.
$$

Der Erwartungswert der Energie ergibt sich, wie wir schon gesehen haben, zu

$$
E[\phi]
= 2\int d1\, \phi^*(1)\,\hat h\,\phi(1)
+ \int d1 \int d2\, |\phi(1)|^2\,\frac{1}{r_{12}}\,|\phi(2)|^2\,.
$$

Das Orbital muss normiert sein:

$$
\int d1\,|\phi(1)|^2 = 1\,.
$$

---

## Variation des Energie-Funktionals

Unter der Annahme, dass ein optimales Orbital existiert, können wir nun eine Bedingung 
herleiten, die dieses Orbital erfüllen muss, indem wir *kleine* Variationen um das 
optimum betrachten. Dazu führen wir die folgende Variation durch:

$$
\phi \rightarrow \phi + \delta\phi, \qquad 
\phi^* \rightarrow \phi^* + \delta\phi^* .
$$

Der Einteilchen-Beitrag zur Energie liefert die Variation

$$
\delta E_1
= 2\int d1 \left[
\delta\phi^*(1)\,\hat h\,\phi(1)
+ \phi^*(1)\,\hat h\,\delta\phi(1)
\right].
$$

Für den Elektron–Elektron-Term gilt

$$
E_{ee} = \int d1\int d2\, |\phi(1)|^2\,\frac{1}{r_{12}}\,|\phi(2)|^2\,,
$$

und seine Variation liefert

$$
\delta E_{ee}
= 2\int d1 
\left[\delta\phi^*(1)\phi(1) + \phi^*(1)\delta\phi(1)\right]
\left(
\int d2\,\frac{|\phi(2)|^2}{r_{12}}
\right).
$$

Wir definieren das **Hartree-Potential**

$$
V_H(1)=\int d2\,\frac{|\phi(2)|^2}{r_{12}}\,.
$$

Damit lässt sich die Variation der Gesamtenergie in der Form

$$
\delta E = 2\int d1\,
\left[
\delta\phi^* \left(\hat h\phi + V_H\phi \right)
+ 
\left(\hat h\phi + V_H\phi \right)^* \delta\phi
\right]
$$

schreiben.

---

## Variation der Normierungsbedingung

Da die Variation des optimalen Orbitals die Normierung nicht zerstören darf, muss gelten:

$$
\delta\left(\int d1\,|\phi|^2\right)
= \int d1\,\big[\delta\phi^*\,\phi + \phi^*\,\delta\phi\big].
$$

Die Bestimmung des optimalen Orbitals ist äquivalent zur Minimierung der Energie 
mit der Normierung als Nebenbedingung. Wie wir bereits gesehen haben, lassen sich solche 
Probleme mithilfe der Methode der Lagrange-Multiplikatoren lösen. Wir definieren das 
Lagrange-Funktional

$$
\mathcal{L}[\phi,\phi^*]
= E[\phi]
- \varepsilon\left(\int d1\,|\phi|^2 -1\right).
$$

Die Variation ergibt

$$
\delta\mathcal L
= \delta E
- \varepsilon \int d1\,\big[\delta\phi^*\,\phi + \phi^*\,\delta\phi\big].
$$

Aus der Variation nach $\delta\phi^*$ folgt

$$
2(\hat h\phi + V_H\phi) - \varepsilon\phi = 0.
$$

---

# Hartree-Gleichung

Damit erhalten wir die Hartree-Gleichung, die die Form des optimalen Orbitals bestimmt:

$$
\require{color}\fcolorbox{red}{#f2d5d3}{
$(\hat h + V_H)\,\phi = \varepsilon\,\phi$
}
$$

In expliziter Form für Helium lautet sie:

$$
\require{color}
\fcolorbox{red}{#f2d5d3}{
$\left(
-\frac{1}{2}\nabla^2
- \frac{Z}{r}
+ \int d^3r'\,\frac{|\phi(\mathbf r')|^2}{|\mathbf r - \mathbf r'|}
\right)\phi(\mathbf r)
= \varepsilon\,\phi(\mathbf r)$.
}
$$


Diese Gleichung ist nichtlinear, da das Hartree-Potential vom Orbital selbst abhängt. 
Die optimalen Orbitale erhält man daher erst durch eine selbstkonsistente Lösung 
(SCF-Verfahren): man wählt ein Startorbital, berechnet daraus $V_H$, löst die Gleichung 
für ein neues Orbital und wiederholt dies bis zur Konvergenz.

## Zur physikalischen Bedeuting des Hartree-Potentials

<div style="background-color:#f2d5d3; padding:15px; border-left:5px solid red; border-radius:4px;">

Das Hartree-Potential beschreibt das gemittelte elektrostatische Feld, das ein Elektron
durch die räumliche Verteilung der übrigen Elektronen erfährt. Anstelle des vollen,
korrelierten Zwei-Elektronen-Problems wird jedes Elektron so behandelt, als bewege es
sich in der kontinuierlichen Ladungswolke mit Dichte $|\phi(\mathbf r)|^2$.  
Damit enthält das effektive Potential bereits die Abschirmung des Kerns
durch die übrigen Elektronen.
</div>

## Lösung der Hartree-Gleichung in der Gauss-Basis

Im nächsten Schritt wollen wir die Hartree-Gleichung nicht mehr direkt im Ortsraum,
sondern in einer endlichen Basis von Einteilchenfunktionen formulieren. Dies erlaubt
eine numerische Lösung, die strukturell genau dem Vorgehen in realen quantenchemischen
Programmen entspricht. Der große Vorteil der gaußschen Basisfunktionen liegt darin,
dass sämtliche für eine quantenchemische Rechnung benötigten Integrale effizient und
vollständig analytisch berechnet werden können.

Als Basis verwenden wir sphärisch symmetrische $s$-artige Gaußfunktionen,
die am Kern zentriert sind. Für Helium genügt aufgrund der
Symmetrie die Betrachtung nur solcher $s$-Funktionen.

Die Basisfunktionen haben die folgende analytische Form:

$$
\chi_\mu(\mathbf r)
=
N_\mu \exp(-\alpha_\mu r^2), 
\qquad r = |\mathbf r|,\quad \mu=1,\dots, M,
$$

wobei $\alpha_\mu > 0$ Exponenten und $N_\mu$ Normierungskonstanten sind. Die Normierungskonstante lautet:

$$
N_\mu 
= \left(\frac{2\alpha_\mu}{\pi}\right)^{3/4}.
$$

Das  Hartree-Orbital entwickeln wir als Linearkombination dieser Basisfunktionen:

$$
\phi(\mathbf r)
=
\sum_{\nu=1}^M c_\nu\,\chi_\nu(\mathbf r),
$$
wobei die Koeffizienten $c_\nu$ variationell zu bestimmen sind.

Die Normierungsbedingung $\int d^3r\,|\phi(\mathbf r)|^2=1$ wird in der Basis zur
Bedingung an die Koeffizienten:

$$
\int d^3r\,\phi^*(\mathbf r)\phi(\mathbf r)
=
\sum_{\mu,\nu} c_\mu^* c_\nu \int d^3r\,\chi_\mu(\mathbf r)\chi_\nu(\mathbf r)
=
\sum_{\mu,\nu} c_\mu^* S_{\mu\nu} c_\nu
=1,
$$

mit der Überlappungsmatrix

$$
S_{\mu\nu}
=
\int d^3r\,\chi_\mu(\mathbf r)\chi_\nu(\mathbf r).
$$


### Matrixelemente des Einteilchenoperators

In der Gaußbasis benötigen wir die Matrixelemente

$$
h_{\mu\nu}
=
\int d^3r\,\chi_\mu(\mathbf r)\,\hat h\,\chi_\nu(\mathbf r)
=
T_{\mu\nu} + V_{\mu\nu},
$$

wobei $T_{\mu\nu}$ der kinetischen Energie
$$
T_{\mu\nu}
=
-\frac{1}{2}
\int d^3r\,\chi_\mu(\mathbf r)\,\nabla^2 \chi_\nu(\mathbf r),
$$

und $V_{\mu\nu}$ dem Kernpotential-Anteil

$$
V_{\mu\nu}
=
-\int d^3r\,\chi_\mu(\mathbf r)\,\frac{Z}{r}\,\chi_\nu(\mathbf r).
$$

entsprechen. 

Aufgrund der Kugelsymmetrie der $s$-Gaußfunktionen können alle Integrale auf radiale
Integrale übergeführt und leicht analtyisch gelöst werden. Für die Überlappung erhält man

$$
S_{\mu\nu}
=
4\pi N_\mu N_\nu
\int_0^\infty dr\,r^2 e^{-(\alpha_\mu + \alpha_\nu) r^2}
=
\left( \frac{2\sqrt{\alpha_\mu \alpha_\nu}}{\alpha_\mu+\alpha_\nu} \right)^{3/2}.
$$

Die kinetische Energie lässt sich durch explizites Auswerten des Laplace-Operators 
auf die Gaußfunktion berechnen. Für eine $s$-Gaußfunktion gilt

$$
\nabla^2 e^{-\alpha r^2}
=
\left(-6\alpha + 4\alpha^2 r^2\right)e^{-\alpha r^2},
$$

woraus man nach Einsetzen und Integration erhält:

$$
T_{\mu\nu}
=
\frac{3\alpha_\mu \alpha_\nu}{\alpha_\mu+\alpha_\nu}\,S_{\mu\nu}.
$$

Für das Kernpotential-Integral

$$
V_{\mu\nu}
=
-4\pi Z N_\mu N_\nu
\int_0^\infty dr\, r\, e^{-(\alpha_\mu + \alpha_\nu) r^2}
$$

ergibt die Berechnung des radialen Integrals

$$
V_{\mu\nu}
=
-2Z\sqrt{\frac{\alpha_\mu+\alpha_\nu}{\pi}}\,S_{\mu\nu}.
$$

Damit sind alle benötigten Einteilchen-Matrixelemente $h_{\mu\nu}$ in analytischer Form
bekannt.


### Elektron–Elektron-Wechselwirkung in der Basis

Für Helium und ein einzelnes Orbital $\phi$ ist das Hartree-Potential gegeben durch

$$
V_H(\mathbf r_1)
=
\int d^3r_2\,\frac{|\phi(\mathbf r_2)|^2}{|\mathbf r_1 - \mathbf r_2|}.
$$

Setzen wir die Basisentwicklung 

$$
\phi(\mathbf r)
=
\sum_{\gamma} c_\gamma \chi_\gamma(\mathbf r)
$$

ein, so folgt für die Elektronendichte des Elektrons 2

$$
|\phi(\mathbf r_2)|^2
=
\sum_{\gamma,\delta} c_\gamma c_\delta^*\,\chi_\gamma(\mathbf r_2)\chi_\delta(\mathbf r_2).
$$

Damit können wir die Matrixelemente des Hartree-Potentials in der Basis berechnen. Diese  sind gegeben durch:

$$
J_{\mu\nu}
=
\int d^3r_1\,\chi_\mu(\mathbf r_1)
\left[
\int d^3r_2\,\frac{|\phi(\mathbf r_2)|^2}{|\mathbf r_1-\mathbf r_2|}
\right]
\chi_\nu(\mathbf r_1).
$$

Einsetzen der oben definiereten Elektronendichte führt auf die **Zwei-Elektronen-Integrale** der Form:

$$
(\mu\nu|\gamma\delta)
=
\int d^3r_1\int d^3r_2\,
\chi_\mu(\mathbf r_1)\chi_\nu(\mathbf r_1)
\frac{1}{|\mathbf r_1-\mathbf r_2|}
\chi_\gamma(\mathbf r_2)\chi_\delta(\mathbf r_2),
$$

so dass wir für die Matrixelemente des Hartree-Potentials den folgenden Ausdruck erhalten:

$$
J_{\mu\nu}
=
\sum_{\gamma,\delta}
(\mu\nu|\gamma\delta)\,c_\gamma c_\delta^*.
$$

Für $s$-Typ-Gaußfunktionen am selben Zentrum mit Exponenten $\alpha_\mu,\alpha_\nu,
\alpha_\gamma,\alpha_\delta$ lassen sich die Integrale $(\mu\nu|\gamma\delta)$ ebenfalls
analytisch berechnen. Verwendet man unnormierte Gaußfunktionen der Form 
$\exp(-\alpha r^2)$, so erhält man für gleiche Zentren den analytischen Ausdruck

$$
(\mu\nu|\gamma\delta)
=
\frac{2\pi^{5/2}}{(\alpha_\mu+\alpha_\nu)(\alpha_\gamma+\alpha_\delta)
\sqrt{\alpha_\mu+\alpha_\nu+\alpha_\gamma+\alpha_\delta}},
$$

die durch Multiplikation mit den entsprechenden Normierungskonstanten $N_\mu,\dots,N_\delta$
auch auf die normierten Funktionen erweitert werden kann. 

### Matrixform der Hartree-Gleichung

Die Hartree-Gleichung  lautet

$$
\big(\hat h + V_H\big)\phi(\mathbf r)
=
\varepsilon\,\phi(\mathbf r).
$$

Setzen wir die Basisentwicklung 
$\phi(\mathbf r) = \sum_\nu c_\nu \chi_\nu(\mathbf r)$ ein und multiplizieren von links
mit $\chi_\mu(\mathbf r)$ und integrieren über den Raum, so erhalten wir

$$
\int d^3r\,\chi_\mu(\mathbf r)\,\hat h\,\phi(\mathbf r)
+
\int d^3r\,\chi_\mu(\mathbf r)\,V_H(\mathbf r)\,\phi(\mathbf r)
=
\varepsilon
\int d^3r\,\chi_\mu(\mathbf r)\,\phi(\mathbf r).
$$

Die Einteilchen-Terme liefern

$$
\int d^3r\,\chi_\mu(\mathbf r)\,\hat h\,\phi(\mathbf r)
=
\sum_\nu c_\nu
\int d^3r\,\chi_\mu(\mathbf r)\,\hat h\,\chi_\nu(\mathbf r)
=
\sum_\nu h_{\mu\nu} c_\nu.
$$

Für den Hartree-Beitrag erhalten wir unter Verwendung der Definition von $J_{\mu\nu}$
und der Expansion von $\phi$

$$
\int d^3r\,\chi_\mu(\mathbf r)\,V_H(\mathbf r)\,\phi(\mathbf r)
=
\sum_\nu c_\nu J_{\mu\nu},
$$

wobei $J_{\mu\nu}$ seinerseits von den Koeffizienten $c_\gamma$ abhängt.

Auf der rechten Seite steht

$$
\varepsilon
\int d^3r\,\chi_\mu(\mathbf r)\,\phi(\mathbf r)
=
\varepsilon
\sum_\nu c_\nu
\int d^3r\,\chi_\mu(\mathbf r)\chi_\nu(\mathbf r)
=
\varepsilon \sum_\nu S_{\mu\nu} c_\nu.
$$

Damit erhalten wir für jedes $\mu$ eine lineare Gleichung der Form

$$
\sum_\nu (h_{\mu\nu} + J_{\mu\nu})\,c_\nu
=
\varepsilon \sum_\nu S_{\mu\nu} c_\nu.
$$

Wir definieren die Hartree-Matrix:

$$
F_{\mu\nu}
=
h_{\mu\nu} + J_{\mu\nu}
=
h_{\mu\nu}
+
\sum_{\gamma,\delta}
(\mu\nu|\gamma\delta)\,c_\gamma c_\delta^*,
$$

so dass die Hartree-Gleichung in der Basis als *verallgemeinertes* Eigenwertproblem erscheint:

$$
\mathbf F\,\mathbf c
=
\varepsilon\,\mathbf S\,\mathbf c.
$$

Hier ist $\mathbf c$ der Koeffizientenvektor, $\mathbf S$ die Überlappungsmatrix und
$\mathbf F$ die von $\mathbf c$ abhängige Hartree-Matrix. Die Nichtlinearität der 
Hartree-Gleichung zeigt sich nun als Nichtlinearität dieses Eigenwertproblems: 
$\mathbf F$ hängt selbst von der gesuchten Eigenlösung $\mathbf c$ ab! Die Lösung der Hartree-Gleichung erfolgt numerisch mit der Methode des selbstkonsistentes Feldes (SCF für *self consistent field*).
