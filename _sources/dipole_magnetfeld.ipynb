# Magnet-Dipol im homogenen Feld 

Wir betrachten eine kreisförmige Schleife mit Radius $R$, in der ein konstanter Strom $I$ fließt.  Die Ebene der Schleife werde um die $x$-Achse um einen Winkel $\theta$ gekippt. Das externe Magnetfeld sei homogen und zeige in $z$-Richtung:

$$
\mathbf B = B\,\mathbf e_z.
$$

Der Normalenvektor der Schleifenebene ergibt sich durch Rotation von $ \mathbf e_z$ um die $x$-Achse:

$$
\mathbf n(\theta) = -\sin\theta\,\mathbf e_y + \cos\theta\,\mathbf e_z.
$$

Einen zweiten Basisvektor in der Ebene bekommen wir als Kreuzprodukt

$$
\mathbf t_2(\theta) = \mathbf n(\theta)\times \mathbf e_x
= \cos\theta\,\mathbf e_y + \sin\theta\,\mathbf e_z.
$$

Damit lässt sich der Ortsvektor eines Punkts der Schleife parametrisieren als

$$
\mathbf r(\varphi)
= R\big[\cos\varphi\,\mathbf e_x
      + \sin\varphi\big(\cos\theta\,\mathbf e_y + \sin\theta\,\mathbf e_z\big)\big].
$$

Das Linienelement der Schleife lautet damit

$$
\mathrm d\mathbf r
= R\Big[-\sin\varphi\,\mathbf e_x
        + \cos\varphi\cos\theta\,\mathbf e_y
        + \cos\varphi\sin\theta\,\mathbf e_z\Big]\mathrm d\varphi.
$$

## Lorentz Kraft auf ein Linienelement

Auf ein Linienelement wirkt die Lorentzkraft

$$
\mathrm d\mathbf F = I\,\mathrm d\mathbf r \times \mathbf B
= I B\,\mathrm d\mathbf r\times\mathbf e_z.
$$

Einsetzen ergibt

$$
\mathrm d\mathbf F(\varphi)
= I B R\Big[\cos\varphi\cos\theta\,\mathbf e_x
            + \sin\varphi\,\mathbf e_y\Big]\mathrm d\varphi.
$$

Wir berechnen nun die Arbeit, die man leisten muss um die Schleifenebene im Magnetfeld um die $x$-Achse zu rotieren. Eine  Rotation um $ \mathbf e_x $ verschiebt jeden Punkt gemäß

$$
\delta\mathbf r = \mathrm d\theta\,(\mathbf e_x\times \mathbf r).
$$

Dies liefert

$$
\delta\mathbf r(\varphi)
= R\sin\varphi\,\mathrm d\theta\,
\big(\cos\theta\,\mathbf e_z - \sin\theta\,\mathbf e_y\big).
$$

Die vom Magnetfeld geleistete Arbeit ist damit

$$
\mathrm dW_\mathrm{mag}
= \int_0^{2\pi} \mathrm d\mathbf F(\varphi)\cdot \delta\mathbf r(\varphi).
$$

Einsetzen beider Ausdrücke ergibt

$$
\mathrm dW_\mathrm{mag}
= -I B R^2 \sin\theta\,\mathrm d\theta
  \int_0^{2\pi}\sin^2\varphi\,\mathrm d\varphi.
$$

Da

$$
\int_0^{2\pi}\sin^2\varphi\,\mathrm d\varphi = \pi,
$$

folgt

$$
\mathrm dW_\mathrm{mag}
= -\pi I B R^2 \sin\theta\,\mathrm d\theta.
$$

Das magnetische Moment einer kreisförmigen Schleife ist

$$
\mu = I\pi R^2.
$$

Damit bekommen wir für die Arbeit das Ergebnis

$$
\mathrm dW_\mathrm{mag}
= -\mu B \sin\theta\,\mathrm d\theta.
$$

## Potentielle Energie

Die potentielle Energie des Dipol–Feld-Systems ändert sich um

$$
\mathrm dU = -\,\mathrm dW_\mathrm{mag}
= \mu B \sin\theta\,\mathrm d\theta.
$$

Integration führt zu

$$
U(\theta) = -\mu B \cos\theta.
$$

Dies ist äquivalent zur kompakten vektoriellen Form

$$
U = -\boldsymbol\mu\cdot\mathbf B,
$$

also dem klassischen Ausdruck für die Energie eines magnetischen Dipols im äußeren Magnetfeld.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

ex = np.array([1,0,0])
ey = np.array([0,1,0])
ez = np.array([0,0,1])

R = 1.0  # Radius des Ringes

def loop_points(theta):
    n = -np.sin(theta)*ey + np.cos(theta)*ez
    t2 = np.cross(n, ex)

    phi = np.linspace(0, 2*np.pi, 300)
    r = (np.cos(phi)[:,None]*ex 
         + np.sin(phi)[:,None]*t2) * R
    return r

def plot_loop(theta_deg=0):
    theta = np.deg2rad(theta_deg)

    pts = loop_points(theta)

    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot(pts[:,0], pts[:,1], pts[:,2], 'b', lw=2)

    n = -np.sin(theta)*ey + np.cos(theta)*ez
    ax.quiver(0,0,0, n[0], n[1], n[2], color='r', lw=3)

    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    ax.set_zlim(-1.2, 1.2)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")

    plt.show()

interact(plot_loop, theta_deg=(0,180,1))
```

# Warum Spin nur an Magnetfelder koppelt: eine Symmetriebetrachtung

In einem Zweizustandssystem lässt sich jeder hermitesche Operator als
$$
O = a_i \sigma_i
$$
schreiben (Einstein-Summation über $i=1,2,3$). Die Pauli-Matrizen bilden damit einen dreidimensionalen Operatorvektor
$$
\boldsymbol{\sigma} = (\sigma_x,\sigma_y,\sigma_z),
$$
der sich unter Rotationen wie ein Vektoroperator transformiert:
$$
U(R)\,\sigma_i\,U(R)^\dagger = R_{ij}\,\sigma_j .
$$

Damit ist jede lineare Wechselwirkung eines solchen Systems mit seiner Umwelt notwendig von der Form
$$
H_{\text{int}} = -\mathbf{K}\cdot \boldsymbol{\sigma},
$$
wobei $\mathbf{K}$ ein äußerer dreidimensionaler Feldvektor ist. Die Frage ist also: *Welche physikalischen Felder können für $\mathbf{K}$ auftreten?*
Die Antwort ergibt sich elegant aus **Parität**.

---

## 1. Die Pauli-Matrizen als axiale Vektoren

Spin ist eine Darstellung der Drehgruppe. Die Operatoren $\sigma_i$ transformieren wie Drehimpulse und sind daher **axiale Vektoren** (Pseudovektoren). Unter einer Paritätstransformation $P$ gilt:
$$
P\,\boldsymbol{\sigma}\,P^{-1}=+\boldsymbol{\sigma}.
$$

Dies ist dieselbe Transformationseigenschaft wie für den klassischen Drehimpuls
$$
\mathbf{L} = \mathbf{r}\times\mathbf{p}
\qquad\Rightarrow\qquad
P\,\mathbf{L}\,P^{-1} = \mathbf{L}.
$$

---

## 2. Transformationseigenschaften von E- und B-Feldern

Aus der Struktur der Maxwellgleichungen folgt:

- Das elektrische Feld $\mathbf{E}$ ist ein **polarer Vektor**:
  $$
  P\,\mathbf{E}\,P^{-1} = -\mathbf{E}.
  $$

- Das magnetische Feld $\mathbf{B}$ ist ein **axialer Vektor**:
  $$
  P\,\mathbf{B}\,P^{-1} = +\mathbf{B}.
  $$

Damit stehen uns zwei Arten äußerer Felder zur Verfügung, die sich unterschiedlich unter Parität verhalten.

---

## 3. Bedingungen an die Kopplung: Paritätsinvarianz des Hamiltonoperators

Der Hamiltonoperator eines physikalischen Systems muss unter Parität ein Skalar sein:
$$
P\,H_{\text{int}}\,P^{-1} = H_{\text{int}}.
$$

Die einzige lineare Kopplung ist
$$
H_{\text{int}} = -\mathbf{K}\cdot \boldsymbol{\sigma}.
$$

Prüfen wir die beiden Kandidaten für $\mathbf{K}$:

### (a) Kopplung an ein elektrisches Feld
$$
P(\mathbf{E}\cdot \boldsymbol{\sigma})P^{-1}
=
(-\mathbf{E})\cdot(+\boldsymbol{\sigma})
=
-\,(\mathbf{E}\cdot \boldsymbol{\sigma}).
$$

Der Ausdruck ist **ungerade unter Parität**.
Damit wäre die Hamiltonfunktion paritätsbrechend – etwas, das in der fundamentalen Beschreibung eines Spins ohne explizite Paritätsverletzung ausgeschlossen ist.

Damit ist der Term
$$
\mathbf{E}\cdot \boldsymbol{\sigma}
$$
nicht erlaubt.

---

### (b) Kopplung an ein magnetisches Feld
$$
P(\mathbf{B}\cdot \boldsymbol{\sigma})P^{-1}
=
(+\mathbf{B})\cdot(+\boldsymbol{\sigma})
=
\mathbf{B}\cdot\boldsymbol{\sigma}.
$$

Der Ausdruck ist **paritätserhaltend**.
Damit entsteht eine zulässige physikalische Kopplung:
$$
H_{\text{int}} = -\gamma\,\mathbf{B}\cdot\boldsymbol{\sigma}.
$$

Dies ist exakt der Hamiltonoperator des magnetischen Moments eines Spin-$1/2$ in einem äußeren Magnetfeld.

---

## 4. Fazit: Nur Magnetfelder besitzen die richtige Symmetrie

Die Kernidee lässt sich in einem Satz zusammenfassen:

**Da Spin ein axialer Vektor ist und das elektrische Feld ein polarer, kann ein Spin-$1/2$ nur mit einem Feld gleicher Transformationseigenschaft koppeln – dem Magnetfeld.**

Oder kurz:
$$
\boldsymbol{\sigma} \;\text{axial}, \qquad
\mathbf{E}\;\text{polar},\qquad
\mathbf{B}\;\text{axial}.
$$

Nur das axiale Produkt
$$
\mathbf{B}\cdot\boldsymbol{\sigma}
$$
ist ein Skalar und damit ein gültiger Term im Hamiltonoperator eines zweidimensionalen internen Freiheitsgrades.

---

## 5. Didaktische Schlussbemerkung

Dieser Zugang benötigt keine mikroskopischen Modellannahmen und zeigt unmittelbar, warum ein zweidimensionaler interner Freiheitsgrad, der sich wie ein Drehimpuls verhält, nur durch eine **magnetische Dipolkopplung** mit externen Feldern wechselwirken kann. Damit erhält man die Standardform
$$
H_{\text{Zeeman}} = -\boldsymbol{\mu}\cdot\mathbf{B}.
$$
