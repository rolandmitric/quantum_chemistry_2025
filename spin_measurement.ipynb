# Spinmessung

Welche möglichen Ergebnisse erhalten wir, wenn wir den Spin eines Elektrons messen?  
Um diese Frage präzise zu beantworten, müssen wir zunächst klären, was unter einer *Spinmessung* zu verstehen ist. Der Spin eines Elektrons ist ein quantenmechanischer Vektoroperator mit drei kartesischen Komponenten,

$$
\vec S = S_x\,\vec e_x + S_y\,\vec e_y + S_z\,\vec e_z .
$$

Wie wir bereits wissen, sind die drei Komponenten $S_x$, $S_y$ und $S_z$ nicht miteinander vertauschbar. Dies bedeutet, dass der Spin nicht gleichzeitig in allen drei Raumrichtungen wohldefiniert sein kann. Eine Spinmessung muss sich daher stets auf eine festgelegte Raumrichtung beziehen, also auf die Projektion des Spinoperators entlang einer bestimmten Achse.

Wir wenden uns nun der zentralen Frage zu:

*Was ist das mögliche Ergebnis einer Messung der Spinprojektion in eine beliebige Richtung im Raum?*


### Messrichtung und Projektionsoperator

Eine Richtung im Raum kann durch einen Einheitsvektor $\vec n$ definiert werden,

$$
\vec n = n_x\,\vec e_x + n_y\,\vec e_y + n_z\,\vec e_z .
$$

Da wir annehmen, dass

$$
\vec n \cdot \vec n = n_x^2 + n_y^2 + n_z^2 = 1 ,
$$

können die Komponenten des Einheitsvektors durch zwei Polarkoordinaten parametrisiert werden,

$$
\vec n =
\sin\theta\cos\varphi\,\vec e_x
+ \sin\theta\sin\varphi\,\vec e_y
+ \cos\theta\,\vec e_z .
$$

Um die Projektion des Spins in Richtung $\vec n$ zu messen, definieren wir den Operator

$$
S_{\vec n} = \vec S \cdot \vec n .
$$

Für ein Spin-$\tfrac12$-Teilchen gilt

$$
\vec S = \frac{\hbar}{2}\,\vec\sigma ,
$$

wobei $\vec\sigma = (\sigma_x,\sigma_y,\sigma_z)$ die Pauli-Matrizen sind. Damit folgt

$$
S_{\vec n}
= \frac{\hbar}{2}
\left(
n_x\sigma_x + n_y\sigma_y + n_z\sigma_z
\right).
$$

In der $\sigma_z$-Basis ergibt sich damit die Matrixdarstellung

$$
S_{\vec n}
= \frac{\hbar}{2}
\begin{pmatrix}
n_z & n_x - i n_y \\
n_x + i n_y & -n_z
\end{pmatrix}.
$$

## Bestimmung der Eigenwerte

Nach den Axiomen der Quantenmechanik sind die möglichen Messergebnisse einer Observable genau deren Eigenwerte.  
Wir müssen daher das Eigenwertproblem

$$
(\vec S\cdot\vec n)\,|\vec n,\lambda\rangle
= \lambda\,|\vec n,\lambda\rangle
$$

lösen.

Zur Bestimmung der Eigenwerte betrachten wir das Quadrat des Operators. Mit der bekannten Identität

$$
(\vec\sigma\cdot\vec a)(\vec\sigma\cdot\vec b)
=
(\vec a\cdot\vec b)\,\mathbb{1}
+ i(\vec a\times\vec b)\cdot\vec\sigma
$$

setzen wir

$$
\vec a=\vec n , \qquad \vec b=\vec n .
$$

Damit ergibt sich

$$
(\vec\sigma\cdot\vec n)^2
=
(\vec n\cdot\vec n)\,\mathbb{1}
+ i(\vec n\times\vec n)\cdot\vec\sigma .
$$

Da

$$
\vec n\cdot\vec n = 1,
\qquad
\vec n\times\vec n = 0,
$$

folgt

$$
(\vec\sigma\cdot\vec n)^2 = \mathbb{1}.
$$

Für den Spinoperator ergibt sich damit

$$
(\vec S\cdot\vec n)^2
=
\left(\frac{\hbar}{2}\right)^2 \mathbb{1}.
$$

Das charakteristische Polynom lautet somit

$$
p(\lambda) = \lambda^2 - \left(\frac{\hbar}{2}\right)^2 ,
$$

woraus die Eigenwerte

$$
\lambda = \pm \frac{\hbar}{2}
$$

folgen.

## Bestimmung der Eigenzustände

Wir bestimmen nun explizit die Eigenzustände
$$
|\vec n+\rangle , \qquad |\vec n-\rangle .
$$


### Fall I: Eigenzustand zu $+\hbar/2$

Wir schreiben das Eigenwertproblem explizit,

$$
S_{\vec n}
\begin{pmatrix}
a \\ b
\end{pmatrix}
=
\frac{\hbar}{2}
\begin{pmatrix}
a \\ b
\end{pmatrix}.
$$

Nach Kürzen des Faktors $\hbar/2$ ergibt sich

$$
\begin{pmatrix}
n_z & n_x - i n_y \\
n_x + i n_y & -n_z
\end{pmatrix}
\begin{pmatrix}
a \\ b
\end{pmatrix}
=
\begin{pmatrix}
a \\ b
\end{pmatrix}.
$$

Dies führt auf das lineare Gleichungssystem

$$
\begin{cases}
(n_z - 1)a + (n_x - i n_y)b = 0, \\
(n_x + i n_y)a - (n_z + 1)b = 0.
\end{cases}
$$

Die Determinante der Koeffizientenmatrix lautet

$$
\det
\begin{pmatrix}
n_z - 1 & n_x - i n_y \\
n_x + i n_y & -(n_z + 1)
\end{pmatrix}
=
1 - (n_x^2 + n_y^2 + n_z^2).
$$

Da $\vec n$ normiert ist, verschwindet die Determinante, und eine nichttriviale Lösung existiert.

Wir bestimmen das Verhältnis $b/a$ aus der ersten Gleichung,

$$
\frac{b}{a}
=
\frac{1 - n_z}{n_x - i n_y}.
$$

Mit

$$
n_z = \cos\theta,
\qquad
n_x - i n_y = \sin\theta\,e^{-i\varphi}
$$

folgt

$$
\frac{b}{a}
=
\frac{1-\cos\theta}{\sin\theta}\,e^{i\varphi}.
$$

Unter Verwendung der trigonometrischen Identitäten

$$
1-\cos\theta = 2\sin^2\frac{\theta}{2},
\qquad
\sin\theta = 2\sin\frac{\theta}{2}\cos\frac{\theta}{2}
$$

erhält man

$$
\frac{b}{a}
=
\tan\frac{\theta}{2}\,e^{i\varphi}.
$$

Da der Eigenzustand nur bis auf einen Normierungsfaktor bestimmt ist, wählen wir

$$
a=\cos\frac{\theta}{2},
\qquad
b=e^{i\varphi}\sin\frac{\theta}{2}.
$$

Damit ergibt sich

$$
|\vec n+\rangle
=
\begin{pmatrix}
\cos\frac{\theta}{2} \\
e^{i\varphi}\sin\frac{\theta}{2}
\end{pmatrix}.
$$

### Fall II: Eigenzustand zu $ -\hbar/2 $

Analog lösen wir

$$
S_{\vec n}
\begin{pmatrix}
a \\ b
\end{pmatrix}
=
-\frac{\hbar}{2}
\begin{pmatrix}
a \\ b
\end{pmatrix}.
$$

Nach Kürzen ergibt sich das Gleichungssystem

$$
\begin{cases}
(1+n_z)a + (n_x - i n_y)b = 0, \\
(n_x + i n_y)a + (1-n_z)b = 0.
\end{cases}
$$

Aus der ersten Gleichung folgt

$$
\frac{b}{a}
=
-\frac{1+n_z}{n_x - i n_y}.
$$

Mit

$$
1+\cos\theta = 2\cos^2\frac{\theta}{2}
$$

erhält man

$$
\frac{b}{a}
=
-\frac{\cos\frac{\theta}{2}}{\sin\frac{\theta}{2}}\,e^{i\varphi}.
$$

Wir wählen daher

$$
a=-e^{-i\varphi}\sin\frac{\theta}{2},
\qquad
b=\cos\frac{\theta}{2}.
$$

Damit ergibt sich
$$
|\vec n-\rangle
=
\begin{pmatrix}
-\,e^{-i\varphi}\sin\frac{\theta}{2} \\
\cos\frac{\theta}{2}
\end{pmatrix}.
$$


Die Eigenzustände der Spinprojektion entlang der Richtung $\vec n$ lauten

$$
|\vec n+\rangle
=
\begin{pmatrix}
\cos\frac{\theta}{2} \\
e^{i\varphi}\sin\frac{\theta}{2}
\end{pmatrix},
\qquad
|\vec n-\rangle
=
\begin{pmatrix}
-\,e^{-i\varphi}\sin\frac{\theta}{2} \\
\cos\frac{\theta}{2}
\end{pmatrix}.
$$

## Spezielle Messrichtungen und physikalische Interpretation

Zur Einordnung der allgemeinen Ergebnisse betrachten wir nun einige besonders wichtige Spezialfälle für die Messrichtung $\vec n$. Diese Fälle erlauben eine direkte Verbindung zwischen der abstrakten Operatorformulierung und konkret beobachtbaren Experimenten.


### Fall 1: Messung entlang der $z$-Achse

Wir wählen

$$
\vec n = \vec e_z .
$$

In diesem Fall gilt

$$
S_{\vec n} = \vec S\cdot\vec n = S_z .
$$

Die zugehörigen Parameter sind

$$
\theta = 0, \qquad \varphi = 0 .
$$

Die Eigenzustände des Operators $S_z$ lauten

$$
|z,+\rangle = |\alpha\rangle, \qquad
|z,-\rangle = |\beta\rangle ,
$$

mit den Eigenwerten

$$
+\frac{\hbar}{2}, \qquad -\frac{\hbar}{2}.
$$

**Interpretation.**  
Die Zustände $|\alpha\rangle$ und $|\beta\rangle$ bilden die Standardbasis der Spin-$\tfrac12$-Theorie. Sie entsprechen Spins, die entlang der $z$-Achse vollständig ausgerichtet sind („spin up“ bzw. „spin down“). Dieser Fall dient als Referenz für alle weiteren Messrichtungen.

### Fall 2: Messung entlang der $x$-Achse

Wir wählen

$$
\vec n = \vec e_x ,
$$

so dass

$$
S_{\vec n} = S_x .
$$

Die zugehörigen Winkel sind
$$
\theta = \frac{\pi}{2}, \qquad \varphi = 0 .
$$

Die Eigenzustände ergeben sich zu

$$
|x+\rangle
= \frac{1}{\sqrt2}\left(|z+\rangle + |z-\rangle\right),
$$

$$
|x-\rangle
= \frac{1}{\sqrt2}\left(-|z+\rangle + |z-\rangle\right).
$$

**Interpretation.**  
Die Zustände $|x+\rangle$ und $|x-\rangle$ sind lineare Kombinationen der $S_z$-Eigenzustände und beschreiben Spins, die entlang der $x$-Achse ausgerichtet sind. Befindet sich das System zunächst in einem Eigenzustand von $S_z$, so ist der Spin bezüglich der $x$-Achse nicht wohldefiniert. Eine Messung von $S_x$ liefert daher ein intrinsisch probabilistisches Ergebnis: Mit gleicher Wahrscheinlichkeit von $50\,\%$ erhält man die beiden möglichen Messwerte

$$
+\frac{\hbar}{2}
\quad\text{oder}\quad
-\frac{\hbar}{2}.
$$


### Fall 3: Messung entlang der $y$-Achse

Wir wählen
$$
\vec n = \vec e_y ,
$$
so dass
$$
S_{\vec n} = S_y .
$$
Die zugehörigen Winkel sind
$$
\theta = \frac{\pi}{2}, \qquad \varphi = \frac{\pi}{2} .
$$

Die Eigenzustände lauten
$$
|y+\rangle
= \frac{1}{\sqrt2}\left(|z+\rangle + i|z-\rangle\right),
$$
$$
|y-\rangle
= \frac{1}{\sqrt2}\left(|z+\rangle - i|z-\rangle\right).
$$

**Interpretation.**  
Auch die $S_y$-Eigenzustände sind Linearkombinationen der $S_z$-Eigenzustände, enthalten jedoch komplexe Koeffizienten. 

## Stern-Gerlach Experiment

Die hier diskutierten Spezialfälle besitzen eine direkte experimentelle Bedeutung im Zusammenhang mit dem Stern–Gerlach-Experiment. In einem inhomogenen Magnetfeld erfährt ein Teilchen mit magnetischem Moment $\vec\mu \propto \vec S$ eine Kraft, die proportional zum Gradienten der Feldkomponente entlang der Messrichtung ist. Für ein Feldgradienten entlang der Richtung $\vec n$ ist diese Kraft proportional zu $S_{\vec n}$.

Wir betrachten einen Elektronenstrahl, der sich in $x$-Richtung ausbreitet und ein inhomogenes Magnetfeld durchläuft. Das Magnetfeld wählen wir so, dass die $z$-Komponente  sich linear mit der Koordinate $z$ verändert und ignorieren die anderen Komponenten,

$$
\vec B = B_z(z)\,\vec e_z,
\qquad
B_z(z) = B_0 + G z ,
$$

wobei $G = \partial B_z / \partial z$ der konstante Feldgradient ist. Das magnetische Moment des Elektrons ist proportional zu seinem Spin,

$$
\vec\mu = \gamma {\vec S},
$$

Für die Wechselwirkungsenergie mit dem Magnetfeld ergibt sich daraus:

$$
V(z) = -\,\vec\mu\cdot\vec B = -\,\mu_z B_z(z).
$$

Einsetzen von $B_z(z)$ ergibt

$$
V(z) = -\,\mu_z (B_0 + G z).
$$


Die auf das Elektron wirkende Kraft bekommt man aus dem Gradienten der Energie als

$$
\vec F = -\nabla V.
$$

Da nur eine $z$-Abhängigkeit vorliegt, bleibt

$$
F_z = -\frac{dV}{dz} = \mu_z\,G.
$$

Setzt man den Operator $S_z$ ein, erhält man

$$
F_z = \gamma \frac{G}{\hbar}\,S_z.
$$




Für ein Elektron besitzt der Operator $S_z$ die Eigenwerte

$$
S_z = \pm \frac{\hbar}{2}.
$$

Damit nimmt die Kraft **nur zwei diskrete Werte** an,

$$
F_z = \mp\,\gamma \hbar G/2
$$


Während der Durchquerung des Magnetfeldes wirkt diese konstante Kraft über eine Zeit $T$. Die dadurch erzeugte Impulsänderung beträgt

$$
\Delta p_z = F_z\,T.
$$

Nach dem Austritt aus dem Feld bewegt sich das Elektron frei weiter, wobei der transversale Impuls zu einer räumlichen Ablenkung führt.
Da $F_z$ nur zwei mögliche Werte besitzt, entstehen **zwei räumlich getrennte Teilstrahlen**.

Ein Stern–Gerlach-Experiment entlang der $z$-Achse führt daher zu einer räumlichen Aufspaltung des Strahls in genau zwei Teilstrahlen, die den Eigenwerten

$$
\pm\frac{\hbar}{2}
$$

von $S_z$ entsprechen. Dreht man die Orientierung des Magnetfeldgradienten, so misst das Experiment entsprechend die Eigenwerte von $S_{\vec n}$ für die gewählte Richtung $\vec n$.

Damit liefert das Stern–Gerlach-Experiment eine unmittelbare physikalische Realisierung der hier entwickelten Theorie:  
Unabhängig von der Messrichtung beobachtet man stets genau zwei diskrete Messergebnisse, während die Orientierung der Messachse ausschließlich die zugehörigen Eigenzustände bestimmt.
