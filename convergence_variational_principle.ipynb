# Über die Konvergenz des linearen Variationsprinzips: Verschachtelungs Satz 

Der **Verschachtelungs Satz (auch Satz von MacDonald)** beschreibt das Konvergenzverhalten der variationell bestimmten Energienäherungen, wenn die Größe der verwendeten Basis systematisch erweitert wird. Er liefert die mathematische Grundlage für die Erwartung, dass eine fortlaufende Erweiterung der Basis die angenäherte Grundzustandsenergie erniedrigt und dass die Näherungen für angeregte Zustände ineinander verschachtelt („interlaced“) sind.

Sei $\{\phi_i\}$ eine Menge normierter, linear unabhängiger Basisfunktionen.  
Für $n = 1$ ergibt sich wie wir schon an Beispielen gesehen haben die einfachste Variationsabschätzung der Energie zu

$$\require{color}\fcolorbox{red}{#f2d5d3}{$\displaystyle
\varepsilon_1^{(1)} = \langle \phi_1 | \hat H | \phi_1 \rangle \ge E_0
$}$$

wobei $E_0$ die exakte Grundzustandsenergie des Systems bezeichnet.  
Das Variationsprinzip garantiert, dass jede normierte Testfunktion eine obere Schranke für $E_0$ liefert.


### Wie ändert sich die Energie des Grundzustandes wenn wir eine zweite Basisfunktion hinzufügen?

Wir erweitern nun die Basis zu $\{\phi_1, \phi_2\}$.  
Sind die Funktionen nicht orthogonal, können sie (z. B. durch das Gram–Schmidt-Verfahren) orthogonalisiert werden, sodass

$$
\langle \phi_1 | \phi_2 \rangle = 0.
$$

Die Variationsenergien erhält man durch Lösung der **säkulären Gleichung**

$$\require{color}\fcolorbox{red}{#f2d5d3}{$\displaystyle
\det
\begin{pmatrix}
\langle \phi_1 | \hat H | \phi_1 \rangle - E & \langle \phi_1 | \hat H | \phi_2 \rangle \\
\langle \phi_2 | \hat H | \phi_1 \rangle & \langle \phi_2 | \hat H | \phi_2 \rangle - E
\end{pmatrix}
= 0
$}$$

Mit den Abkürzungen  
$\varepsilon_1^{(1)} = \langle \phi_1 | \hat H | \phi_1 \rangle$,  
$\varepsilon_2^{(1)} = \langle \phi_2 | \hat H | \phi_2 \rangle$,  
und $H_{12} = \langle \phi_1 | \hat H | \phi_2 \rangle$  
ergibt sich dabei  das quadratische Polynom

$$\require{color}\fcolorbox{red}{#f2d5d3}{$\displaystyle
P^{(2)}(E) = (\varepsilon_1^{(1)} - E)(\varepsilon_2^{(1)} - E) - |H_{12}|^2 = 0
$}$$

Die beiden Wurzeln $\varepsilon_1^{(2)}$ und $\varepsilon_2^{(2)}$ sind die verbesserten Energieabschätzungen.

Für $E = \varepsilon_1^{(1)}$ gilt  
$P^{(2)}(\varepsilon_1^{(1)}) = -|H_{12}|^2 < 0$,  
während für $E \to \pm \infty$ gilt $P^{(2)}(E) \to +\infty$.  
Daraus folgt, dass eine Nullstelle unterhalb und eine oberhalb von $\varepsilon_1^{(1)}$ liegt, also

$$\require{color}\fcolorbox{red}{#f2d5d3}{$\displaystyle
\varepsilon_1^{(2)} \le \varepsilon_1^{(1)} \le \varepsilon_2^{(2)}
$}$$

Die niedrigste Eigenenergie im zweidimensionalen Unterraum liefert somit eine bessere (d. h. tiefere) Näherung von  $E_0$!

Dies können wir nun induktiv fortsetzen. Wir betrachten nun explizit den Fall $n=$
und fügen nun eine dritte orthogonale Funktion $\phi_3$ hinzu.  
Anstatt der ursprünglichen Basis $\{\phi_1, \phi_2, \phi_3\}$ können wir die ersten beiden Funktionen durch ihre „optimalen“ Linearkombinationen ersetzen, die dem bereits diagonalisierten Zweizustandsproblem entsprechen. 
Die effektive Basis lautet dann $\{\tilde\phi_1, \tilde\phi_2, \phi_3\}$, und der Hamiltonoperator hat die Matrixdarstellung

$$
\mathbf{H}^{(3)} =
\begin{pmatrix}
\varepsilon_1^{(2)} & 0 & H_{13} \\
0 & \varepsilon_2^{(2)} & H_{23} \\
H_{13} & H_{23} & \varepsilon_3^{(1)}
\end{pmatrix}.
$$

Die säkuläre Determinante ist nun ein kubisches Polynom:

$$\displaystyle
P^{(3)}(E) =
\det
\begin{pmatrix}
\varepsilon_1^{(2)} - E & 0 & H_{13} \\
0 & \varepsilon_2^{(2)} - E & H_{23} \\
H_{13} & H_{23} & \varepsilon_3^{(1)} - E
\end{pmatrix}
= 0
$$

Durch Ausmultiplizieren erhält man

$$\displaystyle
P^{(3)}(E) =
(\varepsilon_1^{(2)} - E)(\varepsilon_2^{(2)} - E)(\varepsilon_3^{(1)} - E)
- (\varepsilon_1^{(2)} - E)H_{23}^2
- (\varepsilon_2^{(2)} - E)H_{13}^2
$$

Die drei Wurzeln von $P^{(3)}(E)$, bezeichnet mit $\varepsilon_1^{(3)} \le \varepsilon_2^{(3)} \le \varepsilon_3^{(3)}$, liefern die Energieabschätzungen im dreidimensionalen Unterraum.

Aus dem Vorzeichenverlauf von $P^{(3)}(E)$ ergibt sich die **Verschachtelungseigenschaft**

$$\require{color}\fcolorbox{red}{#f2d5d3}{$\displaystyle
\varepsilon_1^{(3)} \le \varepsilon_1^{(2)} \le
\varepsilon_2^{(3)} \le \varepsilon_2^{(2)} \le
\varepsilon_3^{(3)}
$}$$

## Allgemeine Formulierung des Verschachtelungssatzes

Für eine wachsende Folge von $n$-dimensionalen Variationsunterräumen  
$\mathcal{H}_1 \subset \mathcal{H}_2 \subset \dots \subset \mathcal{H}_n$  
gelten für die aus den säkulären Gleichungen erhaltenen Näherungsenergien

$$\require{color}\fcolorbox{red}{#f2d5d3}{$\displaystyle
E_k^{(n+1)} \le E_k^{(n)} \le E_{k+1}^{(n+1)}, \qquad k = 1, \dots, n
$}$$

Jede angenäherte Eigenenergie nimmt also monoton ab (und nähert sich dem exakten Wert von oben), wenn die Basismenge vergrößert wird.  
Das Spektrum für $(n+1)$ Basisfunktionen ist **ineinander verschachtelt** mit dem für $n$ Basisfunktionen.

- Die niedrigste Variationsenergie konvergiert von oben gegen die exakte Grundzustandsenergie:

$$\require{color}\fcolorbox{red}{#f2d5d3}{$\displaystyle
  E_0^{(1)} \ge E_0^{(2)} \ge \dots \ge E_0
$}$$

- Auch die Näherungen für angeregte Zustände konvergieren analog; jede ist nach unten durch die exakte Energie des entsprechenden Zustands und nach oben durch die benachbarte höhere Wurzel des kleineren Unterraums beschränkt.

Damit liefert der Satz von MacDonald eine strenge Begründung für die systematische Verbesserung von Elektronenstrukturmethoden durch hierarchische Basiserweiterungen und stellt das zentrale Prinzip moderner CI-, MCSCF- und variationaler Quantenalgorithmus-Verfahren dar.
