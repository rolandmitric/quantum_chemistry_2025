# Spinzustände eines 2-Elektronensystems

## Spin-Operatoren für ein System aus zwei Elektronen

Wir betrachten nun ein Quantensystem, das aus zwei Elektronen besteht. Jedes Elektron besitzt einen eigenen Spinoperator, den wir mit $\vec S_1$ bzw. $\vec S_2$ bezeichnen. Diese Operatoren sind zunächst auf den jeweiligen Einteilchen-Zustandsräumen definiert. Um Spinmessungen am Gesamtsystem konsistent beschreiben zu können, müssen sie auf den Tensorproduktraum des Zwei-Elektronen-Systems erweitert werden.

Diese Erweiterung erfolgt in natürlicher Weise dadurch, dass ein Operator, der auf einem Teilsystem wirkt, auf dem jeweils anderen Teilsystem als Identitätsoperator wirkt. Für die drei kartesischen Komponenten $i=x,y,z$ definieren wir die entsprechenden Operatoren auf dem Produktraum daher als

$$
\begin{aligned}
S_i^{(1)} &\;\longrightarrow\; S_i^{(1)} \otimes I, \\
S_i^{(2)} &\;\longrightarrow\; I \otimes S_i^{(2)}, \qquad i=x,y,z,
\end{aligned}
$$

wobei $I$ den Einheitsoperator auf dem jeweiligen Einteilchenraum bezeichnet. Diese Definition stellt sicher, dass $S_i^{(1)}$ ausschließlich auf den Spin des ersten Elektrons wirkt, während der Spin des zweiten Elektrons unverändert bleibt, und umgekehrt.

Der Operator für die $i$-te Komponente des Gesamtspins ergibt sich dann als Summe der beiden Beiträge,

$$
S_i = S_i^{(1)} \otimes I + I \otimes S_i^{(2)}, \qquad i=x,y,z.
$$

Insbesondere nimmt der Operator für die $z$-Komponente des Gesamtspins die Form

$$
S_z = S_z^{(1)} \otimes I + I \otimes S_z^{(2)}
$$

an. Entsprechende Ausdrücke gelten analog für $S_x$ und $S_y$. Diese Konstruktion bildet die Grundlage für die Definition des Gesamtspinoperators $\vec S=\vec S_1+\vec S_2$ und erlaubt es, sowohl Einzelspin- als auch Gesamtspin-Eigenschaften im Tensorproduktraum konsistent zu analysieren.

Diesen Operator wenden wir nun explizit auf die Basiszustände des Produktraums an. Für den Zustand mit beiden Spins nach oben ergibt sich

$$
\begin{aligned}
S_z\bigl(|\alpha(1)\rangle \otimes |\alpha(2)\rangle\bigr)
&=
\left(S_z^{(1)} \otimes I + I \otimes S_z^{(2)}\right)
|\alpha(1)\rangle \otimes |\alpha(2)\rangle \\
&=
S_z^{(1)}|\alpha(1)\rangle \otimes I|\alpha(2)\rangle
+
I|\alpha(1)\rangle \otimes S_z^{(2)}|\alpha(2)\rangle \\
&=
\frac{\hbar}{2}\,|\alpha(1)\rangle \otimes |\alpha(2)\rangle
+
\frac{\hbar}{2}\,|\alpha(1)\rangle \otimes |\alpha(2)\rangle \\
&=
\hbar\,|\alpha(1)\rangle \otimes |\alpha(2)\rangle .
\end{aligned}
$$

Für den gemischten Zustand erhält man entsprechend

$$
\begin{aligned}
S_z\bigl(|\alpha(1)\rangle \otimes |\beta(2)\rangle\bigr)
&=
\left(S_z^{(1)} \otimes I + I \otimes S_z^{(2)}\right)
|\alpha(1)\rangle \otimes |\beta(2)\rangle \\
&=
S_z^{(1)}|\alpha(1)\rangle \otimes I|\beta(2)\rangle
+
I|\alpha(1)\rangle \otimes S_z^{(2)}|\beta(2)\rangle \\
&=
\frac{\hbar}{2}\,|\alpha(1)\rangle \otimes |\beta(2)\rangle
-
\frac{\hbar}{2}\,|\alpha(1)\rangle \otimes |\beta(2)\rangle \\
&=
0 \cdot |\alpha(1)\rangle \otimes |\beta(2)\rangle .
\end{aligned}
$$

Völlig analog zeigt man

$$
\begin{aligned}
S_z\bigl(|\beta(1)\rangle \otimes |\alpha(2)\rangle\bigr)
&=
0 \cdot |\beta(1)\rangle \otimes |\alpha(2)\rangle, \\
S_z\bigl(|\beta(1)\rangle \otimes |\beta(2)\rangle\bigr)
&=
-\hbar\,|\beta(1)\rangle \otimes |\beta(2)\rangle .
\end{aligned}
$$

Die Ergebnisse für den $S_z$-Operator lassen sich übersichtlich in der folgenden Tabelle zusammenfassen:

$$
\begin{array}{c|c}
\text{Zustand} & S_z \\
\hline
|\alpha(1)\rangle \otimes |\alpha(2)\rangle & +\hbar \\
|\alpha(1)\rangle \otimes |\beta(2)\rangle & 0 \\
|\beta(1)\rangle \otimes |\alpha(2)\rangle & 0 \\
|\beta(1)\rangle \otimes |\beta(2)\rangle & -\hbar
\end{array}
$$

Wir sehen daraus, dass alle vier Basiszustände des Produktraums Eigenzustände des Operators $S_z$ sind, wobei der zugehörige Eigenwert gerade der Summe der $z$-Spinprojektionen der beiden Elektronen entspricht.


## $S^2$-Operator für zwei Elektronen

Als nächsten Schritt konstruieren wir den Operator $S^2$ für ein System aus zwei Elektronen. Der Gesamtspinoperator ist definiert als

$$
\vec S = \vec S_1 + \vec S_2,
$$

wobei $\vec S_1$ und $\vec S_2$ die Spinoperatoren der beiden Elektronen bezeichnen. Der zugehörige Operator für das Quadrat des Gesamtspins ergibt sich daraus als

$$
S^2 = \vec S \cdot \vec S
= \left(\vec S_1 + \vec S_2\right)\cdot\left(\vec S_1 + \vec S_2\right).
$$

Die Spinoperatoren der einzelnen Elektronen lassen sich komponentenweise schreiben als

$$
\vec S_1
=
S_x^{(1)}\,\vec e_x
+
S_y^{(1)}\,\vec e_y
+
S_z^{(1)}\,\vec e_z,
$$

$$
\vec S_2
=
S_x^{(2)}\,\vec e_x
+
S_y^{(2)}\,\vec e_y
+
S_z^{(2)}\,\vec e_z,
$$

wobei $\vec e_x,\vec e_y,\vec e_z$ die kartesischen Einheitsvektoren im Spinraum bezeichnen.

Wie zuvor diskutiert, wirken die Einzelspinoperatoren auf dem Tensorproduktraum des Zwei-Elektronen-Systems als

$$
S_i^{(1)} \;\longrightarrow\; S_i^{(1)} \otimes I,
\qquad
S_i^{(2)} \;\longrightarrow\; I \otimes S_i^{(2)},
\qquad i=x,y,z.
$$

Damit nimmt der Gesamtspinoperator auf dem Produktraum die explizite Form

$$
\begin{aligned}
\vec S
=&\,
\left(S_x^{(1)} \otimes I + I \otimes S_x^{(2)}\right)\vec e_x \\
&+
\left(S_y^{(1)} \otimes I + I \otimes S_y^{(2)}\right)\vec e_y \\
&+
\left(S_z^{(1)} \otimes I + I \otimes S_z^{(2)}\right)\vec e_z
\end{aligned}
$$

an. Diese Darstellung bildet die Ausgangsbasis für die explizite Berechnung des Operators $S^2$ auf dem Vierdimensionalen Spinraum des Zwei-Elektronen-Systems.


Daraus ergibt sich der Operator $S^2$ in expliziter Form als

$$
\begin{aligned}
S^2
&= S_x S_x + S_y S_y + S_z S_z \\
&=
\left(S_x^{(1)} \otimes I + I \otimes S_x^{(2)}\right)
\left(S_x^{(1)} \otimes I + I \otimes S_x^{(2)}\right) \\
&\quad +
\left(S_y^{(1)} \otimes I + I \otimes S_y^{(2)}\right)
\left(S_y^{(1)} \otimes I + I \otimes S_y^{(2)}\right) \\
&\quad +
\left(S_z^{(1)} \otimes I + I \otimes S_z^{(2)}\right)
\left(S_z^{(1)} \otimes I + I \otimes S_z^{(2)}\right).
\end{aligned}
$$

Nach dem Ausmultiplizieren der einzelnen Terme und Zusammenfassen gleicher Beiträge erhält man den $S^2$-Operator in der kompakten Form

$$
S^2
=
S_1^2 \otimes I
+
I \otimes S_2^2
+
2\left(
S_x^{(1)} \otimes S_x^{(2)}
+
S_y^{(1)} \otimes S_y^{(2)}
+
S_z^{(1)} \otimes S_z^{(2)}
\right),
$$

wobei $S_k^2 = (S_x^{(k)})^2+(S_y^{(k)})^2+(S_z^{(k)})^2$ den Quadratoperator des Einzelspins $k=1,2$ bezeichnet.

Zusätzlich lässt sich der Operator $S^2$ für zwei Spin-$\tfrac12$-Teilchen besonders elegant über den **Permutationsoperator**. Wir definieren den Operator $P_{12}$ durch seine Wirkung auf Produktzustände:

$$
P_{12}\bigl(|u(1)\rangle \otimes |v(2)\rangle\bigr)
=
|v(1)\rangle \otimes |u(2)\rangle.
$$

Dieser Operator ist linear, und er vertauscht die beiden Spins. Um $S^2$ in dieser Sprache zu formulieren, benutzen wir die zuvor hergeleitete Darstellung

$$
S^2
=
S_1^2 \otimes I
+
I \otimes S_2^2
+
2\left(
S_x^{(1)} \otimes S_x^{(2)}
+
S_y^{(1)} \otimes S_y^{(2)}
+
S_z^{(1)} \otimes S_z^{(2)}
\right).
$$

Für Spin-$\tfrac12$ gilt $S_i=\frac{\hbar}{2}\sigma_i$ mit den Pauli-Matrizen $\sigma_i$. Damit wird

$$
S_x^{(1)} \otimes S_x^{(2)}
+
S_y^{(1)} \otimes S_y^{(2)}
+
S_z^{(1)} \otimes S_z^{(2)}
=
\frac{\hbar^2}{4}\sum_{i=x,y,z}\sigma_i \otimes \sigma_i.
$$

Nun zeigt man (und kann leicht durch Nachrechnen in der Basis $|\alpha\alpha\rangle,|\alpha\beta\rangle,|\beta\alpha\rangle,|\beta\beta\rangle$ verifizieren), dass der Permutationsoperator die Identität

$$
P_{12}
=
\frac{1}{2}\left(I\otimes I+\sum_{i=x,y,z}\sigma_i\otimes\sigma_i\right)
$$

erfüllt. Daraus folgt unmittelbar

$$
\sum_{i=x,y,z}\sigma_i\otimes\sigma_i = 2P_{12}-I\otimes I.
$$

Setzen wir dies in $S^2$ ein und verwenden $S_1^2=S_2^2=\frac{3}{4}\hbar^2\,I$ für Spin-$\tfrac12$, so erhalten wir

$$
\begin{aligned}
S^2
&=
\frac{3}{4}\hbar^2\,I\otimes I
+
\frac{3}{4}\hbar^2\,I\otimes I
+
2\cdot \frac{\hbar^2}{4}\left(2P_{12}-I\otimes I\right) \\
&=
\frac{3}{2}\hbar^2\,I\otimes I
+
\frac{\hbar^2}{2}P_{12}
-
\frac{\hbar^2}{4}I\otimes I \\
&=
\hbar^2\left(I\otimes I + P_{12}\right).
\end{aligned}
$$

Damit erhalten wir die besonders kompakte Darstellung

$$
\boxed{
S^2=\hbar^2\left(I\otimes I + P_{12}\right).
}
$$

Die Relation überpüfen wir durch eine symbolische Berechnung.

```python
import sympy as sp
from sympy import I
from sympy import kronecker_product as kron

sigma_x = sp.Matrix([[0,1],[1,0]])
sigma_y = sp.Matrix([[0,-I],[I,0]])
sigma_z = sp.Matrix([[1,0],[0,-1]])
sigma_0 = sp.eye(2)

P = kron(sigma_x,sigma_x)+kron(sigma_y,sigma_y) + kron(sigma_z,sigma_z) + kron(sigma_0,sigma_0)
P/2
```

## Eigenzustände von $S^2$

Wir konstruieren nun die Eigenzustände des Zweielektronen-$S^2$-Operators. Durch die oben gefundene form des Operators reduziert sich die Suche nach seinen Eigenzuständen auf die Bestimmung der Eigenzustände von $P_{12}$. Die Eigenwerte des Permutationsoperator sind $\pm 1$ da $P_{12}^2 = I$.  

Zustände mit Eigenwert $+1$ sind unter Vertauschung symmetrisch, Zustände mit Eigenwert $-1$ antisymmetrisch. Aus

$$
S^2|\Psi\rangle=\hbar^2\left(1+P_{12}\right)|\Psi\rangle
$$

folgt dann unmittelbar

$$
P_{12}|\Psi\rangle=+|\Psi\rangle \;\Rightarrow\; S^2|\Psi\rangle=2\hbar^2|\Psi\rangle,
\qquad
P_{12}|\Psi\rangle=-|\Psi\rangle \;\Rightarrow\; S^2|\Psi\rangle=0\cdot|\Psi\rangle.
$$

Wir konstruieren nun die entsprechenden Eigenzustände. Offensichtlich sind die beiden Zustände $|\alpha(1)\rangle\otimes|\alpha(2)\rangle$ und $|\beta(1)\rangle\otimes|\beta(2)\rangle$ bereits symmetrisch, da eine Vertauschung der beiden Elektronen nichts ändert:

$$
P_{12}\bigl(|\alpha(1)\rangle\otimes|\alpha(2)\rangle\bigr)=|\alpha(1)\rangle\otimes|\alpha(2)\rangle,
\qquad
P_{12}\bigl(|\beta(1)\rangle\otimes|\beta(2)\rangle\bigr)=|\beta(1)\rangle\otimes|\beta(2)\rangle.
$$

Sie sind daher Eigenzustände von $P_{12}$ mit Eigenwert $+1$ und damit Eigenzustände von $S^2$ mit Eigenwert $2\hbar^2$.

Für die beiden gemischten Produktzustände vertauscht $P_{12}$ hingegen die beiden Faktoren:

$$
P_{12}\bigl(|\alpha(1)\rangle\otimes|\beta(2)\rangle\bigr)=|\beta(1)\rangle\otimes|\alpha(2)\rangle,
\qquad
P_{12}\bigl(|\beta(1)\rangle\otimes|\alpha(2)\rangle\bigr)=|\alpha(1)\rangle\otimes|\beta(2)\rangle.
$$

Daraus folgt, dass weder $|\alpha(1)\rangle\otimes|\beta(2)\rangle$ noch $|\beta(1)\rangle\otimes|\alpha(2)\rangle$ Eigenzustände von $P_{12}$ sind. Die passenden Eigenzustände erhält man jedoch durch symmetrische und antisymmetrische Linearkombinationen. Wir definieren

$$
|T_0\rangle=\frac{1}{\sqrt{2}}\left(|\alpha(1)\rangle\otimes|\beta(2)\rangle+|\beta(1)\rangle\otimes|\alpha(2)\rangle\right),
$$

$$
|S_0\rangle=\frac{1}{\sqrt{2}}\left(|\alpha(1)\rangle\otimes|\beta(2)\rangle-|\beta(1)\rangle\otimes|\alpha(2)\rangle\right).
$$

Dann gilt unmittelbar

$$
P_{12}|T_0\rangle
=
\frac{1}{\sqrt{2}}\left(|\beta(1)\rangle\otimes|\alpha(2)\rangle+|\alpha(1)\rangle\otimes|\beta(2)\rangle\right)
=
|T_0\rangle,
$$

$$
P_{12}|S_0\rangle
=
\frac{1}{\sqrt{2}}\left(|\beta(1)\rangle\otimes|\alpha(2)\rangle-|\alpha(1)\rangle\otimes|\beta(2)\rangle\right)
=
-|S_0\rangle.
$$

Damit sind $|T_0\rangle$ und $|S_0\rangle$ Eigenzustände von $P_{12}$ mit Eigenwert $+1$ bzw. $-1$.

Wir fassen damit die Eigenzustände von $S^2$ zusammen. Die drei symmetrischen Zustände (Triplett) sind

$$
|T_+\rangle = |\alpha(1)\rangle\otimes|\alpha(2)\rangle,
\qquad
|T_0\rangle=\frac{1}{\sqrt{2}}\left(|\alpha(1)\rangle\otimes|\beta(2)\rangle+|\beta(1)\rangle\otimes|\alpha(2)\rangle\right),
\qquad
|T_-\rangle = |\beta(1)\rangle\otimes|\beta(2)\rangle,
$$

und sie erfüllen

$$
S^2|T_m\rangle = 2\hbar^2\,|T_m\rangle,\qquad m\in\{+,0,-\}.
$$

Der antisymmetrische Zustand (Singulett) ist

$$
|S_0\rangle=\frac{1}{\sqrt{2}}\left(|\alpha(1)\rangle\otimes|\beta(2)\rangle-|\beta(1)\rangle\otimes|\alpha(2)\rangle\right),
$$

und er erfüllt

$$
S^2|S_0\rangle = 0\cdot|S_0\rangle.
$$

Damit sehen wir: Die Singulett- und Triplettzustände entstehen direkt als antisymmetrische bzw. symmetrische Eigenzustände des Vertauschungsoperators $P_{12}$, und ihre $S^2$-Eigenwerte folgen sofort aus der einfachen Relation $S^2=\hbar^2(I\otimes I+P_{12})$.


<h1> Zusammenfassung </h1>

Aus den vier ungekoppelten Spin-Basisfunktionen eines 2-Elektronensystems lässt sich durch Bildung von Linearkombinationen ein Singulatt und ein Triplett-Zustand (mit drei Komponenten) generieren. Diese Zustände sind gleichzeitig Eigenfunktionen von $S_z$ und $S^2$ für das Gesamtsystem und haben die folgende Form:

\begin{array}{c|c|c|c}
\hline Multiplett & Zustand & S^2/\hbar^2 & S_z/\hbar \\
\hline  Singulett & \frac{1}{\sqrt{2}}(|\alpha(1)\rangle \otimes|\beta(2)\rangle-|\beta(1)\rangle \otimes|\alpha(2)\rangle & 0 & 0 \\
\hline Triplett & |\alpha(1)\rangle \otimes|\alpha(2)\rangle & 2  & +1  \\
  & \frac{1}{\sqrt{2}}(|\alpha(1)\rangle \otimes|\beta(2)\rangle+|\beta(1)\rangle \otimes|\alpha(2)\rangle) & 2  & 0  \\
 & |\beta(1)\rangle \otimes|\beta(2)\rangle & 2  & -1 \\
\hline
\end{array}
