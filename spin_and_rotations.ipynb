# Spin & Rotationen

Wir möchten nun Intuition dafür entwickeln, weshalb der innere Freiheitsgrad des Elektrons als **Spin** bezeichnet wird und in welcher Beziehung diese Größe zu räumlichen Rotationen steht. Bereits die Bezeichnung legt nahe, dass die drei Operatoren

$$
\hat S_x,\; \hat S_y,\; \hat S_z
$$

eine Rolle spielen, die an die kartesischen Komponenten eines Drehimpulses erinnert. Tatsächlich erfüllen sie dieselbe algebraische Struktur wie die Komponenten eines Drehimpulsoperators, wirken jedoch nicht im Ortsraum, sondern im zweidimensionalen Spinraum. Für einen allgemeinen Spinzustand

$$
|\psi\rangle = c_\alpha|\alpha\rangle + c_\beta|\beta\rangle
$$

lassen sich die Erwartungswerte dieser Operatoren berechnen:

$$
\begin{aligned}
\langle S_x\rangle &= \frac{\hbar}{2}\langle\psi|\sigma_x|\psi\rangle,\\
\langle S_y\rangle &= \frac{\hbar}{2}\langle\psi|\sigma_y|\psi\rangle,\\
\langle S_z\rangle &= \frac{\hbar}{2}\langle\psi|\sigma_z|\psi\rangle .
\end{aligned}
$$

Jedem Spinzustand sind damit drei reelle Zahlen zugeordnet. Es liegt daher nahe, diese Erwartungswerte als die Komponenten eines dreidimensionalen Vektors zu interpretieren. Diese Interpretation ist jedoch nur dann physikalisch sinnvoll, wenn sich diese drei Größen unter einer Rotation des Zustands genau so transformieren, wie man es von den Komponenten eines Vektors im dreidimensionalen Raum erwartet. Dies werden wir im Folgenden explizit zeigen.


### Rotation eines dreidimensionalen Vektors

Eine Rotation wird durch einen linearen Operator beschrieben, dessen Wirkung weder Abstände noch Winkel verändert. Mathematisch handelt es sich um eine lineare Isometrie des euklidischen Raums. Betrachten wir einen Vektor mit kartesischen Komponenten in einem orthonormalen Koordinatensystem,

$$
\vec v = v_x\,\vec e_x + v_y\,\vec e_y + v_z\,\vec e_z \in \mathbb{R}^3.
$$

Eine Rotation um den Winkel $\varphi$ um die $z$-Achse ist eindeutig durch ihre Wirkung auf die Basisvektoren festgelegt. Für diese Rotation gilt:

$$
\begin{aligned}
R\,\vec e_x &= \cos\varphi\,\vec e_x + \sin\varphi\,\vec e_y,\\
R\,\vec e_y &= -\sin\varphi\,\vec e_x + \cos\varphi\,\vec e_y,\\
R\,\vec e_z &= \vec e_z.
\end{aligned}
$$

Aus der Linearität des Rotationsoperators folgt unmittelbar, dass sich die Wirkung von $R$ auf einen beliebigen Vektor aus der Wirkung auf die Basisvektoren zusammensetzt. Die Komponenten des Vektors $\vec v$ werden dabei miteinander gemischt, während die Länge des Vektors sowie der Winkel zu anderen Vektoren erhalten bleiben. Für die Komponenten des rotierten Vektors $\vec v' = R\vec v$ ergibt sich:

$$
\begin{aligned}
v_x' &= \cos\varphi\, v_x - \sin\varphi\, v_y,\\
v_y' &= \sin\varphi\, v_x + \cos\varphi\, v_y,\\
v_z' &= v_z.
\end{aligned}
$$

Diese Transformation lässt sich kompakt in Matrixform schreiben:

$$
\begin{pmatrix}
v_x'\\ v_y'\\ v_z'
\end{pmatrix}
=
\begin{pmatrix}
\cos\varphi & -\sin\varphi & 0\\
\sin\varphi & \cos\varphi & 0\\
0 & 0 & 1
\end{pmatrix}
\begin{pmatrix}
v_x\\ v_y\\ v_z
\end{pmatrix}.
$$

Damit können wir dem Rotationsoperator die folgende Matrixdarstellung zuordnen:

$$
R =
\begin{pmatrix}
\cos\varphi & -\sin\varphi & 0\\
\sin\varphi & \cos\varphi & 0\\
0 & 0 & 1
\end{pmatrix}.
$$

Es ist leicht nachzuprüfen, dass diese Rotationsmatrix die Eigenschaft

$$
R^\dagger R = R R^\dagger = \mathbb{1}
$$

erfüllt, wobei $R^\dagger$ die adjungierte Matrix bezeichnet. Matrizen mit dieser Eigenschaft nennt man **unitär**; im rein reellen Fall spricht man von **orthogonalen** Matrizen. Räumliche Rotationen werden daher durch unitäre (bzw. orthogonale) Transformationen beschrieben.

Die Transformationsgesetze der Komponenten eines Vektors unter Rotationen charakterisieren den Begriff des Vektors in der Physik vollständig. Ein Tripel von Größen, das sich unter Rotationen genau in dieser Weise transformiert, nennt man einen **Vektor**. 


### Rotation eines Spinzustands

Wir möchten nun zeigen, dass eine geeignete **unitäre Transformation** im zweidimensionalen Spinraum die Erwartungswerte der Spinoperatoren genau so transformiert wie die Komponenten eines geometrischen Vektors im dreidimensionalen Raum. Dazu müssen wir zunächst präzisieren, was es bedeutet, einen **Spinzustand zu rotieren**.

In Analogie zur Rotation eines Vektors im Ortsraum definieren wir die Rotation eines Spinzustands als eine unitäre Transformation im Spinraum. Für eine Rotation um den Winkel $\varphi$ um die $z$-Achse definieren wir den Rotationsoperator im Spinraum als

$$
R(\varphi) = e^{-\frac{i}{\hbar}\hat S_z \varphi}
= e^{-i\sigma_z\varphi/2}.
$$

Ein Spinzustand transformiert unter dieser Rotation gemäß

$$
|\psi'\rangle = R(\varphi)\,|\psi\rangle.
$$

Wir untersuchen nun, wie sich die Erwartungswerte der Spinoperatoren unter dieser Transformation verändern. Für die transformierten Erwartungswerte gilt

$$
\begin{aligned}
\langle S_x'\rangle &= \frac{\hbar}{2}\langle\psi|\, R^\dagger\sigma_x R\,|\psi\rangle,\\
\langle S_y'\rangle &= \frac{\hbar}{2}\langle\psi|\, R^\dagger\sigma_y R\,|\psi\rangle,\\
\langle S_z'\rangle &= \frac{\hbar}{2}\langle\psi|\, R^\dagger\sigma_z R\,|\psi\rangle.
\end{aligned}
$$

Zur Auswertung verwenden wir die Exponentialdarstellung der Pauli-Matrix

$$
e^{i\sigma_z \varphi/2}
= \cos\!\left(\frac{\varphi}{2}\right)\mathbb{1}
+ i\sin\!\left(\frac{\varphi}{2}\right)\sigma_z.
$$

Mit der Vertauschungsrelation der Pauli-Matrizen erhält man zunächst

$$
\sigma_x\,e^{i\sigma_z\varphi/2}
=
\cos\!\left(\frac{\varphi}{2}\right)\sigma_x
+ \sin\!\left(\frac{\varphi}{2}\right)\sigma_y.
$$

Durch Links-Multiplikation mit $e^{-i\sigma_z\varphi/2}$ folgt daraus

$$
R^\dagger\sigma_x R
= \cos\varphi\,\sigma_x + \sin\varphi\,\sigma_y .
$$

Damit ergibt sich für den transformierten Erwartungswert

$$
\langle S_x'\rangle
= \cos\varphi\,\langle S_x\rangle
+ \sin\varphi\,\langle S_y\rangle.
$$

Eine analoge Rechnung liefert

$$
\langle S_y'\rangle
= -\sin\varphi\,\langle S_x\rangle
+ \cos\varphi\,\langle S_y\rangle,
$$

sowie

$$
\langle S_z'\rangle = \langle S_z\rangle.
$$

Wir sehen somit, dass sich die Erwartungswerte der drei Spinoperatoren unter einer Rotation exakt so transformieren wie die kartesischen Komponenten eines dreidimensionalen Vektors bei einer Rotation um die $z$-Achse. Die Operatoren $\hat S_x,\hat S_y,\hat S_z$ bilden daher gemeinsam einen **Vektoroperator**. Dies rechtfertigt die Interpretation des Spins als eines quantenmechanischen Drehimpulses: Obwohl der Spin kein klassisches Rotationsanalogon besitzt, kodiert seine algebraische Struktur exakt die Transformationsgesetze räumlicher Vektoren unter Rotationen.


### Eine bemerkenswerte Eigenschaft des Spins

Aus der expliziten Darstellung des Rotationsoperators im Spinraum ergibt sich unmittelbar eine zunächst überraschende Eigenschaft von Spinzuständen. Betrachten wir als Spezialfall eine Rotation um den Winkel $2\pi$, so erhalten wir

$$
R(2\pi)
= e^{-\,\frac{i}{\hbar}\hat S_z\,2\pi}
= e^{-i\sigma_z\pi}
= -\mathbb{1}.
$$

Eine Drehung des Spinzustands um $2\pi$ führt also **nicht** auf denselben Zustandsvektor zurück, sondern multipliziert ihn mit dem Faktor $-1$. Erst nach einer weiteren Rotation um $2\pi$, also insgesamt nach einer Drehung um $4\pi$, gilt

$$
R(4\pi) = \mathbb{1},
$$

und der Zustand stimmt wieder exakt mit dem ursprünglichen überein.

Dieses Verhalten ist charakteristisch für Spin-$\tfrac12$-Systeme und besitzt kein klassisches Gegenstück.
