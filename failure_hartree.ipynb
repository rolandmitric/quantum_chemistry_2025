## Das Versagen der Hartree-Methode: Der Fall des Lithiumatoms

Nach der erfolgreichen Behandlung des Heliumatoms im Hartree-Ansatz liegt es nahe,
die Methode auf das Lithiumatom anzuwenden und damit vielleicht bereits erste
Strukturmerkmale des Periodensystems zu erklären. Wir werden jedoch sehen, dass die
Hartree-Methode bei Lithium **grundsätzlich scheitert**.

Der Grund ist **fundamental**: Das Hartree-Produkt besitzt die falsche
Symmetrie und verletzt das Pauli-Prinzip, das wir in den kommenden
Vorlesungen ausführlich behandeln werden!

Um das Versagen zu verdeutlichen, setzen wir alle drei Elektronen in dasselbe
räumliche Orbital. Wir betrachten die normierte wasserstoffartige \(1s\)-Funktion mit Exponent \(Z\),

$$
\phi_{1s}(\mathbf r)
= \left(\frac{Z^3}{\pi}\right)^{1/2}\, e^{-Z r},
\qquad r = |\mathbf r|.
$$

Der elektronische Hamiltonoperator für drei Elektronen lautet in atomaren Einheiten

$$
\hat H
= \sum_{i=1}^3\left(
-\frac12 \nabla_i^2
 - \frac{Z}{r_i}
\right)
+
\sum_{i<j} \frac{1}{r_{ij}}.
$$

Als (bewusst falschen) Versuchszustand verwenden wir das reine Hartree-Produkt

$$
\Psi(\mathbf r_1,\mathbf r_2,\mathbf r_3)
=
\phi_{1s}(\mathbf r_1)\,
\phi_{1s}(\mathbf r_2)\,
\phi_{1s}(\mathbf r_3).
$$

Für die wasserstoffartige \(1s\)-Funktion kennen wir den Erwartungswert des
Einelektronenoperators:

$$
\langle \hat h \rangle_{1s}
=
\left\langle
-\frac12\nabla^2 - \frac{Z}{r}
\right\rangle_{1s}
= -\frac{Z^2}{2}.
$$

Damit ergibt sich für drei Elektronen im selben Orbital

$$
E_{\text{1-el}}
= 3\left(-\frac{Z^2}{2}\right)
= -\frac{3Z^2}{2}.
$$

Die Coulomb-Wechselwirkung zweier Elektronen im selben Orbital beträgt

$$
J = \left\langle \frac{1}{r_{12}} \right\rangle_{1s,1s}
= \frac{5}{8} Z.
$$

Im Lithiumfall gibt es drei mögliche Elektronenpaare, sodass

$$
E_{\text{ee}}
= 3J
= 3 \cdot \frac{5}{8} Z
= \frac{15}{8} Z.
$$

Die Gesamtenergie lautet somit

$$
E(Z)
= -\frac{3Z^2}{2} + \frac{15}{8} Z.
$$

Für Lithium (\(Z=3\)) erhält man explizit

$$
E(3)
= -\frac{27}{2} + \frac{45}{8}
= -\frac{63}{8}
= -7.875\ \text{Hartree}.
$$

Zum Vergleich fassen wir die Werte in einer kleinen Tabelle zusammen:

| Modell / Beschreibung                                 | Energie (Hartree)     |
|--------------------------------------------------------|------------------------|
| **Falsches Hartree-Produkt** (alle 3 e⁻ im selben \(1s\)) | **–7.875**             |
| **Reales Lithiumatom** (experimentell)                  | **≈ –7.478**           |

Auffällig ist, dass das Hartree-Produkt sogar eine **niedrigere**
Energie liefert als der gemessene Grundzustand des realen Lithiumatoms.
Dies scheint auf den ersten Blick dem Variationsprinzip zu widersprechen!
Tatsächlich folgt daraus jedoch, dass der wahre Grundzustand des Li-Atoms
**orthogonal** zum Hartree-Produkt sein muss. Das einfache Hartree-Produkt
liefert also nicht einmal eine qualitativ korrekte Beschreibung des
Dreielektronensystems.
