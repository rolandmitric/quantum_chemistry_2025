## Kets, Bras, Funktionale, Operatoren usw...

### Komplexer Vektorraum der Quantenzustände 

Mathematisch gesehen, sind die quantenmechanischen Zustände aslo **Vektoren in einem komplexen Vektorraum**. Da unsere bisher etwas mysteriöse zweideutige Observable mit zwei Basiszuständen beschrieben werden kann, können wir durch Bildung beliebiger Linearkombinationen der Basiszustände neue Quantenzustände erzeugen: 

$$ \psi = c_\alpha |\alpha \rangle  + c_\beta |\beta \rangle. $$

Bildlich gesprochen, trägt also ein Elektron in ''seiner Seele'' einen zweidimensionalen komplexen Vektorraum! Wir erinnern im folgenden an die formale Definition eines Vektorraumes.

<div style="border:1px solid #1f6feb;border-left-width:8px;background:#f1f8ff;padding:14px 16px;border-radius:8px;">
  <h4 style="margin:0 0 8px 0;">Definition: Vektorraum</h4>
  <p style="margin:0 0 8px 0;">

Sei $ \mathbb{K} $ ein Körper (in der Quantenmechanik typischerweise $ \mathbb{K}=\mathbb{C} $).  
Ein **Vektorraum über $ \mathbb{K} $** besteht aus

1. einer **Menge** $ V $ von Vektoren (genannt Kets $ \lvert\psi\rangle $),
2. einer **Addition von Vektoren**, formal definiert als Abbildung
    
$$
   + : V \times V \longrightarrow V, \qquad 
   (\lvert\psi\rangle,\lvert\phi\rangle) \mapsto \lvert\psi\rangle + \lvert\phi\rangle ,
$$
    
3. einer **Skalarmultiplikation**, definiert als Abbildung  

$$
   \cdot : \mathbb{K} \times V \longrightarrow V,\qquad 
   (a,\lvert\psi\rangle) \mapsto a\,\lvert\psi\rangle ,
$$

so dass für alle $ \lvert\psi\rangle, \lvert\phi\rangle, \lvert\chi\rangle \in V $ und alle $ a,b \in \mathbb{K} $ die folgenden Axiome gelten:

**(1) Kommutativität der Addition**  
    
$$
\lvert\psi\rangle + \lvert\phi\rangle
= \lvert\phi\rangle + \lvert\psi\rangle .
$$

**(2) Assoziativität der Addition**  
    
$$
(\lvert\psi\rangle + \lvert\phi\rangle) + \lvert\chi\rangle
= \lvert\psi\rangle + (\lvert\phi\rangle + \lvert\chi\rangle).
$$

**(3) Existenz des Nullvektors**  
Es existiert ein $ \lvert 0\rangle \in V $ mit  
    
$$
\lvert\psi\rangle + \lvert 0\rangle = \lvert\psi\rangle .
$$

**(4) Existenz additiver Inverser**  
Zu jedem $ \lvert\psi\rangle \in V $ existiert ein $ -\lvert\psi\rangle \in V $ mit 
    
$$
\lvert\psi\rangle + (-\lvert\psi\rangle)=\lvert 0\rangle .
$$

**(5) Assoziativität der Skalarmultiplikation**  
    
$$
a\,(b\,\lvert\psi\rangle)= (ab)\,\lvert\psi\rangle .
$$

**(6) Neutrales Element der Skalarmultiplikation**  
    
$$
1\cdot \lvert\psi\rangle = \lvert\psi\rangle .
$$

**(7) Distributivität über die Vektoraddition** 
    
$$
a(\lvert\psi\rangle + \lvert\phi\rangle)
= a\lvert\psi\rangle + a\lvert\phi\rangle .
$$

**(8) Distributivität über die Skalaraddition**  
    
$$
(a+b)\lvert\psi\rangle
= a\lvert\psi\rangle + b\lvert\psi\rangle .
$$
</div>


Die zwei komplexe Zahlen  $c_\alpha $ and $ c_\beta $ , die den "internen" Zustand des Elektrons vollstädig charakterisieren können wir kompakt durch einen Spaltenvektor darstellen:

$$ \psi\rightarrow\left[\begin{array}{c}
c_{\alpha}\\
c_{\beta}
\end{array}\right]$$

<u> Bemerkung </u> : Einen allgemeinen Vektor in einem zweidimesionalen komplexen Vektorraum bezeichnen man manchmal in der Literatur als **Spinor**.

Diesen Spaltenvektor können wir auch wie folgt zerlegen:

$$\left[\begin{array}{c}
c_{\alpha}\\
c_{\beta}
\end{array}\right]=c_{\alpha}\left[\begin{array}{c}
1\\
0
\end{array}\right]+c_{\beta}\left[\begin{array}{c}
0\\
1
\end{array}\right] $$

darstellen. Daraus sehen wir, dass wir den Basiszuständen $|\alpha \rangle $ und $|\beta \rangle $ auch zwei entsprechende Spaltenvektoren zuordnen können:

$$|\alpha\rangle\rightarrow\left[\begin{array}{c}
1\\
0
\end{array}\right]\quad|\beta\rangle\rightarrow\left[\begin{array}{c}
0\\
1
\end{array}\right]$$



##  Skalarprodukt und quantenmechanische Messungen

Wenn wir quantenmechanische Zustände als Vektoren auffassen, besitzen wir damit noch keinen Zugang zu etwas *Messbarem*. Wir können zwar Zustände kombinieren und neue Zustände bilden, aber wie charakterisieren wir einen einzelnen Zustand? Wir brauchen ein mathematisches Modell mit dem wir Messungen und die Messergebnisse beschreiben können!

Experimentell werden Zustände eines System beschrieben, indem man ihnen Zahlen zuordnet, die dem Ergebnis einer bestimmten Messung entsprechen. Wir benötigen also eine mathematische Struktur, die es erlaubt, einem Zustand eines Quantensystems Zahlen zuzuordnen.

In der elementaren Quantenmechanik haben wir gelernt, dass das Produkt einer Wellenfunktion mit ihrem komplex konjugierten Partner, also $ \psi^\ast(x)\,\psi(x) $, physikalisch einer Wahrscheinlichkeitsdichte für den Ort eines Teilchens entspricht.

Die Definition eines Vektorraums erlaubt es uns, Zustände $ \lvert\psi\rangle $ zu addieren und mit komplexen Zahlen zu skalieren. Aber diese Struktur erlaubt es noch nicht, Produkte von Zuständen zu definieren oder ihnen Zahlen zuzuordnen. Um physikalisch relevante Größen zu berechnen — etwa

- Wahrscheinlichkeitsamplituden $ \langle\phi\mid\psi\rangle $,
- Erwartungswerte von Observablen,
- Übergangswahrscheinlichkeiten zwischen Zuständen,

benötigen wir eine zusätzliche Struktur: eine Möglichkeit, zwei Vektoren zu „paaren“ und daraus eine komplexe Zahl zu erhalten. Dies motiviert die Einführung des **Skalarprodukts** (oder inneren Produkts).

## Definition: Skalarprodukt
<div style="border:1px solid #1f6feb;border-left-width:8px;background:#f1f8ff;padding:14px 16px;border-radius:8px;">
  <h4 style="margin:0 0 8px 0;">Definition: Skalarprodukt</h4>
  <p style="margin:0 0 8px 0;">
Sei $V$ ein komplexer Vektorraum. Ein **Skalarprodukt** auf $ V $ ist eine Abbildung

$$
\langle \,\cdot\,\mid\,\cdot\,\rangle : V \times V \rightarrow \mathbb{C},
$$

die jedem Paar von Vektoren $\lvert\psi\rangle, \lvert\phi\rangle$ in $V$ eine komplexe Zahl $\langle\psi\mid\phi\rangle$ zuordnet und für alle $\lvert\psi\rangle, \lvert\phi\rangle, \lvert\chi\rangle$ in $V$ sowie alle $a\in\mathbb{C}$ folgende Axiome erfüllt:

**(1) Hermitizität:**  
    
$$\langle\psi\mid\phi\rangle = \langle\phi\mid\psi\rangle^\ast$$

**(2) Linearität im zweiten Argument:**  
    
$$\langle\psi\mid(a\,\phi + \chi)\rangle = a\,\langle\psi\mid\phi\rangle + \langle\psi\mid\chi\rangle $$

**(3) Homogenität im zweiten Argument:**  
      
$$\langle\psi\mid(a\,\phi)\rangle = a\,\langle\psi\mid\phi\rangle $$

**(4) Positive Definitheit:**  
    
$$ 
\langle\psi\mid\psi\rangle \ge 0 
$$ 
      
und  
    
$$ 
\langle\psi\mid\psi\rangle = 0 \Longleftrightarrow \lvert\psi\rangle = \lvert 0\rangle 
$$

**Hinweis:** In der Mathematik wird die Linearität häufig im ersten Argument definiert. In der Physik ist es jedoch üblich, sie (wie hier) im zweiten Argument einzuführen. Ein Vektorraum, der mit einem Skalarprodukt ausgestattet ist, heißt **unitärer Raum**. 
</div>

Mit dieser Definition können den Vektorraum der inneren Zustände eines Elektrons zu einem unitären Raum erweitern indem wir ein Skalarprodukt wie folgt einführen. Betrachten wir zwei beliebige Vektoren, die durch folgende Spaltenvektoren dargestellt werden können: 

$$|\psi\rangle\rightarrow\left[\begin{array}{c}
c_{\alpha}\\
c_{\beta}
\end{array}\right]\\
|\phi\rangle\rightarrow\left[\begin{array}{c}
d_{\alpha}\\
d_{\beta}
\end{array}\right] $$

Wir *definieren*  deren Skalarprodukt als :

$$ \langle\phi|\psi\rangle=d_{\alpha}^{*}c_{\alpha}+d_{\beta}^{*}c_{\beta} $$

Diese Definition erfüllt alle Definitionseigenschaften eines Skalarprodukts wie man direkt überprüfen kann:

Betrachten wir einen allgemeinen Zustand $|\psi\rangle=c_\alpha|\alpha\rangle+c_\beta|\beta\rangle$, dem wir in der Matrix-Darstellung den Spaltenvektor

$$
|\psi\rangle \rightarrow\left[\begin{array}{l}
c_\alpha \\
c_\beta
\end{array}\right]
$$

zugeordnet haben. Laut unserer Definition wäre das Skalarprodukt $\langle\psi \mid \psi\rangle$ durch die folgende Beziehung gegeben:

$$
\begin{aligned}
\langle\psi \mid \psi\rangle & =c_\alpha^* c_\alpha+c_\beta^* c_\beta \\
& =\left|c_\alpha^2\right|^2+\left|c_\beta\right|^2
\end{aligned}
$$

Dies ist aber immer positiv und erfüllt damit die Definitionseigenschaft 1. Es gilt also

$$
\langle\psi \mid \psi\rangle \geqslant 0 \quad \forall \psi \text {. }
$$

Des Weiteren ist $\langle\psi \mid \psi\rangle=0$ äquivalent mit $c_\alpha=c_\beta=0$, womit auch die Eigenschaft 2 erfüllt ist. Die Linearitat lässt sich ebenfalls durch direkte Anwendung der Definition überprüfen, so dass wir hiermit eine valide Definition eines Skalarprodukts auf unserem Vektorraum haben!

Betrachten wir einen beliebigen, aber fixierten Vektor $|\phi\rangle \in V$. Mit Hilfe dieses Vektors definieren wir die Abbildung

$$
F_{|\phi\rangle} : V \longrightarrow \mathbb{K}, \qquad 
F_{|\phi\rangle}(|\psi\rangle) := \langle \phi | \psi \rangle .
$$

Diese Zuordnung ordnet jedem Vektor $|\psi\rangle$ die Zahl zu, die durch das Skalarprodukt mit $|\phi\rangle$ entsteht. Wir wollen nun zeigen, dass $F_{|\phi\rangle}$ eine **lineare Abbildung** ist, d.\,h. dass **Additivität** und **Homogenität** erfüllt sind.

Zunächst zeigen wir die **Additivität**. Für beliebige $|\psi\rangle, |\chi\rangle \in V$ gilt

$$
F_{|\phi\rangle}(|\psi\rangle + |\chi\rangle)
= \langle \phi \mid \psi + \chi \rangle .
$$

Aus der Linearität des Skalarprodukts im zweiten Argument folgt

$$
\langle \phi \mid \psi + \chi \rangle
= \langle \phi \mid \psi \rangle + \langle \phi \mid \chi \rangle,
$$

und damit

$$
F_{|\phi\rangle}(|\psi\rangle + |\chi\rangle)
= F_{|\phi\rangle}(|\psi\rangle) + F_{|\phi\rangle}(|\chi\rangle).
$$

Nun prüfen wir die **Homogenität**. Für ein beliebiges $a \in \mathbb{K}$ und $|\psi\rangle \in V$ erhalten wir

$$
F_{|\phi\rangle}(a\,|\psi\rangle)
= \langle \phi \mid a\,\psi \rangle .
$$

Wiederum aus der Linearität des Skalarprodukts im zweiten Argument folgt

$$
\langle \phi \mid a\,\psi \rangle
= a\,\langle \phi \mid \psi \rangle,
$$

also

$$
F_{|\phi\rangle}(a\,|\psi\rangle)
= a\,F_{|\phi\rangle}(|\psi\rangle).
$$

Damit sind sowohl Additivität als auch Homogenität gezeigt, also ist $F_{|\phi\rangle}$ linear.
Solche Abbildungen $F : V \to \mathbb{K}$, die linear sind, bezeichnet man als **lineare Funktionale**. Sie stellen die einfachsten Beispiele für lineare Operatoren dar!

<div style="border:1px solid #1f6feb;border-left-width:8px;background:#f1f8ff;padding:14px 16px;border-radius:8px;">
  <h4 style="margin:0 0 8px 0;">Definition: Dualraum</h4>
  <p style="margin:0 0 8px 0;">
      
Der Dualraum eines Vektorraums $V$ über einem Körper $\mathbb{K}$ ist definiert als die Menge aller linearen Abbildungen von $V$ nach $\mathbb{K}$:
      
$$
V^\ast := \{\, F : V \to \mathbb{K} \mid F \text{ ist linear} \,\}.
$$
      
Zusammen mit der punktweisen Addition und der üblichen Skalarmultiplikation bildet $V^\ast$ selbst wieder einen Vektorraum. Jedes lineare Funktional $F_{|\phi\rangle}$, das durch ein Skalarprodukt mit einem fixierten Vektor entsteht, ist also ein Element des Dualraums.
      
</div>

Wie wir gerade gesehen haben kann man jedem Vektor $|\phi\rangle \in V$ ein lineares Funktional der Form 

$$
F_{|\phi\rangle} : V \to \mathbb{K}, \qquad
F_{|\phi\rangle}(|\psi\rangle) = \langle \phi | \psi \rangle
$$

zuordnen. Die naheliegende Frage ist nun, ob umgekehrt **jedes** lineare Funktional $F \in V^\ast$ auf diese Weise durch einen Vektor $|u\rangle$ beschrieben werden kann, also ob es zu jedem $F$ einen Vektor $|u\rangle$ gibt mit

$$
F(|v\rangle) = \langle u | v \rangle \quad \text{für alle } |v\rangle \in V.
$$

Diese Frage wird durch den folgenden wichtigen Satz beantwortet.

### Riesz’scher Darstellungssatz (endlichdimensionale Version)

Sei $(V,\langle \cdot \mid \cdot \rangle)$ ein endlichdimensionaler unitärer Raum. Dann gilt:

> **Zu jedem linearen Funktional $F \in V^\ast$ existiert genau ein Vektor $|u\rangle \in V$, so dass**

$$
F(|v\rangle) = \langle u \mid v \rangle 
\qquad \text{für alle } |v\rangle \in V.
$$

Der Vektor $|u\rangle$ heißt die **Riesz-Darstellung** des Funktionals $F$.

###  Beweis

Sei $\{\,|e_1\rangle,\dots,|e_n\rangle\,\}$ eine Orthonormalbasis von $V$.  
Jeder Vektor $|v\rangle \in V$ lässt sich eindeutig als Linearkombination dieser Basisvektoren schreiben:

$$
|v\rangle = \sum_{k=1}^n v_k\,|e_k\rangle, 
\qquad v_k = \langle e_k \mid v \rangle.
$$

Wendet man das gegebene lineare Funktional $F$ auf $|v\rangle$ an, so erhält man wegen der Linearität von $F$:

$$
F(|v\rangle)
= F\!\left(\sum_{k=1}^n v_k\,|e_k\rangle\right)
= \sum_{k=1}^n v_k\,F(|e_k\rangle).
$$

Nun wollen wir $F(|v\rangle)$ in der Form eines Skalarprodukts $\langle u \mid v \rangle$ schreiben. Dazu definieren wir den folgenden Vektor:

$$
|u\rangle = \sum_{k=1}^n F(|e_k\rangle)^\ast\,|e_k\rangle.
$$

Damit folgt für jedes $|v\rangle$:
$$
\langle u \mid v \rangle
= \sum_{k=1}^n u_k^\ast\,v_k
= \sum_{k=1}^n F(|e_k\rangle)\,v_k
= F(|v\rangle).
$$

Damit haben wir gezeigt, dass es zu jedem linearen Funktional $F$ tatsächlich einen Vektor $|u\rangle$ gibt, so dass

$$
F(|v\rangle) = \langle u \mid v \rangle 
\qquad \text{für alle } |v\rangle \in V.
$$

### Eindeutigkeit

Die **Eindeutigkeit** von $|u\rangle$ folgt aus der Nichtentartung des Skalarprodukts.  
Angenommen, es gäbe zwei Vektoren $|u\rangle$ und $|u'\rangle$ in $V$ mit

$$
\langle u \mid v \rangle = F(|v\rangle) = \langle u' \mid v \rangle 
\qquad \text{für alle } |v\rangle \in V.
$$

Dann gilt für alle $|v\rangle$:

$$
\langle u - u' \mid v \rangle = 0.
$$

Insbesondere folgt für $|v\rangle = |u - u'\rangle$:

$$
\langle u - u' \mid u - u' \rangle = 0.
$$

Da das Skalarprodukt positiv definit ist, folgt $|u - u'\rangle = 0$, also $|u\rangle = |u'\rangle$.  
Die Darstellung ist somit eindeutig.

Jedem Vektor $|u\rangle$ können wir ein lineares Funktional $F_{|u\rangle}$ zuordnen, und umgekehrt besitzt jedes lineare Funktional $F$ genau eine Riesz-Darstellung $|u\rangle$. Die Abbildung

$$
V \longrightarrow V^\ast, \qquad
|u\rangle \mapsto F_{|u\rangle}, 
\qquad F_{|u\rangle}(|v\rangle) = \langle u \mid v \rangle,
$$

ist damit ein Isomorphismus von Vektorräumen. In der Sprache der Physik motiviert der Rieszsche Satz die Einführung der sog. Dirac-Notation.

### Dirac-Notation

Die Dirac-Notation trennt systematisch zwischen Vektoren und den zugehörigen linearen Funktionalen.  Ein **Ket** $|v\rangle$ bezeichnet dabei einen Vektor in $V$. Durch den Rieszschen Darstellungssatz entspricht jedem Ket ein eindeutig bestimmtes lineares Funktional, das man als **Bra** schreibt:

$$
\langle v| : V \rightarrow \mathbb{C}, 
\qquad
\langle v|\,|w\rangle = \langle v \mid w \rangle.
$$

Der Bra $\langle v|$ ist also nichts anderes als die Riesz-Darstellung des Funktionals 
$F_{|v\rangle}$. Die Bra–Ket-Schreibweise

$$
\langle v|w\rangle
$$

steht damit schlicht für das Skalarprodukt der beiden Vektoren. Auf diese Weise liefert der Rieszsche Darstellungssatz die mathematische Grundlage der Dirac-Notation:  
Er erlaubt, jedem Vektor $|v\rangle$ eindeutig ein „linkswirkendes“ lineares Funktional $\langle v|$ zuzuordnen und so die Dualität zwischen Vektoren und Funktionalen in einer besonders eleganten und physikalisch intuitiven Form zu schreiben. Dies illustrieren wir nun am Beispiel des zweidimensionalen Vektroraumes mit dem wir die inneren Zustände des Elektrons beschreiben. 



Sei $\{\,|\alpha\rangle,|\beta\rangle\,\}$ eine Orthonormalbasis des zweidimensionalen Vektorraums $V$. Jeder Vektor kann dann als

$$
|\psi\rangle = c_\alpha\,|\alpha\rangle + c_\beta\,|\beta\rangle
$$

geschrieben werden. Nach dem Riesz-Theorem korrespondiert hierzu genau ein Bra

$$
\langle\psi|
= c_\alpha^\ast\,\langle\alpha|
+ c_\beta^\ast\,\langle\beta|,
$$

also ein Element des Dualraums $V^\ast$, das durch komplexe Konjugation der Komponenten entsteht.

In Matrixschreibweise entsprechen die Basisvektoren den Spaltenvektoren

$$
|\alpha\rangle \;\leftrightarrow\; 
\begin{bmatrix}1 \\ 0\end{bmatrix},
\qquad
|\beta\rangle \;\leftrightarrow\;
\begin{bmatrix}0 \\ 1\end{bmatrix},
$$

während ihre zugehörigen Bras als Zeilenvektoren dargestellt werden:

$$
\langle\alpha| \;\leftrightarrow\; [\,1 \;\; 0\,], 
\qquad
\langle\beta| \;\leftrightarrow\; [\,0 \;\; 1\,].
$$

Die beiden Zeilenvektoren bilden damit eine Basis des Dualraums $V^\ast$.

In dieser Notation nimmt das Skalarprodukt die Form einer üblichen Matrixmultiplikation an. Für die Basisvektoren erhält man daher

$$
\langle\alpha|\alpha\rangle = [1\;\;0]\begin{bmatrix}1 \\ 0\end{bmatrix} = 1,\qquad
\langle\beta|\beta\rangle = [0\;\;1]\begin{bmatrix}0 \\ 1\end{bmatrix} = 1,
$$

$$
\langle\alpha|\beta\rangle = [1\;\;0]\begin{bmatrix}0 \\ 1\end{bmatrix} = 0,\qquad
\langle\beta|\alpha\rangle = [0\;\;1]\begin{bmatrix}1 \\ 0\end{bmatrix} = 0.
$$

Damit zeigt die Matrixdarstellung anschaulich die Dualität zwischen Kets und Bras:  
Spaltenvektoren repräsentieren Elemente des Vektorraums $V$, während Zeilenvektoren die zugehörigen Funktionale des Dualraums $V^\ast$ darstellen.

## Lineare Operatoren und Observablen

Nachdem wir nun Kets als Vektoren, Bras als zugehörige Funktionale und damit die Dualität zwischen $V$ und $V^\ast$ verstanden haben, wenden wir uns nun den **linearen Operatoren** zu, also den Abbildungen, die Vektoren wieder auf Vektoren abbilden und in der Quantenmechanik die Observablen repräsentieren.

### Definition: Linearer Operator

Ein **linearer Operator** auf einem komplexen Vektorraum $V$ ist eine Abbildung

$$
\hat A : V \longrightarrow V,
$$

die für alle $|v\rangle, |w\rangle \in V$ und alle $a,b \in \mathbb{C}$ die folgende Eigenschaft hat:

$$
\hat A\,(a|v\rangle + b|w\rangle)
= a\,\hat A|v\rangle + b\,\hat A|w\rangle.
$$


### Matrixdarstellung linearer Operatoren in endlichdimensionalen Vektorräumen

Sei $V$ ein $n$-dimensionaler komplexer Vektorraum mit einer Basis $\{\,|i\rangle\,\}_{i=1}^n$. Ein linearer Operator $\hat A: V \to V$ ist vollständig bestimmt durch seine Wirkung auf die Basisvektoren,

$$
\hat A |j\rangle , \qquad j = 1,\dots,n.
$$

Da $\hat A|j\rangle$ wieder ein Element von $V$ ist, kann es eindeutig in der gewählten Basis zerlegt werden,

$$
\hat A |j\rangle = \sum_{i=1}^n A_{ij} |i\rangle.
$$

Die Koeffizienten $A_{ij}\in\mathbb C$ sind die Matrixelemente des Operators $\hat A$ bezüglich der Basis $\{|i\rangle\}$ und definieren die Matrixdarstellung des Operators

$$
\hat A \;\longleftrightarrow\;
\left( A_{ij} \right)_{i,j=1}^n .
$$

Ein beliebiger Vektor $|\psi\rangle\in V$ lässt sich in derselben Basis schreiben als

$$
|\psi\rangle = \sum_{j=1}^n c_j |j\rangle
\quad \leftrightarrow \quad
\begin{bmatrix}
c_1 \\ \vdots \\ c_n
\end{bmatrix}.
$$

Die Wirkung des Operators auf $|\psi\rangle$ ergibt sich aus der Linearität,

$$
\hat A|\psi\rangle
= \sum_{i,j=1}^n A_{ij} c_j |i\rangle,
$$

was exakt der Matrix-Vektor-Multiplikation entspricht,

$$
\hat A|\psi\rangle
\quad\leftrightarrow\quad
\left( A_{ij} \right)
\begin{bmatrix}
c_1 \\ \vdots \\ c_n
\end{bmatrix}.
$$

Damit ist jedem linearen Operator auf einem endlichdimensionalen Vektorraum eindeutig eine Matrix zugeordnet.

### Beispiel: Matrixdarstellung im innerem Raum eines Elektrons

Der Raum der internen Zustände eines Elektrons ist zweidimensional. 
Ein Operator $\hat A$ ist vollständig bestimmt durch seine Wirkung auf die zwei Basiszustände,

$$
\hat A|\alpha\rangle , \qquad \hat A|\beta\rangle.
$$

Die Bilder lassen sich eindeutig in der Basis entwickeln,

$$
\hat A|\alpha\rangle
= A_{\alpha\alpha}|\alpha\rangle
+ A_{\beta\alpha}|\beta\rangle,
$$

$$
\hat A|\beta\rangle
= A_{\alpha\beta}|\alpha\rangle
+ A_{\beta\beta}|\beta\rangle.
$$

Die vier komplexen Zahlen

$$
A_{\alpha\alpha},\; A_{\beta\alpha},\;
A_{\alpha\beta},\; A_{\beta\beta}
$$

sind die Matrixelemente des Operators. Entsprechend ergibt sich die Matrixdarstellung

$$
\hat A \;\longleftrightarrow\;
\begin{pmatrix}
A_{\alpha\alpha} & A_{\alpha\beta} \\
A_{\beta\alpha} & A_{\beta\beta}
\end{pmatrix}.
$$

Ein beliebiger Zustand

$$
|\psi\rangle = c_\alpha|\alpha\rangle + c_\beta|\beta\rangle
\quad \leftrightarrow \quad
\begin{bmatrix}
c_\alpha \\ c_\beta
\end{bmatrix}
$$

wird durch den Operator gemäß

$$
\hat A|\psi\rangle
\quad\leftrightarrow\quad
\begin{pmatrix}
A_{\alpha\alpha} & A_{\alpha\beta} \\
A_{\beta\alpha} & A_{\beta\beta}
\end{pmatrix}
\begin{bmatrix}
c_\alpha \\ c_\beta
\end{bmatrix}
$$

transformiert.  

## Das adjungierte (hermitesche konjugierte) eines Operators

Um Observablen beschreiben zu können, benötigen wir den Begriff des **Adjungierten** eines Operators. Für jeden Operator $\hat A$ definieren wir den adjungierten Operator $\hat A^\dagger$ durch die folgende Bedingung:

$$
\langle u|\hat A v\rangle = \langle \hat A^\dagger u | v\rangle
\qquad\text{für alle } |u\rangle, |v\rangle \in V.
$$

Der adjungierte Operator ist also derjenige, der die Wirkung von $\hat A$ „auf das Bra umwälzt“.  
Diese Gleichung ist jedoch zunächst nur eine Charakterisierung. Es ist nicht offensichtlich, dass ein solcher Operator $\hat A^\dagger$ tatsächlich existiert. Dies folgt aus dem Rieszschen Darstellungssatz.

Sei dazu $|u\rangle \in V$ fest gewählt. Wir betrachten die folgende Abbildung

$$
f_u : V \to \mathbb C, \qquad
f_u(|v\rangle) := \langle u | \hat A v \rangle.
$$

Wie man leicht zeigen kann, ist diese Abbildung linear und damit ein lineares Funktional auf $V$. Nach dem Rieszschen Darstellungssatz existiert zu jedem solchen Funktional ein eindeutig bestimmter Vektor $|w_u\rangle \in V$, so dass

$$
\langle u | \hat A v \rangle = \langle w_u | v \rangle
\qquad\text{für alle } |v\rangle \in V.
$$

Die Zuordnung $|u\rangle \mapsto |w_u\rangle$ ist linear in $|u\rangle$ und definiert somit selbst einen linearen Operator auf $V$. Diesen Operator bezeichnen wir als $\hat A^\dagger$ und schreiben

$$
|w_u\rangle = \hat A^\dagger |u\rangle.
$$

Damit gilt für alle $|u\rangle, |v\rangle \in V$ die Relation

$$
\langle u|\hat A v\rangle = \langle \hat A^\dagger u | v\rangle,
$$

womit die Existenz und Eindeutigkeit des adjungierten Operators gezeigt ist. 


Die Matrixdarstellung dieses Operators erhält man indem man die Matrix des Operators $\hat A$  **konjugiert und transponiert**. In dem Fall des zweidimensionalen Vektorraumes den wir als Beispiel nehme, hätte $ A^\dagger$ die folgende Matrixdarstellung:

$$
\hat A^\dagger \;\longleftrightarrow\;
\begin{pmatrix}
A_{\alpha\alpha} & A_{\alpha\beta} \\
A_{\beta\alpha} & A_{\beta\beta}
\end{pmatrix}^\dagger
=
\begin{pmatrix}
A_{\alpha\alpha}^\ast & A_{\beta\alpha}^\ast \\
A_{\alpha\beta}^\ast & A_{\beta\beta}^\ast
\end{pmatrix}.
$$

## Hermitesche Operatoren (Observablen)

Ein Operator $\hat A$ heißt **hermitesch** (oder selbstadjungiert),
$$
\hat A^\dagger = \hat A.
$$

### Hermitesche Operatoren im zweidimensionalen Vektorraum

Da wir nun die Matrixdarstellung eines Operators $\hat A$ sowie seines Adjungierten $\hat A^\dagger$ kennen, können wir die allgemeine Struktur hermitescher Operatoren im Raum der internen Zustände des Elektrons systematisch analysieren. Ein Operator heißt **hermitesch**, wenn er mit seinem Adjungierten übereinstimmt,
$$
\hat A = \hat A^\dagger.
$$

Ausgedrückt in der Orthonormalbasis $\{|\alpha\rangle,|\beta\rangle\}$ bedeutet diese Bedingung für die Matrixelemente:

$$
A_{\alpha\alpha}=A_{\alpha\alpha}^\ast,\qquad
A_{\beta\beta}=A_{\beta\beta}^\ast,\qquad
A_{\beta\alpha}=A_{\alpha\beta}^\ast.
$$

Daraus folgt unmittelbar, dass die Diagonalelemente reell sein müssen, während die nichtdiagonalen Elemente komplexe Konjugierte voneinander sind. 


Wenn wir nun der Idee von Wolfgang Pauli folgen, müssen wir zur Beschreibung der ''Zweideutigkeit'' einen hermiteschen Operator einführen, der nur zwei mögliche Eigenzustände und Eigenwerte hat. Dieser Operator, den wir zunächst mit $\hat{\sigma}$ bezeichnen, sollte durch die folgende Wirkung auf die Basiszustände definiert sein: 

$$\hat{\sigma}|\alpha\rangle=+1|\alpha\rangle$$

$$ \hat{\sigma}|\beta\rangle=-1|\beta\rangle $$ 

Diesem Operator können wir also in der Matrix-Darstellung die folgende 2x2 Matrix zuordnen:


$$ \hat{\sigma}\rightarrow\left[\begin{array}{cc}
1 & 0\\
0 & -1
\end{array}\right] $$


Aus der allgemeinen Form eines hermiteschen Operators auf einem zweidimensionalen komplexen Vektorraum folgt unmittelbar, dass ein solcher Operator vollständig durch **vier reelle Parameter** beschrieben wird.  

$$
A=
\begin{pmatrix}
a & c - i d \\
c + i d & b
\end{pmatrix},
$$

wobei $a,b,c,d \in \mathbb{R}$ frei wählbar sind. Dies lässt sich besonders elegant darstellen, wenn man die Matrix $A$ als Linearkombination geeigneter Basisoperatoren schreibt. Zunächst trennen wir Diagonal- und Off-Diagonalanteile:

$$
A=
\begin{pmatrix}
a & 0 \\
0 & b
\end{pmatrix}
+
\begin{pmatrix}
0 & c - i d \\
c + i d & 0
\end{pmatrix}.
$$

Nun schreiben wir die Diagonalmatrix als Summe von Spur- und Spur-freiem Anteil:

$$
\begin{pmatrix}
a & 0 \\
0 & b
\end{pmatrix}
=
\frac{a+b}{2}
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
+
\frac{a-b}{2}
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix},
$$

während der nichtdiagonale Teil wie folgt zerlegt werden kann:

$$
\begin{pmatrix}
0 & c - i d \\
c + i d & 0
\end{pmatrix}
=
c\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
+
d\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix}.
$$

Damit lässt sich **jeder** hermitesche Operator im zweidimensionalen Raum als Linearkombination von vier Basisoperatoren schreiben:

$$
A=
\frac{a+b}{2}
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
+
\frac{a-b}{2}
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix}
+
c\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
+
d\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix}.
$$

Zur Vereinfachung, führen wir nun die neuen reellen Parameter

$$
a_0 = \frac{a+b}{2}, \qquad
a_3 = \frac{a-b}{2}, \qquad
a_1 = c, \qquad
a_2 = d
$$
ein.


Mit den neuen Parametern lässt sich der Operator $A$ besonders kompakt schreiben:

$$
A
= a_0\,\underbrace{\begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix}}_{I}
+ a_1\,\underbrace{\begin{pmatrix}0 & 1 \\ 1 & 0\end{pmatrix}}_{\sigma_1}
+ a_2\,\underbrace{\begin{pmatrix}0 & -i \\ i & 0\end{pmatrix}}_{\sigma_2}
+ a_3\,\underbrace{\begin{pmatrix}1 & 0 \\ 0 & -1\end{pmatrix}}_{\sigma_3}.
$$

Damit besitzt jeder hermitesche Operator im zweidimensionalen komplexen Raum eine eindeutige Darstellung als Linearkombination von vier Matrizen. Diese bilden eine Basis des Raumes aller hermiteschen Operatoren:

$$
\begin{aligned}
I &= \begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix}, \\
\sigma_1 &= \begin{pmatrix}0 & 1 \\ 1 & 0\end{pmatrix}, \\
\sigma_2 &= \begin{pmatrix}0 & -i \\ i & 0\end{pmatrix}, \\
\sigma_3 &= \begin{pmatrix}1 & 0 \\ 0 & -1\end{pmatrix}.
\end{aligned}
$$

Die drei Matrizen $\sigma_1, \sigma_2, \sigma_3$ bezeichnet man als **Pauli-Matrizen**.  
Sie spielen eine zentrale Rolle in der Quantenmechanik, da sie die grundlegenden Observablen eines Zweizustandssystems darstellen!


Welchen physikalischen Inhalt haben diese mathematischen Einsichten über die allgemeine Form eines hermiteschen Operators im zweidimensionalen komplexen Vektorraum?

**Jede physikalische Observable** eines Quantensystems, das nur zwei orthonormale Basiszustände besitzt, lässt sich als Linearkombination von genau drei nichttrivialen Basisobservablen darstellen (zusätzlich zum Identitätsoperator $I$). Die Pauli-Annahme, dass es eine physikalische Eigenschaft mit nur zwei möglichen Messwerten gibt, führt daher auf natürlichem Wege nicht nur zu einem einzelnen Operator, sondern zu **drei** eng miteinander verknüpften hermiteschen Operatoren, die durch die drei Pauli-Matrizen repräsentiert werden. Diese drei Matrizen bilden gewissermaßen die *elementarsten*  Observablen, die in einem zweidimensionalen Quantensystem existieren können.

Doch damit stellen sich unmittelbar grundlegende Fragen:  
Welche physikalische Größe wird durch diese drei Operatoren beschrieben?  
Wie lässt sich diese Observable experimentell messen?  
Und ist es überhaupt möglich, alle drei $\sigma$–Operatoren gleichzeitig zu messen, oder schließen sie sich gegenseitig aus?

Die Beantwortung dieser Fragen führt direkt zur Entdeckung einer neuen Form des Drehimpulses, des **Spins**, einer quantenmechanischen Eigenschaft ohne klassisches Analogon. 

<h1> Algebra der Pauli-Matrizen </h1>

Wir untersuchen im Folgenden einige algebraische Eigenschaften der Pauli-Matrizen.

Zunächst stellen wir fest, dass alle drei Pauli-Matrizen hermitesch sind und somit selbst Observablen repräsentieren. Die möglichen Messergebnisse einer quantenmechanischen Observable entsprechen den Eigenwerten des zugehörigen hermiteschen Operators. Es liegt daher nahe zu fragen:

**Welche Eigenwerte besitzen die Operatoren $\sigma_1$, $\sigma_2$ und $\sigma_3$?**

Eine erste wichtige Beobachtung erhält man durch direkte Multiplikation. Für alle drei Pauli-Matrizen gilt:

$$
\sigma_i^2 = I, \qquad i=1,2,3.
$$

Zum Beispiel ergibt die explizite Rechnung für $\sigma_1$:

$$
\sigma_1^2
= 
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
=
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
= I.
$$

Aus der Beziehung $\sigma_i^2 = I$ folgt unmittelbar, dass die Eigenwerte der Pauli-Matrizen nur die Werte $\pm 1$ annehmen können. Dies sieht man wie folgt: Sei $|v\rangle$ ein Eigenvektor von $\sigma_i$ zum Eigenwert $\lambda$,

$$
\sigma_i |v\rangle = \lambda |v\rangle.
$$

Wenden wir erneut $\sigma_i$ an, so ergibt sich

$$
\sigma_i^2 |v\rangle
= \sigma_i(\lambda |v\rangle)
= \lambda\, (\sigma_i |v\rangle)
= \lambda^2 |v\rangle.
$$

Da $\sigma_i^2 = I$ gilt, folgt

$$
I |v\rangle = \lambda^2 |v\rangle,
$$

also

$$
\lambda^2 = 1 \quad \Longrightarrow \quad \lambda = \pm 1.
$$

Damit besitzen alle drei Pauli-Operatoren genau zwei Eigenwerte, nämlich $+1$ und $-1$. 


### Kommutatoren der Pauli-Matrizen

Wir haben gesehen, dass die Pauli-Matrizen hermitesch sind und somit Observablen darstellen. Eine natürliche Frage lautet nun: **Können verschiedene Pauli-Operatoren gleichzeitig exakt gemessen werden?**  
Dies ist genau dann möglich, wenn sie gemeinsame Eigenvektoren besitzen – und das wiederum ist äquivalent dazu, dass die Operatoren miteinander **kommutieren**.  

Um dies zu untersuchen, bestimmen wir den Kommutator zweier Pauli-Matrizen.

Wir beginnen mit dem Produkt $\sigma_1\sigma_2$:

$$
\sigma_1\sigma_2
=
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix}
=
\begin{pmatrix}
i & 0 \\
0 & -i
\end{pmatrix}
= 
i
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix}
= i\,\sigma_3.
$$

Nun berechnen wir das vertauschte Produkt $\sigma_2\sigma_1$:

$$
\sigma_2\sigma_1
=
\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix}
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
=
\begin{pmatrix}
-i & 0 \\
0 & i
\end{pmatrix}
=
-\,i
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix}
= -i\,\sigma_3.
$$

Aus den beiden Rechnungen folgt:

$$
\sigma_1\sigma_2 = i\sigma_3, 
\qquad
\sigma_2\sigma_1 = -i\sigma_3.
$$

Damit ergibt sich für den Kommutator:

$$
[\sigma_1,\sigma_2]
=
\sigma_1\sigma_2 - \sigma_2\sigma_1
=
i\sigma_3 - (-i\sigma_3)
=
2i\,\sigma_3.
$$

Wir sehen also unmittelbar:

Die Pauli-Matrizen kommutieren nicht miteinander. Damit können sie auch **keine gemeinsamen Eigenvektoren** besitzen und sind folglich **nicht gleichzeitig beliebig scharf messbar**.
Vielmehr erfüllen die Die Pauli-Matrizen bis auf eine Konstante dieselben Vertauschungsrelationen wie die Komponenten des quantenmechanischen Drehimpulsoperators. In der Theorie der quantenmechanischen Rotation gilt nämlich

$$
[L_x,\,L_y] = i\hbar\, L_z \quad \text{usw.}
$$

und die Struktur ist identisch (bis auf den Faktor $2$ und den Wert von $\hbar$).

Dieser Vergleich legt den physikalischen Gehalt der Pauli-Matrizen offen:

Die von Pauli ursprünglich postulierte „zweideutige“ Eigenschaft des Elektrons, die er zur Erklärung des atomaren Aufbauprinzips zunächst ad hoc einführte, folgt exakt derselben algebraischen Struktur wie ein **Drehimpuls**. Es handelt sich jedoch um eine neue Form des Drehimpulses, die nicht mit einer räumlichen Rotation eines klassischen Teilchens verknüpft ist und daher kein direktes klassisches Analogon besitzt. Diese intrinsische  urquantenmechanische Eigenschaft des Elektrons bezeichnet man in Anlehnung an den Drehimpuls als **Spin**. Der Spin wird nicht durch anschauliche Modelle, sondern allein durch die Algebra der Pauli-Matrizen eindeutig charakterisiert.


Die Vertauschungsrelationen der Pauli-Matrizen lassen sich vollständig in die Sprache des quantenmechanischen Drehimpulses übertragen, indem wir die folgenden Operatoren einführen:

$$
S_x = \frac{\hbar}{2}\,\sigma_1, 
\qquad 
S_y = \frac{\hbar}{2}\,\sigma_2, 
\qquad 
S_z = \frac{\hbar}{2}\,\sigma_3.
$$

Diese Operatoren unterscheiden sich von den Pauli-Matrizen lediglich durch den sinnvoll gewählten Faktor $\hbar/2$, der sicherstellt, dass die Operatoren dieselbe physikalische Dimension und dieselben Vertauschungsrelationen besitzen wie der gewöhnliche Drehimpuls.

Mit diesen Definitionen erhalten wir beispielsweise:

$$
[S_x,S_y] = \left[\frac{\hbar}{2}\sigma_1,\;\frac{\hbar}{2}\sigma_2\right] \\
= \left(\frac{\hbar}{2}\right)^2[\sigma_1,\sigma_2] \\
= \left(\frac{\hbar}{2}\right)^2(2i\sigma_3) \\
= i\,\frac{\hbar^2}{2}\,\sigma_3 \\
= i\hbar\left(\frac{\hbar}{2}\sigma_3\right) \\
= i\hbar\,S_z .
$$


Damit haben die Operatoren $S_x$, $S_y$ und $S_z$ **exakt dieselbe Algebra wie der mechanische Drehimpulsoperator** $\hat{\mathbf{L}}$:

$$
[S_x, S_y] = i\hbar S_z, \qquad
[S_y, S_z] = i\hbar S_x, \qquad
[S_z, S_x] = i\hbar S_y.
$$

Wir haben also gezeigt, dass die aus den Pauli-Matrizen konstruierten Operatoren $S_x$, $S_y$ und $S_z$ die mathematischen Eigenschaften eines Drehimpulses erfüllen. Dies bestätigt ausdrücklich die Interpretation des Spins als eine besondere Form des Drehimpulses. 


### Weitere nützliche Relationen zwischen Pauli-Matrizen

Nachdem wir den Kommutator der Pauli-Matrizen berechnet haben, bemerken wir auch die folgende Relation:

$$
\begin{aligned}
& \sigma_1 \sigma_2+\sigma_2 \sigma_1=0 \\
& \sigma_2 \sigma_3+\sigma_3 \sigma_2=0 \\
& \sigma_3 \sigma_1+\sigma_1 \sigma_3=0
\end{aligned}
$$

Multiplikation von Pauli-Matrizen ist also antikommutativ! Wir definieren den <b>Anti-Kommutator</b> als:

$$
\left\{\sigma_i, \sigma_j\right\} \equiv \sigma_i \sigma_j+\sigma_j \sigma_i=0 \text { wenn } i \neq j
$$

Die Vertauschungsrelationen für den Drehimpuls und Spin können mit Hilfe des <b>Levi-Civita-Symbols</b> in einer kompakten allgemeinen Form umschreiben, die viele spätere Berechnungen erleichtert. Wir definieren das Levi-Civita-Symbol $\varepsilon_{i j k}$ als

$$
\varepsilon_{i j k}=\left\{\begin{aligned}
1, & \text { wenn } i j k=123,231,312 \\
-1, & \text { wenn } i j k=213,132,321 \\
0, & \text { ansonsten }
\end{aligned}\right.
$$

Das Levi-Civita-Symbol hat insgesamt 27 Komponenten. Wenn mindestens zwei Indizes gleich sind, ist jedoch $\varepsilon_{i j k}=0$. Für gerade Permutationen aus der Menge $\{1,2,3\}$ ist $\varepsilon_{i j k}=+1$ und für ungerade Permutationen ist $\varepsilon_{i j k}=-1$.

Die Vertauschungsrelationen zwischen Pauli-Matrizen können mit Hilfe des Levi-Civitas Symbols in der folgenden kompakten Form geschrieben werden:

$$
\left[\sigma_i, \sigma_j\right]=2 i \sum_k \varepsilon_{i j k} \sigma_k
$$

oder noch kürzer unter Verwendung der Einsteinschen Summenkonvention (über doppelt auftretende Indizes innerhalb eines Produkts wird summiert):

$$
\left[\sigma_i, \sigma_j\right]=2 i \varepsilon_{i j k} \sigma_k \quad \text { (Summation über k implizit!) }
$$

Wir überprüfen an einem Beispiel, ob die obige Relation tatsächlich stimmt:

$$
\begin{aligned}
{\left[\sigma_1, \sigma_2\right] } & =2 i(\underbrace{\varepsilon_{121} \sigma_1}_{=0}+\underbrace{\varepsilon_{122} \sigma_2}_{=0}+\underbrace{\varepsilon_{123}}_{=1} \sigma_3) \\
& =2 i \sigma_3
\end{aligned}
$$

Eine weitere nützliche Beziehung erhalten wir, indem wir das Produkt der Pauli-Matrizen durch die Einführung des Levi-Civita-Symbols ausdrücken:

$$
\begin{aligned}
\sigma_i \sigma_j & =\frac{1}{2}\left(\sigma_i \sigma_j+\sigma_j \sigma_i\right)+\frac{1}{2}\left(\sigma_i \sigma_j-\sigma_j \sigma_i\right) \\
& =\frac{1}{2}\left\{\sigma_i, \sigma_j\right\}+\frac{1}{2}\left[\sigma_i, \sigma_j\right] \\
& =\frac{1}{2}\left(2 \delta_{i j} I\right)+\frac{1}{2} \cdot 2 i \sum_k \varepsilon_{i j k} \sigma_k
\end{aligned}
$$

Daraus folgt die Pauli-Identität:
$$
\sigma_i \sigma_j=\delta_{i j}+i \sum_k \varepsilon_{i j k} \sigma_k
$$
