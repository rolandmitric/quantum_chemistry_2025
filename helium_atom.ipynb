# Das Helium-Atom

Nachdem wir das Variationsprinzip als Werkzeug zur näherungsweisen Lösung komplexer quantenmechanischer Probleme eingeführt haben, wenden wir uns nun dem ersten System zu, das bereits die für quantenchemische Fragestellungen typische Komplexität aufweist: dem Heliumatom. Obwohl es lediglich zwei Elektronen enthält, erweist es sich als deutlich schwieriger zu behandeln als das Wasserstoffatom. Dieses Beispiel soll vor allem dazu dienen, zentrale Konzepte und Ideen der Quantenchemie zu motivieren – an einem noch überschaubaren System, bei dem ein Teil der Rechnungen „von Hand“ nachvollzogen werden kann. Viele der hier am Heliumatom entwickelten Ansätze werden wir anschließend auf größere Mehrelektronen-Atome und molekulare Systeme übertragen und verallgemeinern. 

## Der atomare Hamiltonoperator


Das Heliumatom besteht aus einem Kern mit Ladungszahl $Z = 2$ und zwei Elektronen.  
Im **nichtrelativistischen** Fall lautet der Hamiltonoperator in Atomaren Einheiten  
(a.u., also $\hbar = m_e = e = 4\pi\varepsilon_0 = 1$):

$$
H = \sum_{i=1}^2 \left(-\tfrac12 \nabla_i^2 - \frac{Z}{r_i}\right) + \frac{1}{r_{12}},
\qquad r_i = |\mathbf r_i|,\quad r_{12}=|\mathbf r_1-\mathbf r_2|.
$$

Die **exakte** Lösung der Schrödingergleichung ist in diesem Fall nicht möglich, weil der Term $1/r_{12}$ die beiden Elektronen koppelt.  
Als Ausgangspunkt für unsere Behandlung des He-Atoms lassen wir im ersten Schritt die Elektron–Elektron-Abstoßung vollständig **weg** und erhalten damit den folgenden vereinfachten Hamiltonian:

$$
H_0 = \sum_{i=1}^2 \left(-\tfrac12 \nabla_i^2 - \frac{Z}{r_i}\right).
$$

In diesem Hamiltonian sind beide Elektronen vollständig entkoppelt, und der Operator kann als Summe zweier Terme geschrieben werden, die jeweils nur von den Koordinaten des einen oder des anderen Elektrons abhängen.  
Solche Terme bezeichnet man in der Quantenchemie als **Ein-Elektron-Terme**. Es gilt also

$$
H_0 = h(1) + h(2),
$$

mit dem bekannten wasserstoffähnlichen Hamiltonian

$$
h = -\tfrac12 \nabla^2 - \frac{Z}{r}.
$$

<div style="background-color:#e6f0ff; border:2px solid #3366cc; border-radius:8px; padding:10px; margin-top:1em;">
<b>Verallgemeinerung: Ungekoppelter Hamiltonoperator für ein $N$-Elektronen-Atom</b><br><br>
Lässt man auch bei einem allgemeinen $N$-Elektronen-Atom die Elektron–Elektron-Wechselwirkung fort, so ergibt sich der ungekoppelte Hamiltonoperator zu
$$
H_0 = \sum_{i=1}^{N} \left(-\tfrac12 \nabla_i^2 - \frac{Z}{r_i}\right)
    = \sum_{i=1}^{N} h(i),
$$
wobei jeder Summand $h(i)$ den Ein-Elektron-Hamiltonian für das $i$-te Elektron beschreibt:
$$
h(i) = -\tfrac12 \nabla_i^2 - \frac{Z}{r_i}.
$$

## Separationsansatz (Hartree-Produkt)

Für den Grundzustand des He-Atoms wählen wir den **Produktansatz**

$$
\Psi(\mathbf r_1,\mathbf r_2) = \phi(\mathbf r_1)\,\phi(\mathbf r_2),
$$

wobei $\phi$ ein 1s-ähnliches Orbital am Kern ist.  
Setzt man diesen Ansatz in die zeitunabhängige Schrödingergleichung mit dem vereinfachten Hamiltonoperator

$$
H_0 = h(1) + h(2), \qquad h = -\tfrac12 \nabla^2 - \frac{Z}{r},
$$

ein, erhält man

$$
H_0\,\Psi(\mathbf r_1,\mathbf r_2)
= [\,h(1) + h(2)\,]\,\phi(\mathbf r_1)\phi(\mathbf r_2)
= [\,h(1)\phi(\mathbf r_1)\,]\phi(\mathbf r_2) + \phi(\mathbf r_1)[\,h(2)\phi(\mathbf r_2)\,].
$$

Da beide Terme identisch sind, genügt es, das Ein-Elektron-Problem zu betrachten.  
Erfüllt das Orbital $\phi$ die Einteilchen-Gleichung

$$
h\,\phi = \varepsilon\,\phi,
$$

so folgt unmittelbar

$$
H_0\,\Psi = 2\varepsilon\,\Psi.
$$

Damit ist $\varepsilon$ die Eigenenergie eines Elektrons im wasserstoffähnlichen Potential eines Kerns der Ladungszahl $Z$.  
Für den 1s-Grundzustand gilt

$$
\varepsilon = E_{1s}(Z) = -\frac{Z^2}{2},
$$

und somit ergibt sich für das Gesamtsystem ohne Elektron–Elektron-Abstoßung

$$
E = 2\,E_{1s}(Z) = 2\!\left(-\frac{Z^2}{2}\right) = -Z^2.
$$

Für Helium ($Z = 2$) folgt daher $E = -4$ Hartree. Der **exakte nichtrelativistische Wert** aus hochpräzisen numerischen Rechnungen beträgt

$$
E_\text{exakt} \approx -2.9037\ \text{Hartree},
$$

und das **experimentell bestimmte Ionisationspotential** des Heliumgrundzustands entspricht einer Gesamtenergie von etwa

$$
E_\text{exp} \approx -2.9034\ \text{Hartree}.
$$

Die Differenz von rund **1.1 Hartree** zeigt deutlich, dass die einfache Näherung ohne Elektron–Elektron-Abstoßung das System **zu stark bindet**.  
Sie verdeutlicht damit den erheblichen Beitrag der Elektron–Elektron-Wechselwirkung zur realen Bindungsenergie des Heliumatoms. Wir lernen daraus, dass die Elektron-Elektron Wechselwirkung definitiv keine Vernachlässigbare Größe ist und die orbitalbasierte Produktansätze keine gute Näherungen für die exakte Wellenfunktion sind !



## Erwartungwert der Energie

Anstatt die Schrödingergleichung direkt zu lösen, können wir den Energieeigenwert auch über den **Erwartungswert des Hamiltonoperators** bestimmen. Für eine gegebene (nicht notwendigerweise exakte) Wellenfunktion $\Psi$ ist der zugehörige Energieerwartungswert definiert als

$$
E[\Psi] = \frac{\langle \Psi | H | \Psi \rangle}{\langle \Psi | \Psi \rangle}.
$$

Dieser Ausdruck bildet die Grundlage der Anwendung des **Variationsprinzips** auf die atomaren und molekularen Systeme.  

Wenden wir dies auf das Heliumatom im Näherungshamiltonian $H_0 = h(1) + h(2)$ an, so ergibt sich für den Produktansatz

$$
\Psi(\mathbf r_1,\mathbf r_2) = \phi(\mathbf r_1)\,\phi(\mathbf r_2)
$$

der Erwartungswert

$$
E[\Psi] = \frac{\langle \phi(1)\phi(2) |\, h(1) + h(2)\, | \phi(1)\phi(2) \rangle}{\langle \phi(1)\phi(2) | \phi(1)\phi(2) \rangle}.
$$

Da $h(1)$ und $h(2)$ auf verschiedene Koordinaten wirken, kann der Ausdruck separiert werden:

$$
E[\Psi] = 
\frac{\langle \phi(1) | h(1) | \phi(1) \rangle}{\langle \phi(1) | \phi(1) \rangle}
+
\frac{\langle \phi(2) | h(2) | \phi(2) \rangle}{\langle \phi(2) | \phi(2) \rangle}
= 2\,\langle \phi | h | \phi \rangle = 2\,\varepsilon.
$$

Damit bestätigt der Erwartungswert formal das zuvor erhaltene Ergebnis:  
Die Gesamtenergie ist gleich der Summe der beiden Einteilchenenergien. Erst wenn die Elektron–Elektron-Wechselwirkung berücksichtigt wird, liefert das Variationsprinzip eine Möglichkeit, realistischere Energien durch geeignete Wahl der Wellenfunktion zu bestimmen.

## Symbolische Auswertung mit SymPy

Für interessierte Studierende lässt sich die Energie des Heliumatoms mit vereinfachten Hamiltonian (ohne Elektron–Elektron-Abstoßung) auch **symbolisch** berechnen.  
Wir wählen als Näherung für die räumliche Wellenfunktion das Produkt zweier exponentiell abfallender 1s-Funktionen

$$
\Psi(\mathbf r_1,\mathbf r_2) = e^{-Z(r_1 + r_2)}.
$$

Da wir s-Zustände betrachten,können die Winkelintegrationen für beide Elektronen sofort durchgeführt werden:  
Das **Volumenelement** separiert zu

$$
\mathrm d^3 r = r^2 \sin\theta\, \mathrm dr\, \mathrm d\theta\, \mathrm d\phi,
\qquad \int \mathrm d\Omega = 4\pi,
$$

und für zwei Elektronen ergibt sich nach Winkelintegration das radiale Maß

$$
(4\pi)^2 \int_0^\infty\!\!\int_0^\infty r_1^2 r_2^2\, f(r_1,r_2)\,\mathrm dr_1\,\mathrm dr_2,
$$

wobei die konstanten Faktoren $(4\pi)^2$ in der Energieratio herauskürzen und daher ohne Einfluss bleiben.

Der **Laplace-Operator** reduziert sich für sphärisch symmetrische Funktionen auf seine radiale Form

$$
\nabla^2 f(r) = \frac{\mathrm d^2 f}{\mathrm d r^2} + \frac{2}{r}\frac{\mathrm d f}{\mathrm d r},
$$

sodass der nichtrelativistische Hamiltonoperator

$$
H_0 = -\tfrac12 (\nabla_1^2 + \nabla_2^2) - \frac{Z}{r_1} - \frac{Z}{r_2}
$$

im Code direkt mit radialen Ableitungen implementiert werden kann.

Das folgende **SymPy-Skript** führt dies aus, integriert über beide Elektronenradien analytisch und berechnet schließlich den Erwartungswert der Energie symbolisch:

```python
import sympy as sp
from sympy import oo
r1, r2, Z = sp.symbols('r1 r2 Z', positive=True, real=True)

def laplacian_radial(psi, r):
    return psi.diff(r, 2) + (2/r)*psi.diff(r)

def T(psi):
    return -sp.S.Half * (laplacian_radial(psi, r1) + laplacian_radial(psi, r2))

def V_ne(psi):
    return -(Z/r1 + Z/r2) * psi

def H0(psi):
    return T(psi) + V_ne(psi)

psi0 = sp.exp(-Z*(r1 + r2))  

def integrate_6d(f):
    integrand = (4*sp.pi)**2 * r1**2 * r2**2 * f
    I_r2 = sp.integrate(integrand, (r2, 0, sp.oo))
    I_r1 = sp.integrate(I_r2, (r1, 0, sp.oo))
    return sp.simplify(I_r1)

def energy(psi, H):
    num = integrate_6d(psi * H(psi))
    den = integrate_6d(psi * psi)
    return sp.simplify(num/den)

E_Z = energy(psi0, H0)            
```

```python
E_He = sp.simplify(E_Z.subs(Z, 2))  # Helium-Ergebnis

print(f"Energie = {E_He} Hartree")
```

## Wie groß ist die Elektron–Elektron-Wechselwirkung? Eine erste Abschätzung.

Nachdem wir die Grundenergie des Heliumatoms ohne Elektron–Elektron-Abstoßung berechnet haben, wollen wir nun abschätzen, **wie groß der Beitrag der Elektron–Elektron-Wechselwirkung tatsächlich ist**.  
Dazu berechnen wir den Erwartungswert des Coulomb-Terms $1/r_{12}$ mit unserer Produktwellenfunktion

$$
\Psi(\mathbf r_1,\mathbf r_2) = e^{-Z(r_1 + r_2)}.
$$

Der Erwartungswert der Elektron–Elektron-Wechselwirkung lautet

$$
\langle V_{ee} \rangle 
= \big\langle \Psi \big| \frac{1}{r_{12}} \big| \Psi \big\rangle
= \int \Psi^*(\mathbf r_1,\mathbf r_2)\,
\frac{1}{|\mathbf r_1 - \mathbf r_2|}\,
\Psi(\mathbf r_1,\mathbf r_2)\,
\mathrm d^3 r_1\,\mathrm d^3 r_2.
$$

Statt über die sechs kartesischen Koordinaten $(\mathbf r_1,\mathbf r_2)$ zu integrieren, ist es für s-Zustände deutlich einfacher, auf **Koordinaten der Abstände** umzusteigen:  
$r_1 = |\mathbf r_1|$, $r_2 = |\mathbf r_2|$ und $r_{12} = |\mathbf r_1 - \mathbf r_2|$.  
Zwischen ihnen gilt der **Kosinussatz**

$$
r_{12}^2 = r_1^2 + r_2^2 - 2r_1r_2\cos\theta_{12},
$$

woraus sich die **Dreiecksgrenzen**

$$
|r_1 - r_2| \le r_{12} \le r_1 + r_2
$$

ergeben. Nach der Winkelintegration reduziert sich das Maß auf das folgende Volumenelement:

$$
\mathrm d^3r_1\,\mathrm d^3r_2 = 8\pi^2\, r_1\,r_2\,r_{12}\,\mathrm dr_1\,\mathrm dr_2\,\mathrm dr_{12}.
$$

Damit lautet das zu berechnende Integral:

$$
\langle V_{ee} \rangle =
8\pi^2 \int_0^\infty \!\!\int_0^\infty \!\!\int_{|r_1-r_2|}^{r_1+r_2}
r_1 r_2 r_{12}\, \frac{e^{-2Z(r_1+r_2)}}{r_{12}}\,
\mathrm d r_{12}\, \mathrm d r_2\, \mathrm d r_1.
$$

Zum Vergleich benötigen wir zusätzlich das Normintegral

$$
\langle \Psi|\Psi\rangle =
8\pi^2 \int_0^\infty \!\!\int_0^\infty \!\!\int_{|r_1-r_2|}^{r_1+r_2}
r_1 r_2 r_{12}\, e^{-2Z(r_1+r_2)}\,
\mathrm d r_{12}\, \mathrm d r_2\, \mathrm d r_1.
$$

Beide Integrale können symbolisch mit **SymPy** ausgewertet werden.  
Der folgende Code führt diese Berechnung exakt aus:

```python
import sympy as sp
r1, r2, r12, Z = sp.symbols('r1 r2 r12 Z', positive=True, real=True)

# Produktansatz der Wellenfunktion
psi0 = sp.exp(-Z*(r1 + r2))

# Winkelintegriertes Volumenelement in (r1, r2, r12)
dV = 8*sp.pi**2 * r1 * r2 * r12

# Normintegral S = <Psi|Psi>
S = sp.integrate(
    dV * psi0**2,
    (r12, sp.Abs(r1 - r2), r1 + r2),
    (r2, 0, sp.oo),
    (r1, 0, sp.oo)
)

# Erwartungswert <V_ee> = <Psi|1/r12|Psi>
Vee = sp.integrate(
    dV * psi0**2 * (1/r12),
    (r12, sp.Abs(r1 - r2), r1 + r2),
    (r2, 0, sp.oo),
    (r1, 0, sp.oo)
)

# Verhältnis <V_ee>/<Psi|Psi>
result = sp.simplify(Vee/S)
print(result)
```

Die symbolische Berechnung ergibt schließlich

$$
\langle V_{ee}\rangle = \frac{5Z}{8}.
$$

Für Helium ($Z=2$) beträgt der Erwartungswert der Elektron–Elektron-Abstoßung also  
$\langle V_{ee}\rangle = 1{,}25$ Hartree. Addiert man diesen Beitrag zu der Energie zweier unabhängiger Elektrone ($E_0 = -4$ Hartree), erhält man für die Gesamtenergie

$$
E \approx -4 + 1{,}25 = -2{,}75\ \text{Hartree},
$$

was bereits eine vernünftige Näherung des exakten Grundzustandswerts  
$E_\text{exakt} = -2{,}9037$ Hartree darstellt.Damit wird deutlich, dass die Elektron–Elektron-Wechselwirkung einen Beitrag von etwa **1 Hartree**, also rund einem Drittel der gesamten Bindungsenergie des He-Atoms, liefert.
